#### 手順1.RMDP問題をラグランジュ定式化するためのいくつかの準備
* f-DivergenceのUncertainty setについて、凸最適化問題を解きます。ラグランジュ形式にすることで、制約条件を含んだ1つの目的関数として表すことができます。

**目的関数**

$$
\inf _P \sum_{s \in S} P(s) V(s)
$$


**制約条件**

$$
D\left(P \| P^*\right) \leqq \rho
$$


$$
P \in \Delta(S)
$$


$$
P \ll P^*
$$

* 次に$r(s)=\frac{P(s)}{P^*(s)}$とおいて、目的関数と制約条件を書き直します。

**目的関数** 

$$\inf_{P} \sum P(s) V(s) \rightarrow \inf_{r} \sum r(s) P^*(s) V(s)$$

**制約条件1**

$$\sum_{s \in S} f(r(s)) P^*(s) \leqq \rho$$

* 導出
$$
\begin{aligned}
& D(p\|q) = \sum q(x)f(\frac{p(x)}{q(x)})なので\\
& q(x)=P^*(s), p(x)=P(s), \frac{p(x)}{q(x)}=r(s)とすると \\
& D\left(P \| P^*\right)=\sum_s P^*(s) \cdot f(r(s)) \leqq P
\end{aligned}
$$

**制約条件2**

$$\sum_{s \in S} r(s) P^*(s)=1$$

* $P(s) = P*(s)r(s)$であり、$P \in \delta(S)$なので、$P$は確率分布である。つまり、$\sum P(s)= 1$である。
* $P(s)$を$P*(s)r(s)$で書き換えると、$\sum_{s}r(s)P*(s)=1$となる。

**制約条件3**

$$
r(s) \geq 0 \text { for all } s \in S
$$

#### 手順2.ラグランジュ定式化(LemmaB.1)
 $r \geqslant 0$、$\lambda \geqslant 0$、$\eta \in \mathbb{R}$ を用いて、いかのようにラグランジュ定式化を行います。右辺の2項以降は制約条件における不等式の各要素を左辺にもってきています。
$$
\left.L(r, \lambda, \eta)=\sum_{s \in \mathcal{S}} r(s) V(s) P^{*}(s)+\lambda\left(\sum_{s \in \mathcal{S}} f(r(s))\right) P^{*}(s)-\rho\right)-\eta\left(\sum_{s \in \mathcal{S}} r(s) P^{*}(s)-1\right)
$$

* 式の置き換え

$$
\begin{aligned}
& L(r, \lambda, \eta) \\
& =\sum_{s \in S} r(s) V(s) P^*(s)+\lambda \sum_{s \in S} f(r(s)) P^*(s)-\lambda \rho-\eta \sum_{s \in s} r(s) P^*(s)+\eta \\
& =\sum_{s \in S} r(s) V(s) P^*(s)+\lambda \sum_{s \in S} f(r(s)) P^*(s)-\eta \sum_{s\in S} r(s) P^*(s)-\lambda \rho+\eta \\
& =\sum_{s\in S} P^*(s)(r(s) V(s)+\lambda f(r(s))-\eta r(s))-\lambda \rho+\eta \\
& =\lambda \sum_{s \in S} \frac{P^*(S)(-\eta r(s)+r(s) V(S)+\lambda f(r(s)))}{\lambda}-\lambda \rho+\eta \\
& =\lambda \sum_{s \in S} P^*(s)\left(-\frac{\eta r(s)}{\lambda}+\frac{r(s)V(s)}{\lambda}+f(r(s))\right)-\lambda \rho+\eta\\
& =\lambda \sum_{s \in S} P^*(s)\left(f(r(s))-\left(\frac{\eta -V(s)}{\lambda}\right)r(s)\right)-\lambda \rho+\eta
\end{aligned}
$$

* $\inf$をとる

$$
\inf_{r \geq 0} L(r, \lambda, \eta) = \lambda \inf_{s \in S}\sum_{s \in S} P^*(s)\left(f(r(s))-\left(\frac{\eta -V(s)}{\lambda}\right)r(s)\right)-\lambda \rho+\eta
$$

* $r$についての$\inf$は$s= r(s)$より、$s$についての$\inf$と同義であるため、

$$
\inf_{r \geq 0} L(r, \lambda, \eta) = \lambda \sum_{s \in S} P^*(s)\inf_{s \in S}\left(f(r(s))-\left(\frac{\eta -V(s)}{\lambda}\right)r(s)\right)-\lambda \rho+\eta
$$

* $f^{*}(t)=-\inf _{s \geqslant 0}(f(s)-s t)$とすると、$f(s)=f(r(s)), s=r(s), t=\frac{\eta-V(s)}{\lambda}$と置き換えることができるので、


$$
\inf_{r \geq 0} L(r, \lambda, \eta) = -\lambda \sum_{s \in S} P^*(s) f^*\left(\frac{\eta-V(s)}{\lambda}\right)-\lambda \rho+\eta
$$
