### 凸最適化の基礎


#### 目次
* いくつかの基礎事項
* 


#### 正則 
* $\boldsymbol{A}$が正則であるための条件
$$\boldsymbol{AB} = \boldsymbol{BA} = \boldsymbol{I}$$

* $\boldsymbol{A}$が正則であるための必要十分条件
$$\operatorname{det} \boldsymbol{A} \neq 0$$
* $\boldsymbol{B}$ : **逆行列**


#### 行列式の性質
* $\operatorname{det} \boldsymbol{A} = \operatorname{det} \boldsymbol{AB}$
* $\operatorname{det} \boldsymbol{A}^{-1} = 1 / \operatorname{det} \boldsymbol{A}$

#### トレース$\operatorname{tr} \boldsymbol{A}$
* $a_{11} + a_{22} + \cdots + a_{nn}$
  * 行列の$n$行$n$列の要素をすべて足したやつです。あとで結構出てきます。  

#### クラメールの公式
* 正則な行列を持つ$\boldsymbol{Ax} = \boldsymbol{b}$の解 $\boldsymbol{x} = \boldsymbol{A}^{-1} \boldsymbol{b}$ の第 $i$ 成分
* $x_i = \operatorname{det}[\boldsymbol{A}|\boldsymbol{b}|i] / \operatorname{det} \boldsymbol{A}$
* 正方行列 $\boldsymbol{A}, \boldsymbol{B} \in \mathbb{R}^{n \times n}$ に対して、次式が成立する。

$$
\operatorname{tr}\left[\boldsymbol{A}^{-1} \boldsymbol{B}\right]=\sum_{i=1}^{n} \frac{\operatorname{det}\left[\boldsymbol{A}\left|\boldsymbol{B}^{[i]}\right| i\right]}{\operatorname{det} \boldsymbol{A}}
$$


#### 固有値と固有ベクトル
$$
\boldsymbol{Ax} = \lambda \boldsymbol{x}
$$
* $\lambda$ : $\boldsymbol{A}$ の固有値

* $\boldsymbol{x}$ : $\lambda$ に対する $\boldsymbol{A}$ の固有ベクトル

* $\lambda$は特性方程式$
\operatorname{det}(\lambda \boldsymbol{I} - \boldsymbol{A}) = 0
$の根(方程式を満たす解)である。

* $\lambda$について以下が成り立つ。
$$
\lambda_1 + \lambda_2 + \cdots + \lambda_n = \operatorname{tr} \boldsymbol{A}, \quad \lambda_1 \lambda_2 \cdots \lambda_n = \operatorname{det} \boldsymbol{A}
$$

#### 直交行列

$n \times n$ 行列 $\boldsymbol{Q}$ が以下の条件を満たすとき、直交行列と呼ぶ。

$$
\boldsymbol{QQ}^\top = \boldsymbol{Q}^\top \boldsymbol{Q} = \boldsymbol{I}
$$

#### 実対称行列の対角化

$n$ 次の実対称行列 $\boldsymbol{A} \in \mathcal{S}^n$ は常に**直交行列 $\boldsymbol{Q}$ を用いて対角化可能**である。

$$
\boldsymbol{Q}^\top \boldsymbol{AQ} = \operatorname{diag}[\lambda_1, \ldots, \lambda_n] 
$$

ここで、

* $\lambda_1, \ldots, \lambda_n$ は $\boldsymbol{A}$ の固有値（すべて実数）
* $\operatorname{diag}[\lambda_1, \ldots, \lambda_n]$ は対角成分に $\lambda_1, \ldots, \lambda_n$ を持つ対角行列

#### 半正定値行列と正定値行列

実対称行列 $\boldsymbol{A} \in \mathcal{S}^n$ について、

* 任意のベクトル $\boldsymbol{x} \in \mathbb{R}^n$ に対して $\langle \boldsymbol{x}, \boldsymbol{Ax} \rangle \geqq 0$ を満たすとき、 **半正定値行列**と呼び、 $\boldsymbol{A} \succeq \boldsymbol{O}$ と書く。
* 任意の非ゼロベクトル $\boldsymbol{x} \in \mathbb{R}^n$ に対して $\langle \boldsymbol{x}, \boldsymbol{Ax} \rangle > 0$ を満たすとき、 **正定値行列**と呼び、 $\boldsymbol{A} \succ \boldsymbol{O}$ と書く。

$\boldsymbol{A}$ が **半正定値行列 (正定値行列) であることと、$\boldsymbol{A}$ のすべての固有値が非負 (正) であることは同値**である。

#### 半正定値行列の平方根

半正定値行列 $\boldsymbol{A}$ は、対角化を用いて平方根 $\boldsymbol{A}^{\frac{1}{2}}$ を定義できる。

$$
\boldsymbol{A}^{\frac{1}{2}} = \boldsymbol{Q} \operatorname{diag}[\lambda_i^{\frac{1}{2}}] \boldsymbol{Q}^\top
$$

このとき、

* $\boldsymbol{A}^{\frac{1}{2}}$ も対称かつ半正定値行列
*  $\boldsymbol{A} = \left( \boldsymbol{A}^{\frac{1}{2}} \right)^2$ 

#### 半正定値行列のトレース

二つの半正定値行列 $\boldsymbol{A}, \boldsymbol{B} \in \mathcal{S}^n$ に対して、

$$
\operatorname{tr}[\boldsymbol{AB}] = \operatorname{tr}\left[\boldsymbol{B}^{\frac{1}{2}} \boldsymbol{AB}^{\frac{1}{2}}\right] \geqq 0
$$

が成り立つ。






#### アフィン集合

```python
import numpy as np
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display

# 2点を定義
A = np.array([1, 2])
B = np.array([4, 1])

# スライダー
theta_slider = widgets.FloatSlider(
    value=0.5, min=-1, max=2, step=0.01, description='θ:',
    continuous_update=True
)

# 描画関数
def update_plot(theta):
    affine_point = theta * A + (1 - theta) * B
    
    fig = go.Figure()
    # 点A, B
    fig.add_trace(go.Scatter(x=[A[0]], y=[A[1]], mode='markers+text', name='A',
                             text=["A"], textposition='top center'))
    fig.add_trace(go.Scatter(x=[B[0]], y=[B[1]], mode='markers+text', name='B',
                             text=["B"], textposition='top center'))
    
    # AとBを結ぶ線分（アフィン直線）
    line_t = np.linspace(-1, 2, 100)
    points = np.outer(line_t, A) + np.outer(1 - line_t, B)
    fig.add_trace(go.Scatter(x=points[:, 0], y=points[:, 1], mode='lines', name='Affine line'))

    # 現在のアフィン結合点
    fig.add_trace(go.Scatter(x=[affine_point[0]], y=[affine_point[1]],
                             mode='markers+text', name='Affine Combo',
                             marker=dict(size=10, color='red'),
                             text=[f'θ={theta:.2f}'], textposition='bottom right'))
    
    fig.update_layout(title="2Dアフィン集合の可視化",
                      xaxis_title="X",
                      yaxis_title="Y",
                      width=600,
                      height=600,
                      showlegend=False)
    fig.show()

# スライダーに連動
widgets.interact(update_plot, theta=theta_slider)
```

```python
import numpy as np
import plotly.graph_objects as go

# 3点を定義（アフィン集合の生成元）
A = np.array([0, 0, 0])
B = np.array([1, 0, 1])
C = np.array([0, 1, 2])

# アフィン結合の点群を生成
theta1 = np.linspace(0, 1, 30)
theta2 = np.linspace(0, 1, 30)
theta1, theta2 = np.meshgrid(theta1, theta2)
theta3 = 1 - theta1 - theta2

# 有効な範囲（theta3 >= 0）だけ使用
valid = theta3 >= 0
theta1 = theta1[valid]
theta2 = theta2[valid]
theta3 = theta3[valid]

# アフィン結合点
X = theta1 * A[0] + theta2 * B[0] + theta3 * C[0]
Y = theta1 * A[1] + theta2 * B[1] + theta3 * C[1]
Z = theta1 * A[2] + theta2 * B[2] + theta3 * C[2]

# 描画
fig = go.Figure()

# アフィン集合（平面状の点群）
fig.add_trace(go.Mesh3d(
    x=X, y=Y, z=Z,
    opacity=0.5,
    color='lightblue',
    alphahull=0,
    name='Affine Set'
))

# 点A, B, C
for point, name in zip([A, B, C], ['A', 'B', 'C']):
    fig.add_trace(go.Scatter3d(
        x=[point[0]], y=[point[1]], z=[point[2]],
        mode='markers+text',
        marker=dict(size=5),
        text=[name],
        textposition='top center'
    ))

fig.update_layout(
    title="3Dアフィン集合の可視化",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    width=700,
    height=600
)

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# 点A, B, C を定義
A = np.array([0, 0, 0])
B = np.array([1, 0, 1])
C = np.array([0, 1, 2])

# スライダー（theta1, theta2）。theta3 = 1 - theta1 - theta2
theta1_slider = widgets.FloatSlider(value=0.3, min=0, max=1, step=0.01, description='θ₁')
theta2_slider = widgets.FloatSlider(value=0.3, min=0, max=1, step=0.01, description='θ₂')

# 更新関数
def update(theta1, theta2):
    theta3 = 1 - theta1 - theta2
    if theta3 < 0:
        print("θ₁ + θ₂ ≤ 1 にしてください")
        return

    P = theta1 * A + theta2 * B + theta3 * C  # アフィン結合点

    fig = go.Figure()

    # 三角形（アフィン平面）
    triangle_x = [A[0], B[0], C[0], A[0]]
    triangle_y = [A[1], B[1], C[1], A[1]]
    triangle_z = [A[2], B[2], C[2], A[2]]
    fig.add_trace(go.Scatter3d(
        x=triangle_x, y=triangle_y, z=triangle_z,
        mode='lines', name='Affine Plane', line=dict(color='lightblue', width=4)
    ))

    # 結合点のプロット
    fig.add_trace(go.Scatter3d(
        x=[P[0]], y=[P[1]], z=[P[2]],
        mode='markers+text',
        marker=dict(size=5, color='red'),
        text=["Affine Point"],
        textposition="top center"
    ))

    # 元の点も表示
    for pt, label in zip([A, B, C], ['A', 'B', 'C']):
        fig.add_trace(go.Scatter3d(
            x=[pt[0]], y=[pt[1]], z=[pt[2]],
            mode='markers+text',
            marker=dict(size=5),
            text=[label],
            textposition='bottom center'
        ))

    fig.update_layout(
        title="操作可能なアフィン結合（3D）",
        scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'),
        width=700, height=600,
        showlegend=False
    )
    fig.show()

# スライダー表示
widgets.interact(update, theta1=theta1_slider, theta2=theta2_slider)
```

### 超平面の移動

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# スライダー定義
a1_slider = widgets.FloatSlider(value=1.0, min=-5, max=5, step=0.1, description='a₁')
a2_slider = widgets.FloatSlider(value=1.0, min=-5, max=5, step=0.1, description='a₂')
b_slider  = widgets.FloatSlider(value=1.0, min=-10, max=10, step=0.1, description='b')

@widgets.interact(a1=a1_slider, a2=a2_slider, b=b_slider)
def plot_hyperplane(a1, a2, b):
    plt.figure(figsize=(6, 6))
    x = np.linspace(-10, 10, 400)

    if a2 != 0:
        y = (b - a1 * x) / a2
        plt.plot(x, y, label=f'{a1:.1f}x + {a2:.1f}y = {b:.1f}')
    elif a1 != 0:
        # a2 = 0 のときは x = b/a1 の垂直線
        x0 = b / a1
        plt.axvline(x=x0, label=f'x = {x0:.1f}')
    else:
        plt.text(0, 0, "a₁ = a₂ = 0 は定義できません", fontsize=12)
        plt.axis("off")
        return

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("2D hyperplane")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

### アフィン集合、部分空間(線形空間)、超平面

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 2次元部分空間（原点を通る平面）を定義: span{v1, v2}
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, 1])

# アフィン集合：部分空間を平行移動させた平面（z方向に+1）
offset = np.array([0, 0, 1])

# 超平面：法線ベクトルを定義し、a^T x = b の形に
a = np.array([1, 1, 1])
b = 4

# グリッド作成
u = np.linspace(-2, 2, 20)
v = np.linspace(-2, 2, 20)
U, V = np.meshgrid(u, v)

# 部分空間平面：z = u + v
X1 = U * v1[0] + V * v2[0]
Y1 = U * v1[1] + V * v2[1]
Z1 = U * v1[2] + V * v2[2]

# アフィン平面：z = u + v + 1
X2 = X1 + offset[0]
Y2 = Y1 + offset[1]
Z2 = Z1 + offset[2]

# 超平面：a1*x + a2*y + a3*z = b を z について解く
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X3, Y3 = np.meshgrid(x, y)
Z3 = (b - a[0]*X3 - a[1]*Y3) / a[2]

# プロット
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 部分空間
ax.plot_surface(X1, Y1, Z1, alpha=0.5, color='skyblue', label='Subspace')
ax.text(0, 0, 0, "origin", color='black')

# アフィン集合（平行移動された部分空間）
ax.plot_surface(X2, Y2, Z2, alpha=0.5, color='lightgreen')

# 超平面
ax.plot_surface(X3, Y3, Z3, alpha=0.4, color='salmon')

# 装飾
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.tight_layout()
plt.show()
```

### 互換

```python
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# スライダーでベクトル v の成分を操作
x_slider = widgets.FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.5, description='x')
y_slider = widgets.FloatSlider(value=2.0, min=-5.0, max=5.0, step=0.5, description='y')
z_slider = widgets.FloatSlider(value=3.0, min=-5.0, max=5.0, step=0.5, description='z')

def plot_transposition(x, y, z):
    v = np.array([x, y, z])
    v_swapped = np.array([y, x, z])  # (1 2) 互換: x <-> y

    fig = go.Figure()

    # 原点
    origin = np.array([0, 0, 0])

    # 元のベクトル
    fig.add_trace(go.Scatter3d(
        x=[origin[0], v[0]],
        y=[origin[1], v[1]],
        z=[origin[2], v[2]],
        mode='lines+markers+text',
        name='v',
        line=dict(color='blue', width=6),
        marker=dict(size=4),
        text=["", "v"],
        textposition="top center"
    ))

    # 互換後のベクトル
    fig.add_trace(go.Scatter3d(
        x=[origin[0], v_swapped[0]],
        y=[origin[1], v_swapped[1]],
        z=[origin[2], v_swapped[2]],
        mode='lines+markers+text',
        name='σ(v)',
        line=dict(color='red', width=6),
        marker=dict(size=4),
        text=["", "σ(v)"],
        textposition="top center"
    ))

    # x = y 平面の表示（zは固定）
    plane_x, plane_y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
    plane_z = np.zeros_like(plane_x) + z  # z固定高さ
    fig.add_trace(go.Surface(
        x=plane_x,
        y=plane_y,
        z=plane_z,
        showscale=False,
        opacity=0.2,
        colorscale='Greys',
        name='x=y plane'
    ))

    fig.update_layout(
        title="3D 互換 (x ↔ y) の操作可視化",
        scene=dict(
            xaxis_title="X",
            yaxis_title="Y",
            zaxis_title="Z",
            aspectmode='cube'
        ),
        width=800,
        height=600
    )
    fig.show()

ui = widgets.VBox([x_slider, y_slider, z_slider])
out = widgets.interactive_output(plot_transposition, {
    'x': x_slider,
    'y': y_slider,
    'z': z_slider
})

display(ui, out)
```

### 凸集合の可視化と操作


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# 凸集合（閉区間を含む三角形）
convex_set = np.array([
    [1, 1],
    [4, 1],
    [2.5, 4]
])

def project_point_onto_convex_set(px, py):
    from scipy.optimize import minimize

    p = np.array([px, py])

    # 凸集合の各辺に対して射影をとって最短距離の点を選ぶ
    def dist_to_segment(A, B):
        AB = B - A
        t = np.dot(p - A, AB) / np.dot(AB, AB)
        t = np.clip(t, 0, 1)
        return A + t * AB

    projections = []
    for i in range(len(convex_set)):
        A = convex_set[i]
        B = convex_set[(i + 1) % len(convex_set)]
        proj = dist_to_segment(A, B)
        projections.append(proj)

    # 最も近い射影点を選ぶ
    dists = [np.linalg.norm(p - proj) for proj in projections]
    closest_proj = projections[np.argmin(dists)]

    # 描画
    fig, ax = plt.subplots(figsize=(6, 6))
    triangle = plt.Polygon(convex_set, color='lightblue', alpha=0.4, label='Convex Set')
    ax.add_patch(triangle)

    ax.plot(*p, 'ro', label='Point P')
    ax.plot(*closest_proj, 'go', label='Projection')
    ax.plot([p[0], closest_proj[0]], [p[1], closest_proj[1]], 'k--', label='Projection Line')

    ax.set_xlim(0, 5)
    ax.set_ylim(0, 5)
    ax.set_aspect('equal')
    ax.legend()
    ax.grid(True)
    plt.show()

# スライダーで点を動かす
interact(project_point_onto_convex_set,
         px=FloatSlider(value=3, min=0, max=5, step=0.1, description='px'),
         py=FloatSlider(value=3, min=0, max=5, step=0.1, description='py'))
```
