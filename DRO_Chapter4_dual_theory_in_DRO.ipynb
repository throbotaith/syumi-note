# このページについて
このページでは、Distributionally Robust Optimization(https://arxiv.org/abs/2411.02549)の第4章の勉強内容をまとめていこうと思います。
## 内側最大化と双対理論

DROにおける内側最大化問題は次のように表されます。

$$
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] \quad (*) \tag{1}
$$

3章であったように、$(1)$は簡単に解けない場合があります。4章では、双対理論を使ってこの問題を解いていくことを考えます。

### 双対問題の構築
**1：問題を値関数の形で表現**

元の最適化問題の最適値を、ある**値関数 $h$** を使って表現することを考えます。$\mathcal{P}$ のパラメータを $u$ として元の最適化問題を $\sup -h(u)$ のような形に書き換えます。ここで、$h(u)$ は、$u$ をパラメータとして持つ**最小化**問題の最適値として定義します。
モーメント不確実性集合の場合、以下のように変換します。
$$
\begin{equation*}
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)]=\sup _{u \in \mathcal{F}} \sup _{\mathbb{P} \in \mathcal{P}_{f}(\mathcal{Z})}\left\{\mathbb{E}_{\mathbb{P}}[\ell(Z)]: \mathbb{E}_{\mathbb{P}}[f(Z)]=u\right\}=\sup _{u \in \mathcal{F}}-h(1, u) \tag{2}
\end{equation*}
$$
 $u$ を指定したときの最小化問題(の負の値)として $h(1, u)$ を考え、元の問題は $\sup_{u \in \mathcal{F}} -h(1, u)$ となっています。

**2：値関数の凸性を証明(補題 4.1)**

次に値関数が凸関数であることを示します。補題 4.1 (最適値関数の凸性) を用います。

*   **補題 4.1** $H(u, v)$ が $(u, v)$ に関して凸な関数であるとき、$h(u) = \inf_{v} H(u, v)$ で定義される関数 $h(u)$ も $u$ に関して凸になります。

元の問題は $\mathbb{P}$ に関する複雑な最適化問題ですが、$h(u)$ の最小化問題は$v$と $(u_0, u)$ に関して凸構造を持っていれば、$h(u)$ 自体の凸性が保証されるため、解きやすい問題になります。

**3：双対問題を構築し、強い双対性を得る(補題 4.2)**

$h(u)$ が凸関数であるならば、双対問題を構築することが可能です。ここでフェンシェル・モローの定理が登場します。

*   前提：凸共役と双共役
    *   凸関数 $h(u)$ に対して、その**凸共役** 関数 $h^*(\lambda)$ は $h^*(\lambda) = \sup_{u} \{ \lambda^{\top} u - h(u) \}$ で定義されます。 $h(u)$ を、双対空間の変数 $\lambda$ で表現したものです。
    *   $h^*$ の凸共役 $(h^*)^*(\cdot)$ を $h^{**}(\cdot)$ と書き、これを $h$ の双共役関数といいます。
*   **補題 4.2**
    1.  常に $h^{**}(u) \le h(u)$ が成り立つ(弱い双対性)。
    2.  $h$ が閉じた (closed)凸関数であれば、$h^{**}(u) = h(u)$ が成り立つ(少なくとも $h$ の定義域の内部 $\operatorname{rint}(\operatorname{dom}(h))$ では $h$ は連続であり、$h^{**}=h$ となる)。

$h(u)$ が凸関数であれば、一度双対空間に$h^*$として移し、元の空間に戻しても元の $h$ が復元される、ということです。

**4：双対問題の最終構築**

元の問題が $\sup -h(u)$ とします。フェンシェル・モローの定理から $h(u) \ge h^{**}(u)$ なので、$\sup -h(u) \le \sup -h^{**}(u)$ が成り立ちます。ここで、$h^{**}(u)$ の定義は、

$$
h^{**}(u) = \sup_{\lambda} \{ u^{\top} \lambda - h^*(\lambda) \}
$$

なので、元の問題は

$$
\sup -h(u) \le \sup \left\{ - \sup_{\lambda} \{ u^{\top} \lambda - h^*(\lambda) \} \right\} = \sup \left\{ \inf_{\lambda} \{ -u^{\top} \lambda + h^*(\lambda) \} \right\}
$$

となります。この右辺 $\inf_{\lambda} \{ -u^{\top} \lambda + h^*(\lambda) \}$ が、元の問題 $\sup -h(u)$ の**双対問題**として現れるわけです。

そして、フェンシェル・モローの定理の2は、元の問題と双対問題の最適値が一致する、すなわち**強い双対性**が成り立つための条件を示します。つまり閉じた凸関数かどうかが重要ですね。

それでは、モーメント不確実性集合(定理4.5)と$\phi$ダイバージェンス不確実性集合(定理4.14)について、双対問題を見ていきましょう。
