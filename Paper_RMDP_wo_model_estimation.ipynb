## 記法など
* $\Delta(\mathcal{X})$: 有限集合 $\mathcal{X}$ 上の確率分布の集合
* $Q \ll P$: 確率分布 $Q$ は $P$ に関して絶対連続
	* 任意の $x \in \mathcal{X}$ に対して，$P(x)=0$ ならば $Q(x)=0$ 
* $f: \Omega \rightarrow \mathbb{R} \cup\{-\infty,+\infty\}$:関数
* 距離(のようなもの)の定義
	* $f$-ダイバージェンス　ただし$f(1)=0$ を満たす凸関数 $f$に対して
	*  $D_{f}(Q \| P):=\sum_{x \in \mathcal{X}} f\left(\frac{Q(x)}{P(x)}\right) P(x)$　ただし$Q \ll P$ の場合
	*  $D_{f}(Q \| P):=+\infty$ 　ただし$Q \ll P$でない場合 
* $f^{*}(t):=\sup _{s \in \Omega}\{s t-f(s)\}$ : 関数 $f$ の凸共役
* $\sigma(X)$: 確率変数 $X$ によって生成される$\sigma$代数
* $\sigma\left(\left\{X_{t}\right\}_{t=1}^{T}\right):=\sigma\left(\bigcup_{t=1}^{T} \sigma\left(X_{t}\right)\right)$ :確率変数の列 $\left\{X_{t}\right\}_{t=1}^{T}$ に対して，$\left\{X_{t}\right\}_{t=1}^{T}$ によって生成される$\sigma$代数
*   **双対形式の目的関数:**
	* $f_{\eta, \lambda, V}^{\dagger}\left(s^{\prime}\right) = f^{*}\left(\frac{\eta-V\left(s^{\prime}\right)}{\lambda}\right)$，$V(s') = \max_{a'} Q(s', a')$
$$
J^{(s, a)}(\eta, V):=-\lambda \sum_{s^{\prime} \in \mathcal{S}} P^{*}\left(s^{\prime} \mid s, a\right) f_{\eta, \lambda, V}^{\dagger}\left(s^{\prime}\right)+\eta
$$

*   **双対目的関数の確率的評価 ($J_{t}^{(s, a)}\left(\eta, V ; s_{t}^{\prime}(s, a)\right)$):**
$$
J_{t}^{(s, a)}\left(\eta, V ; s_{t}^{\prime}(s, a)\right):=-\lambda f_{\eta, \lambda, V}^{\dagger}\left(s_{t}^{\prime}(s, a)\right)+\eta
$$

## 分布ロバスト最適化 (DRO)

**制約付きDRO問題**: 
$$
\mathcal{R}_{c}\left(P^{*}, V\right):=\inf _{P \in \Delta(\mathcal{X}), D_{f}\left(P \| P^{*}\right) \leq \rho} \sum_{x \in \mathcal{X}} P(x) V(x) \tag{1}
$$
* $\sum P V = E_P[V]$と同義です．$V$の期待値を最小化するような，$P$を選びます．ここでは，$P^*$から一定距離以下の確率分布しか採用しないように制約がかかっています．

**制約付きDROの双対形式**
$\lambda, \eta$ :双対変数，$f_{\eta, \lambda, V}^{\dagger}(x)=f^{*}\left(\frac{\eta-V(x)}{\lambda}\right)$
$$
\sup _{\lambda \geq 0, \eta \in \mathbb{R}}\left[-\lambda \sum P^{*}(x) f_{\eta, \lambda, V}^{\dagger}(x)-\lambda \rho+\eta\right] \tag{2}
$$
$f$には以下のように仮定を置きます．
**仮定2.1**:
$f(t)$は，
* $\mathbb{R}$上の凸関数
* $t<0$ のとき
	* $f(1)=0$ 
	*  $f(t):=+\infty$
* $\mathbb{R}_{+}$上で微分可能

(2)は非有界の勾配問題になってしまうため，制約をペナルティに置き換えて制約なし問題として解くことを考えます．([この論文読む.](https://arxiv.org/abs/1610.03425))
**ペナルティDRO問題**: 
$$
\mathcal{R}_{p}\left(P^{*}, V\right):=\inf _{P \in \Delta(\mathcal{X})} \sum_{x \in \mathcal{X}} P(x) V(x)+\lambda D_{f}\left(P \| P^{*}\right) \tag{3}
$$
**ペナルティDROの双対形式**:
$$
\sup _{\eta \in \mathbb{R}}\left[-\lambda \sum P^{*}(x) f_{\eta, \lambda, V}^{\dagger}(x)+\eta\right] \tag{4}
$$
 
## ロバストマルコフ決定過程 (RMDP)
ほとんど通常のRMDPの定義と同じですが，制約がついているところが違います．

### RMDPの定義
* $\left\langle\mathcal{S}, \mathcal{A}, P^{*}, R, \gamma\right\rangle$ (状態空間, 行動空間, 基準遷移関数, 報酬関数, 割引率)
* **($s, a$)-rectangular不確実性集合**: $\mathcal{P}:=\otimes_{s, a \in \mathcal{S} \times \mathcal{A}} \mathcal{P}_{s, a}(\rho)$
* **制約付きロバスト価値関数**:
$$
V_{\text {rob }, \mathrm{c}}^{\pi}(s):=\inf _{P \in \mathcal{P}} V_{P}^{\pi}(s)
$$
* **制約付き最適ロバスト価値関数**:
$$
V_{\text {rob,c }}^{*}:=\max _{\pi} V_{\text {rob,c }}^{\pi}
$$
* **制約付きロバストベルマン作用素**:
$$
\mathcal{T}_{\text {rob }, \mathrm{c}} V(s):=\max _{a \in \mathcal{A}}\left(R(s, a)+\gamma \inf _{P(\cdot \mid s, a) \in \mathcal{P}_{s, a}(\rho)} \sum_{s^{\prime} \in \mathcal{S}} P\left(s^{\prime} \mid s, a\right) V\left(s^{\prime}\right)\right)\tag{5}
$$
* **制約付きロバストベルマン作用素の双対形式**:
$$
\begin{aligned}
&\mathcal{T}_{\text {rob }, \mathrm{c}} V(s) \\&:=\max _{a \in \mathcal{A}}\left(R(s, a)+\gamma \sup _{\lambda \geq 0, \eta \in \mathbb{R}}\left[-\lambda \rho+\eta-\lambda \sum_{s^{\prime} \in \mathcal{S}} P^{*}\left(s^{\prime} \mid s, a\right) f_{\eta, \lambda, V}^{\dagger}\left(s^{\prime}\right)\right]\right)(6)
\end{aligned}
$$
[tips]IyengarのロバストDPの論文では，最適ロバスト価値関数は最適ベルマン方程式$V_{\text {rob }, \mathrm{c}}^{*}=\mathcal{T}_{\text {rob }, \mathrm{c}} V_{\text {rob,c }}^{*}$ を満たします．

### $P^*$ の推定：
真の$P^*$が分からない場合，データセットから$\widehat{P}$を推定できます．それを中心に不確実性集合を以下のように定義します．
$$
\widehat{\mathcal{P}}_{s, a}(\rho):=\left\{P(\cdot \mid s, a) \in \Delta(\mathcal{S}) \mid D_{f}(P(\cdot \mid s, a) \| \widehat{P}(\cdot \mid s, a)) \leq \rho\right\},
$$
$\widehat{\mathcal{P}}_{s, a}(\rho)$ を用いた場合の制約付きロバストベルマン作用素は以下のように定義されます．
$$
\widehat{\mathcal{T}}_{\text {rob }, \mathrm{c}} V(s):=\max _{a \in \mathcal{A}}\left(R(s, a)+\gamma \inf _{P \in \widehat{\mathcal{P}}_{s, a}(\rho)} \sum_{s^{\prime} \in \mathcal{S}} P\left(s^{\prime} \mid s, a\right) V\left(s^{\prime}\right)\right) .
$$

### 解きやすい形式に変更
(6)を確率的勾配法で解くと，非有界勾配問題が発生するため，DROの場合のように，ペナルティ付きのロバスト価値関数を考えます．

* **ペナルティロバスト価値関数:**
$$
\begin{aligned}
V_{\mathrm{rob},\mathrm{p}}^{\pi}(s)&:= \\
&\inf _{P \in \Delta\left(\mathcal{S})^{|\mathcal{S}||\mathcal{A}|}\right.} \mathbb{E}_{P, \pi}\left[\sum_{t=0}^{\infty} \gamma^{t}\left(R\left(s_{t}, a_{t}\right)+\lambda \gamma D_{f}\left(P\left(\cdot \mid s_{t}, a_{t}\right) \| P^{*}\left(\cdot \mid s_{t}, a_{t}\right)\right)\right) \mid s_{0}=s\right](7)
\end{aligned}
$$

*   **ペナルティロバストベルマン作用素:**
$$
\begin{aligned}
\mathcal{T}_{\text {rob }, \mathrm{p}} V(s)&:=\\
&\max _{a \in \mathcal{A}}\left(R(s, a)+\gamma \inf _{P(\cdot \mid s, a) \in \Delta(\mathcal{S})}\left[\sum_{s^{\prime} \in \mathcal{S}} P\left(s^{\prime} \mid s, a\right) V\left(s^{\prime}\right)+\lambda D_{f}\left(P(\cdot \mid s, a) \| P^{*}(\cdot \mid s, a)\right)\right]\right)(8)
\end{aligned}
$$

* **ペナルティロバストベルマン作用素の双対形式**:
$$
\mathcal{T}_{\text {rob }, \mathrm{p}} V(s):=\max _{a \in \mathcal{A}}\left(R(s, a)+\gamma \sup _{\eta \in \mathbb{R}}\left[-\lambda \sum_{s^{\prime} \in \mathcal{S}} P^{*}\left(s^{\prime} \mid s, a\right) f_{\eta, \lambda, V}^{\dagger}(x)+\eta\right]\right) .
$$

*   **ペナルティロバストベルマン行動作用素 :**
$$
\mathcal{T}_{\text {rob }, \mathrm{p}} Q(s, a):=R(s, a)+\gamma \sup _{\eta \in \mathbb{R}}\left[-\lambda \mathbb{E}_{s^{\prime} \sim P_{s, a}^{*}} f^{*}\left(\frac{\eta-\max _{a^{\prime}} Q\left(s^{\prime}, a^{\prime}\right)}{\lambda}\right)+\eta\right](9)
$$


## 定理など
## 補題
### 補題A.1

確率測度 $P$ を$\mathcal{S}$ 上の確率分布とし，$V \in [0, \frac{1}{1-\gamma}]$ を価値関数とします．$f(s)=(s-1)^2$ ( $s \ge 0$ の場合) のとき，ペナルティ付きDRO問題
$$
\mathcal{R}_{p}(P, V):=\inf _{Q \in \Delta(\mathcal{S})} \left( \sum_{s' \in \mathcal{S}} Q(s') V(s')+\lambda D_{f}(Q \| P) \right)
$$
の双対問題
$$
\mathcal{R}_{p}(P, V) = \sup _{\eta \in \mathbb{R}} \left[-\lambda \mathbb{E}_{s \sim P} f^{*}\left(\frac{\eta-V(s)}{\lambda}\right)+\eta\right]
$$
を考えます．このとき，最適双対変数 $\eta^*$ は区間 $\Theta=\left[-\lambda, \frac{2}{1-\gamma}+2 \lambda\right]$ 内に存在します．


## 命題
### 命題 3.1

$\mathcal{T}_{\text {rob }, p}$ は価値関数の空間 $\mathcal{V} = [0, 1/(1-\gamma)]^{|\mathcal{S}|}$ 上の $\gamma$-縮小作用素である．したがって，一意な固定点 $V_{\text {rob }, p}^{*}$ が存在し，それは最適ペナルティロバスト価値関数 $V_{r o b, p}^{*}=\max _{\pi} V_{r o b, p}^{\pi}$ と一致する．
* ペナルティロバストベルマン作用素を適用していくと，最適ペナルティロバスト価値関数は一つかつそれに収束します．

## 定理
### 定理 3.1
 
 $f(\cdot)$-ダイバージェンスを持つRMDPにおいて，
1.  任意の制約パラメータ $\rho > 0$ に対して，あるペナルティパラメータ $\lambda > 0$ が存在し，与えられた初期分布 $\mu \in \Delta(\mathcal{S})$ に対して $V_{\text {rob,c }}^{*}(\mu)=V_{\text {rob }, p}^{*}(\mu)$ となる．
2.  逆に，任意のペナルティパラメータ $\lambda > 0$ に対して，ある制約パラメータ $\rho > 0$ が存在し，$V_{\text {rob }, p}^{*}(\mu)=V_{\text {rob,c}}^{*}(\mu)$ となる．

つまり，RMDPにおけるminmax(コスト最小化)とmaxmin(報酬最大化)は同値であるということが言えます．


### 定理 3.2

Generative modelを用い，遷移確率を $n$ 個のサンプルで経験的に推定 $\widehat{P}\left(s^{\prime} \mid s, a\right)=\frac{1}{n} \sum_{i=1}^{n} \mathbf{1}\left(X_{i}^{(s, a)}=s^{\prime}\right)$ する場合 ($X_{i}^{(s, a)} \sim P^{*}(\cdot \mid s, a)$)$f(s)=(s-1)^{2}$ (カイ二乗ダイバージェンス不確実性集合) を選択すると，確率 $1-\delta$ で以下が成り立つ:

**大域的最適性**
$$
\left\|\widehat{V}_{r o b, p}^{*}-V_{r o b, p}^{*}\right\|_{\infty} \leq \widetilde{\mathcal{O}}\left(\frac{\max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\}}{(1-\gamma) \sqrt{n}}\right)
$$

**サンプル複雑度 ($\lambda=\mathcal{O}(1-\gamma)$ の場合)**
$$
\widetilde{\Omega}\left(\frac{|\mathcal{S}||\mathcal{A}| \lambda^{2}}{\varepsilon^{2}(1-\gamma)^{3}}\right)
$$
**サンプル複雑度 ($\lambda=\Omega(1-\gamma)$ の場合)**
$$
\widetilde{\Omega}\left(\frac{|\mathcal{S}||\mathcal{A}|}{\varepsilon^{2}(1-\gamma)^{3}} \min \left\{\frac{1}{16}, \frac{\lambda \gamma(1-\gamma)}{2 \gamma-1}\right\}\right)
$$
### 定理3.2の証明



#### 上界の証明

まず， $\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty}$ を評価します．
$\widehat{V}_{\text {rob }, \mathrm{p}}^{*}$ は $\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}}$ の不動点であり，$V_{\text {rob }, \mathrm{p}}^{*}$ は $\mathcal{T}_{\text {rob }, \mathrm{p}}$ の不動点です．つまり，
$$
\widehat{V}_{\text {rob }, \mathrm{p}}^{*} = \widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}} \widehat{V}_{\text {rob }, \mathrm{p}}^{*}
$$
$$
V_{\text {rob }, \mathrm{p}}^{*} = \mathcal{T}_{\text {rob }, \mathrm{p}} V_{\text {rob }, \mathrm{p}}^{*}
$$
が成り立ちます．これらを用いると，

$$
\begin{aligned}
\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty} & =\left\|\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}} \widehat{V}_{\text {rob }, \mathrm{p}}^{*}-\mathcal{T}_{\text {rob,p }} V_{\text {rob,p }}^{*}\right\|_{\infty} \\
& \leq\left\|\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}} \widehat{V}_{\text {rob }, \mathrm{p}}^{*}-\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}} V_{\text {rob,p }}^{*}\right\|_{\infty}+\left\|\widehat{\mathcal{T}}_{\text {rob,p }} V_{\text {rob,p }}^{*}-\mathcal{T}_{\text {rob }, \mathrm{p}} V_{\text {rob,p }}^{*}\right\|_{\infty} \quad (\text{三角不等式}) \\
& \leq \gamma\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty}+\left\|\widehat{\mathcal{T}}_{\text {rob,p }} V_{\text {rob,p }}^{*}-\mathcal{T}_{\text {rob }, \mathrm{p}} V_{\text {rob,p }}^{*}\right\|_{\infty} \quad (\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}} \text{ の } \gamma \text{-縮小性})
\end{aligned}
$$

となります． $\widehat{\mathcal{T}}_{\text {rob }, \mathrm{p}}$ もまた $\gamma$-縮小性を持ちます．
上の不等式を整理すると，

$$
\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty} \leq \frac{1}{1-\gamma}\left\|\widehat{\mathcal{T}}_{\text {rob,p }} V_{\text {rob,p }}^{*}-\mathcal{T}_{\text {rob }, \mathrm{p}} V_{\text {rob,p }}^{*}\right\|_{\infty}
$$

が得られます．右辺のノルムを評価します．$\mathcal{T}_{\text {rob,p }}$ の定義とその $\widehat{\mathcal{T}}_{\text {rob,p }}$ の定義から，
$$
\begin{aligned}
& \left\|\widehat{\mathcal{T}}_{\text {rob,p }} V_{\text {rob,p }}^{*}-\mathcal{T}_{\text {rob,p }} V_{\text {rob,p }}^{*}\right\|_{\infty} \\
= & \max _{s \in \mathcal{S}} \left| \max_{a \in \mathcal{A}} \left( R(s, a) + \gamma \inf _{Q \in \Delta(\mathcal{S})} \left( \mathbb{E}_{s^{\prime} \sim Q} V_{\text {rob,p }}^{*}(s^{\prime})+\lambda D_{f}\left(Q \| \widehat{P}_{s, a}\right) \right) \right) \right. \\
& \left. - \max_{a \in \mathcal{A}} \left( R(s, a) + \gamma \inf _{Q \in \Delta(\mathcal{S})} \left( \mathbb{E}_{s^{\prime} \sim Q} V_{\text {rob,p }}^{*}(s^{\prime})+\lambda D_{f}\left(Q \| P_{s, a}^{*}\right) \right) \right) \right| \\
\leq & \gamma \max _{s, a}\left|\inf _{Q \in \Delta(\mathcal{S})} \mathbb{E}_{s^{\prime} \sim Q} V_{\text {rob,p }}^{*}\left(s^{\prime}\right)+\lambda D_{f}\left(Q \| \widehat{P}_{s, a}\right)-\inf _{Q \in \Delta(\mathcal{S})} \mathbb{E}_{s^{\prime} \sim Q} V_{\text {rob,p }}^{*}\left(s^{\prime}\right)+\lambda D_{f}\left(Q \| P_{s, a}^{*}\right)\right|
\end{aligned}
$$

となります．ここで $\widehat{P}_{s,a}(\cdot) = \widehat{P}(\cdot | s, a)$， $P_{s,a}^*(\cdot) = P^*(\cdot | s, a)$ です．
$f(s)=(s-1)^2$ の場合，式(4)と補題A.1を用いると，内部問題は次のように書けます．

$$
\inf _{Q \in \Delta(\mathcal{S})} \mathbb{E}_{s^{\prime} \sim Q} V_{\mathrm{rob}, \mathrm{p}}^{*}\left(s^{\prime}\right)+\lambda D_{f}\left(Q \| P_{s, a}^{*}\right)=\sup _{\eta \in \Theta} \left( -\frac{1}{4 \lambda} \mathbb{E}_{s \sim P_{s, a}^{*}}\left(\eta-V_{\mathrm{rob}, \mathrm{p}}^{*}(s)\right)_{+}^{2}-\lambda+\eta \right) .
$$

ここで $(x)_+ = \max(x, 0)$ であり， $\Theta = \left[-\lambda, \frac{2}{1-\gamma}+2 \lambda\right]$ です．

$g(\eta, P) := -\frac{1}{4 \lambda} \mathbb{E}_{s \sim P}\left(\eta-V_{\text {rob }, \mathrm{p}}^{*}(s)\right)_{+}^{2}-\lambda+\eta$ とおきます．

$\widehat{P}(s)=\frac{1}{n} \sum_{k=1}^{n} \mathbf{1}\left(X_{k}=s\right)$ （ここで $X_k \sim P^*(\cdot | s, a)$ は i.i.d.）を用いると，評価したいのは $\max_{s,a} |\sup_{\eta \in \Theta} g(\eta, \widehat{P}_{s,a}) - \sup_{\eta \in \Theta} g(\eta, P_{s,a}^*)|$ です．これは $\sup_{\eta \in \Theta} |g(\eta, \widehat{P}_{s,a}) - g(\eta, P_{s,a}^*)|$ で上から抑えられます．

$Y_k := -\frac{1}{4 \lambda} \sum_{s \in \mathcal{S}} \mathbf{1}\left(X_{k}=s\right)\left(\eta-V_{\text {rob,p}}^{*}(s)\right)_{+}^{2}$ とおくと，
$g(\eta, \widehat{P}) = \frac{1}{n} \sum_{k=1}^n Y_k - \lambda + \eta$ であり，$g(\eta, P) = \mathbb{E} Y_1 - \lambda + \eta$ です．したがって，
$$
|g(\eta, P)-g(\eta, \widehat{P})|=\left|\frac{1}{n} \sum_{k=1}^{n} Y_{k}-\mathbb{E} Y_{1}\right|
$$
となります．

補題A.1より $\eta \in \Theta = \left[-\lambda, \frac{2}{1-\gamma}+2 \lambda\right]$ であり，$V_{\text {rob,p }}^{*}(s) \in [0, \frac{1}{1-\gamma}]$ なので，

$$
0 \leq \eta - V_{\text {rob,p }}^{*}(s) \leq \frac{2}{1-\gamma}+2\lambda - 0 = \frac{2}{1-\gamma}+2\lambda
$$
$$
0 \leq (\eta - V_{\text {rob,p }}^{*}(s))_+ \leq \frac{2}{1-\gamma}+2\lambda
$$
$$
0 \leq -Y_k = \frac{1}{4\lambda} (\eta - V_{\text {rob,p }}^{*}(X_k))_+^2 \leq \frac{1}{4\lambda} (\frac{2}{1-\gamma}+2\lambda)^2 = \frac{1}{4\lambda} (4(\frac{1}{1-\gamma})^2 + 8\frac{\lambda}{1-\gamma} + 4\lambda^2) = \frac{1}{\lambda(1-\gamma)^2} + \frac{2}{1-\gamma} + \lambda
$$
となります．ここでは， $\frac{\left(\frac{2}{1-\gamma}+4 \lambda\right)^{2}}{4 \lambda} \leq 16 \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\}$ と評価しています

よって，$Y_k$ は有界な確率変数です．Hoeffdingの不等式を用いると，固定された $\eta$ に対して，確率 $1-\delta$ で，

$$
|g(\eta, P)-g(\eta, \widehat{P})| = \left|\frac{1}{n} \sum_{k=1}^{n} Y_{k}-\mathbb{E} Y_{1}\right| \leq 16 \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \sqrt{\frac{\ln \frac{2}{\delta}}{2 n}}
$$

が成り立ちます．

次に，この評価を $\eta \in \Theta$ 全体で一様に成り立たせるために，$\varepsilon$-net を用います．

 $h(\eta) = \frac{1}{4 \lambda} (\eta - V)_+^2$ の $\eta$ に関する微分は $\frac{1}{2\lambda}(\eta - V)_+$ であり，これは $\frac{1}{2\lambda}$ リプシッツ連続です．したがって，$g(\eta, P)$ は $\eta$ に関して $\frac{1}{2\lambda}$ リプシッツ連続です．

$g(\eta, P)$ が $L = 8 \max \left\{\frac{1}{\lambda^{2}(1-\gamma)^{2}}, 1\right\}$ リプシッツ連続であるとします．
$\Theta$ の $\varepsilon$-net $\mathcal{N}_{\varepsilon}$ を考えます．その要素数は $|\mathcal{N}_{\varepsilon}| \leq 1 + \frac{\operatorname{diam}(\Theta)}{\varepsilon} = 1 + \frac{\frac{2}{1-\gamma}+4\lambda}{\varepsilon} \leq 1 + \frac{8 \max\{\frac{1}{1-\gamma}, \lambda\}}{\varepsilon}$ です．
任意の $\eta \in \Theta$ に対して，$\|\eta - \eta'\| \leq \varepsilon$ となる $\eta' \in \mathcal{N}_{\varepsilon}$ が存在します．

三角不等式と リプシッツ連続性から，

$$
\begin{aligned}
|g(\eta, \widehat{P})-g(\eta, P)| &\leq |g(\eta, \widehat{P}) - g(\eta', \widehat{P})| + |g(\eta', \widehat{P}) - g(\eta', P)| + |g(\eta', P) - g(\eta, P)| \\
&\leq 2 L \varepsilon + |g(\eta', \widehat{P}) - g(\eta', P)|
\end{aligned}
$$

よって，
$$
\sup _{\eta \in \Theta}|g(\eta, \widehat{P})-g(\eta, P)| \leq 2 L \varepsilon+\sup _{\eta' \in \mathcal{N}_{\varepsilon}}|g(\eta', \widehat{P})-g(\eta', P)| .
$$

Union boundを用いて，$\mathcal{N}_{\varepsilon}$ の全ての点 $\eta'$ で Hoeffding の不等式が成り立つ確率を評価します．
$\mathbb{P}\left(\sup _{\eta' \in \mathcal{N}_{\varepsilon}}|g(\eta', \widehat{P})-g(\eta', P)| \geq C \sqrt{\frac{\ln \frac{2|\mathcal{N}_{\varepsilon}|}{\delta}}{2 n}}\right) \leq \delta$
ここで $C = 16 \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\}$ です．
$\varepsilon = \frac{\lambda}{\sqrt{2n}}$ と選びます．（これは最適ではありませんが，最終的なオーダーには影響しません．）
すると，確率 $1-\delta$ で，
$$
\begin{aligned}
& \sup _{\eta \in \Theta}|g(\eta, \widehat{P})-g(\eta, P)| \\
\leq & 16 \max \left\{\frac{1}{\lambda^{2}(1-\gamma)^{2}}, 1\right\} \frac{\lambda}{\sqrt{2n}} + 16 \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \sqrt{\frac{\ln \frac{2|\mathcal{N}_{\varepsilon}|}{\delta}}{2 n}} \\
\leq & 16 \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \left( \max \left\{\frac{1}{\lambda(1-\gamma)^2}, 1 \right\} \frac{\lambda}{\sqrt{2n}} + \sqrt{\frac{\ln \frac{2|\mathcal{N}_{\varepsilon}|}{\delta}}{2 n}} \right) \\
= & \widetilde{\mathcal{O}}\left( \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \frac{1}{\sqrt{n}} \right)
\end{aligned}
$$
となります．ここで $\ln |\mathcal{N}_{\varepsilon}|$ の項は $\ln n$ のオーダーであり，$\widetilde{\mathcal{O}}$ に吸収されます．
最終的に，$\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty}$ の評価に戻ると，確率 $1-\delta$ で，

$$
\begin{aligned}
\left\|\widehat{V}_{\text {rob }, \mathrm{p}}^{*}-V_{\text {rob }, \mathrm{p}}^{*}\right\|_{\infty} &\leq \frac{1}{1-\gamma} \left\|\widehat{\mathcal{T}}_{\text {rob,p }} V_{\text {rob,p }}^{*}-\mathcal{T}_{\text {rob }, \mathrm{p}} V_{\text {rob,p }}^{*}\right\|_{\infty} \\
&\leq \frac{\gamma}{1-\gamma} \sup_{s,a} \sup_{\eta \in \Theta} |g(\eta, \widehat{P}_{s,a}) - g(\eta, P_{s,a}^*)| \\
&\leq \widetilde{\mathcal{O}}\left( \frac{\gamma}{(1-\gamma)\sqrt{n}} \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \right) \\
&= \widetilde{\mathcal{O}}\left( \frac{1}{(1-\gamma)\sqrt{n}} \max \left\{\frac{1}{\lambda(1-\gamma)^{2}}, \lambda\right\} \right)
\end{aligned}
$$

が成り立ちます．
