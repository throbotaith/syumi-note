# 強化学習における不等式について

## Performance difference lemma(性能差補題)

これは、方策間の価値関数の差($V^\pi(s_0)-V^{\pi'}(s_0)$)について、いい感じに上界を与えてくれる手法です。

性能差補題では、アドバンテージ関数が出てきます。というかこれがかなり重要なので式を書きましょう。

アドバンテージ関数は、各行動の価値が、その平均値よりもどれだけ高いかどうかを示した関数になります。

$$
A^{\pi'}(s,a)=Q^{\pi'}(s,a) - V^{\pi'}(s)
$$
ここで、
$$
Q^{\pi'}(s,a) = r(s,a) + \gamma \mathbb{E}_{s' \sim P(\cdot |s,a) }[V^{\pi'}(s')]
$$
なので、
$$
\begin{aligned}
    A^{\pi'}(s,a) + V^{\pi'}(s) &= r(s,a) + \gamma \mathbb{E}_{s' \sim P(\cdot |s,a) }[V^{\pi'}(s')] \\
    r(s,a)& = A^{\pi'}(s,a) + V^{\pi'}(s) -\gamma \mathbb{E}_{s' \sim P(\cdot |s,a) }[V^{\pi'}(s')]
\end{aligned}
$$

この式は後で使います。

### 状態価値関数の差分から始める

$$
V^{\pi}(s_0) = \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^tr(s_t,a_t)|s_0]
$$

が定義です。次にさっきの報酬関数を使って上の式の中を置き換えましょう。

$$
V^{\pi}(s_0) = \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^t(A^{\pi'}(s_t,a_t) + V^{\pi'}(s_t) -\gamma \mathbb{E}_{s_{t+1} \sim P(\cdot |s,a) }[V^{\pi'}(s_{t+1})])|s_0]
$$

期待値の線形性を使いますと、

$$
V^{\pi}(s_0) = \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^tA^{\pi'}(s_t,a_t)|s_0] +  \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^tV^{\pi'}(s_t)|s_0]  -\gamma \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^t\mathbb{E}_{s_{t+1} \sim P(\cdot |s,a) }[V^{\pi'}(s_{t+1})])|s_0]]
$$

ここからどうするかですが、先程はすべて分解してしまったのですがどうも二項目と三項目はテレスコーピングサムをつかって値を打ち消してくれそうですね。第二項と第三項を戻して第二項としましょう。第二項について考えると、
$$
\text{第二項}= \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^t(V^{\pi'}(s_t) - \mathbb{E}_{s_{t+1} \sim P(\cdot |s,a) }[V^{\pi'}(s_{t+1})])|s_0]]
$$
ここで、第二項の第二項について、期待値と関係がない$\sum$を除いてその形を見ると、

$$
\mathbb{E}_{\tau \sim \pi}[\mathbb{E}_{s_{t+1} \sim P(\cdot |s,a) }[V^{\pi'}(s_{t+1})])|s_0]]
$$

期待値のタワー法則っていうのがあって、$\mathbb{E}[\mathbb{E}[X|Y]]=\mathbb{E}[X]$となります。これを使うと、
$$
\mathbb{E}_{\tau \sim \pi}[\mathbb{E}_{s_{t+1} \sim P(\cdot |s,a) }[V^{\pi'}(s_{t+1})])|s_0]] = \mathbb{E}_{\tau \sim \pi}[V^{\pi'}(s_{t+1})])]
$$
になります。先程の第２項に戻りますと、
$$
\text{第二項}= \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^t(V^{\pi'}(s_t) - V^{\pi'}(s_{t+1})]
$$
これを展開すると、
$$
\text{第二項} = (V^{\pi'}(s_0) - \gamma V^{\pi'}(s_1)) + (\gamma V^{\pi'}(s_1) - \gamma^2 V^{\pi'}(s_2)) + (\gamma^2 V^{\pi'}(s_2) - \dots)
$$
最初の項$V^{\pi'}(s_0)$と最後の項$\gamma^\infty V^{\pi'}(s_\infty)$が残りますが、あれ最後の項ってこれ定義できるんか？まあとりあえず最後らへんの項は$\gamma^\infty$が係数としてかかっているので$0$になります。よって、
$$
\text{第二項} = \mathbb{E}_{\tau \sim \pi} [V^{\pi'}(s_0)]= V^{\pi'}(s_0)
$$
です。なぜ期待値が消えるかというと、$\tau$という確率変数は$\pi$(と状態遷移確率)からサンプルされていますが、期待値の中身は$V^{\pi'}(s_0)$なので、$\tau \sim \pi$に依存しないんですね。だから消えます。

では、第二項を使って、V^{\pi}(s_0)の式に代入すると、
$$
\begin{align*}
    V^{\pi}(s_0) &= \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^tA^{\pi'}(s_t,a_t)|s_0] +  V^{\pi'}(s_0) \\
    V^{\pi}(s_0) - V^{\pi'}(s_0) & = \mathbb{E}_{\tau \sim \pi}[\sum_{t=0}^{\infty}\gamma^tA^{\pi'}(s_t,a_t)|s_0]
\end{align*}
$$
最後に、$\tau \sim \pi$を$s \sim d_{s_o}^\pi$にします。
$$
\mathbb{E}_{\tau \sim \pi} \left[ \sum_{t=0}^\infty \gamma^t A^{\pi'}(s_t, a_t) \right] = \frac{1}{1-\gamma} \mathbb{E}_{s \sim d_{s_0}^\pi} \left[ \mathbb{E}_{a \sim \pi(\cdot|s)} [A^{\pi'}(s,a)] \right]
$$

よって、Performance difference lemmaが導かれました。
$$
V^{\pi}\left(s_{0}\right)-V^{\pi^{\prime}}\left(s_{0}\right)=\frac{1}{1-\gamma} \mathbb{E}_{s \sim d_{s_{0}}^{\pi}} \mathbb{E}_{a \sim \pi(\cdot \mid s)}\left[A^{\pi^{\prime}}(s, a)\right]
$$
