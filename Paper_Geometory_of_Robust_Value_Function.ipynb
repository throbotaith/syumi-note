# この研究について

[ロバスト価値関数のポリトープ](https://arxiv.org/abs/2201.12929)について，まとめていきます．価値関数のポリトープと同じように，価値関数の空間における探索を続けましょう．方策を入力として，価値関数を出力とした写像の出力先は，凸でないポリトープになることを前の論文で確認しました．では，ロバスト価値関数はどのような幾何学的構造になるのでしょうか？
そこで，本研究では，非ロバストな価値関数の空間を状態ごとに分解します．そこからロバスト価値関数空間が円錐超曲面(多様体とかも考慮した超平面の一般化)によって構成されるということ(図1)，ロバスト価値関数空間が不確実性集合の端点から構成される部分集合で完全に決定できるということが示されます．これはRectangularといい感じに交わってそうな話です．

## (価値関数空間までの)準備

公な定義($S$が状態集合)などは飛ばします．
* 空でない集合：$\mathcal{U}$
* 極錐：$\mathcal{U}^{*}$
$$
\begin{equation*}
\mathcal{U}^{*}:=\{y \mid\langle y, x\rangle \leq 0, \forall x \in \mathcal{U}\} \tag{1}
\end{equation*}
$$
* 凸包：$\operatorname{conv}(\cdot)$
* 凸集合の端点：$\operatorname{ext}(\cdot)$

**MDP**
* $\operatorname{MDP}\left(\mathcal{S}, \mathcal{A}, P, r, \gamma, p_{0}\right)$
* 状態遷移確率(,行列表記)： $P_{s, a} \in \Delta_{\mathcal{S}}$ ， $P:=\left(P_{s, a}\right)_{s \in \mathcal{S}, a \in \mathcal{A}} \in\left(\Delta_{\mathcal{S}}\right)^{\mathcal{S} \times \mathcal{A}}$
* 報酬関数(,行列表記)： $r_{s, a} \in \mathbb{R}$ ， $r:=\left(r_{s, a}\right)_{s \in \mathcal{S}, a \in \mathcal{A}} \in \mathbb{R}^{\mathcal{S} \times \mathcal{A}}$ 
* ある状態に関する遷移確率ベクトル：$P_{s}:=\left(P_{s, a}\right)_{a \in \mathcal{A}} \in\left(\Delta_{S}\right)^{\mathcal{A}}$
* ある状態に関する報酬ベクトル：$r_{s}:=\left(r_{s, a}\right)_{a \in \mathcal{A}} \in \mathbb{R}^{\mathcal{A}}$
* 決定論的方策($\pi_{s, a}=1$ を除き全てゼロ)：$d_{s, a} \in \Delta_{\mathcal{A}}$
* ある方策$\pi$に従った場合の状態遷移確率行列，ベクトル
$$
\begin{align*}
P^{\pi} & :=\left(P^{\pi_{s}}\right)_{s \in \mathcal{S}} \in\left(\Delta_{\mathcal{S}}\right)^{\mathcal{S}}, \quad \text { ここで } \\
P^{\pi_{s}} & :=P_{s} \pi_{s}=\sum_{a \in \mathcal{A}} \pi_{s, a} P_{s, a} \in \Delta_{\mathcal{S}} \tag{2}
\end{align*}
$$
* ある$\pi$に従った場合の報酬関数行列，ベクトル
$$
\begin{align*}
r^{\pi} & :=\left(r^{\pi_{s}}\right)_{s \in \mathcal{S}} \in \mathbb{R}^{\mathcal{S}}, \quad \text { ここで } \\
r^{\pi_{s}} & :=r_{s}^{\top} \pi_{s}=\sum_{a \in \mathcal{A}} \pi_{s, a} r_{s, a} \in \mathbb{R} . \tag{3}
\end{align*}
$$
* ある$\pi$，状態遷移確率$P$に従った場合の状態価値関数
$$
\begin{equation*}
V^{\pi, P}(s):=\mathbb{E}_{P \pi}\left[\sum_{t=0}^{\infty} \gamma^{t} r_{s_{t}, a_{t}} \mid s_{0}=s\right] \tag{4}
\end{equation*}
$$
**RMDP**
* 不確実性集合：$\mathcal{P} \subseteq\left(\Delta_{\mathcal{S}}\right)^{\mathcal{S} \times \mathcal{A}}$ 
* ロバスト状態価値関数
$$
\begin{align*}
V^{\pi, \mathcal{P}}(s) & :=\min _{P \in \mathcal{P}} V^{\pi, P}(s)  \tag{5}\\
\end{align*}
$$
* 最適ロバスト状態価値関数
$$
\begin{align*}
V^{\star, \mathcal{P}}(s) & :=\max _{\pi \in \Pi} V^{\pi, \mathcal{P}}(s) \tag{6}
\end{align*}
$$
* (s,a)-Rectangular不確実性集合
$$
\begin{equation*}
\mathcal{P}=\left\{P \mid P_{s, a} \in \mathcal{P}_{s, a}, \forall s \in \mathcal{S}, \forall a \in \mathcal{A}\right\} \tag{7}
\end{equation*}
$$
* s-Rectangular不確実性集合
$$
\begin{equation*}
\mathcal{P}=\left\{P \mid P_{s} \in \mathcal{P}_{s}, \forall s \in \mathcal{S}\right\} \tag{8}
\end{equation*}
$$
## 価値関数空間

* 方策集合(空間)から(ロバスト)価値関数集合(空間)への写像
$$
\begin{align*}
f_{P}(U) & :=\left\{V^{\pi, P} \mid \pi \in U\right\}  \tag{10}\\
f_{\mathcal{P}}(U) & :=\left\{V^{\pi, \mathcal{P}} \mid \pi \in U\right\} \tag{11}
\end{align*}
$$
* 定常確率論的方策集合：$\Pi=\left(\Delta_{\mathcal{A}}\right)^{S}$ 
* 価値関数，ロバスト価値関数(価値関数の集合)空間
$$
\begin{align*}
\mathcal{V}^{P} & :=f_{P}(\Pi)  \tag{12}\\
\mathcal{V}^{\mathcal{P}} & :=f_{\mathcal{P}}(\Pi) \tag{13}
\end{align*}
$$
* $s$で一致する方策の集合
$$
\begin{equation*}
Y^{\pi_{s}}:=\left\{\pi^{\prime} \mid \pi_{s}^{\prime}=\pi_{s}\right\} \tag{14}
\end{equation*}
$$
* $I-\gamma P^{\pi}$ の状態 $s$ に対応する行
$$
\begin{equation*}
L^{\pi_{s}, P_{s}}:=\mathbf{e}_{s}-\gamma P^{\pi_{s}}=\mathbf{e}_{s}-\gamma P_{s} \pi_{s} \tag{15}
\end{equation*}
$$
**価値関数ポリトープ**
 $V^{\pi, P}$ を分解して超平面として表しましょう．
$$
\begin{equation*}
H^{\pi_{s}, P_{s}}:=\left\{\mathbf{x} \in \mathbb{R}^{\mathcal{S}} \mid\left\langle\mathbf{x}, L^{\pi_{s}, P_{s}}\right\rangle=r^{\pi_{s}}\right\} \tag{16}
\end{equation*}
$$
ここで，$\mathbf{x} \in \mathbb{R}^{\mathcal{S}}$は状態関数です．以下のような演算を満たす$\mathbf{x}$ を求めます．
$$
r^\pi=\left(I-\gamma P^\pi\right) V^\pi
$$
図にあるように何次元でも，$\pi$上の超平面はすべて $V^{\pi, P}$を通ります．$V^{\pi, P}$ はすべての状態で，内積結果が$r^{\pi_{s}}$になる$\mathbf{x}$です．
![](https://cdn.mathpix.com/cropped/2025_04_10_e9019ac9733584f1c7dag-3.jpg?height=394&width=854&top_left_y=215&top_left_x=1042)
$V^{\pi, P}$をすべての $\pi \in \Pi$ 分集めたもの(和集合)は，
$$
\begin{equation*}
\nu^{P}=\bigcup_{\pi \in \Pi} \bigcap_{s \in \mathcal{S}} H^{\pi_{s}, P_{s}} \tag{18}
\end{equation*}
$$
なので価値関数空間です．

ここで，2つの補題をまとめて説明します．まとめた都合上，数式番号が論文に対して前後します．
**補題3.1**
全ての状態に対する，超平面において交わる唯一の点は$V^{\pi, P}$です．$(18)$の主張を支持します．
 $\pi$ と$P$ を考える．このとき，次が成り立ちます．

$$
\begin{equation*}
\left\{V^{\pi, P}\right\}=\bigcap_{s \in \mathcal{S}} H^{\pi_{s}, P_{s}} \tag{17}
\end{equation*}
$$
**補題3.2**
(18)まんまです．和集合と積集合の演算？は交換できます．が，$\bigcup_{\pi \in \Pi}$と$\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}}$の部分だけ異なります．なぜなら，**TODO**
$$
\begin{equation*}
\mathcal{V}^{P}=\bigcup_{\pi \in \Pi} \bigcap_{s \in \mathcal{S}} H^{\pi_{s}, P_{s}}=\bigcap_{s \in \mathcal{S}} \bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} H^{\pi_{s}, P_{s}} \tag{19}
\end{equation*}
$$
また，定義に戻ります．
* $H^{\pi_{s}, P_{s}}$ による閉じた上下半空間
$$
\begin{align*}
\mathcal{H}_{+}^{\pi_{s}, P_{s}^{s}} & :=\left\{\mathbf{x} \in \mathbb{R}^{\mathcal{S}} \mid\left\langle\mathbf{x}, L^{\pi_{s}, P_{s}}\right\rangle \geq r^{\pi_{s}}\right\}, \\
\mathcal{H}_{-}^{\pi_{s}, P_{s}} & :=\left\{\mathbf{x} \in \mathbb{R}^{\mathcal{S}} \mid\left\langle\mathbf{x}, L^{\pi_{s}, P_{s}}\right\rangle \leq r^{\pi_{s}}\right\} . \tag{20}
\end{align*}
$$
* 半空間で表現した状態価値関数空間
$$
\begin{equation*}
\mathcal{V}_{P}=\bigcap_{s \in \mathcal{S}} \bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{+}^{\pi_{s}, P_{s}} \cap \mathcal{H}_{-}^{\pi_{s}, P_{s}} . \tag{21}
\end{equation*}
$$
**価値関数空間の作り方**
下図は価値関数空間を作っている様子です．
(a)：(無数にある)$V^{\pi, P}$のプロットと，そこから発生する超平面をすべて書きます．
(b)：異なる状態ごとに超平面を色分けします．
(c)：各状態に対して，超平面の上半空間の和集合(線矢印)，超平面のした半空間の和集合(点線矢印)を作ります．
(d)：各状態に対して，異なる行動に対する超平面$H^{d_{s, a}, P_{s}}$ をプロットします．方策が決定論的になっていることに注意．

## 価値関数空間をポリトープで表現したい．
凸多面体は半空間たちの有限な共通部分，ポリトープは凸多面体の有界な**有限和集合**です．(21)において，半空間の有限共通部分は，$\mathcal{H}_{+}^{\pi_{s}, P_{s}} \cap \mathcal{H}_{-}^{\pi_{s}, P_{s}}$です．しかし，$\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{+}^{\pi_{s}, P_{s}} \cap \mathcal{H}_{-}^{\pi_{s}, P_{s}}$は**無限和集合**になってしまいます．なぜなら行動上の確率単体($\Delta_{\mathcal{A}}$)は**非可算無限集合**になるからです．そこで，無限和集合を取り除けば価値関数空間をポリトープとして表現できそうです．
### ポリトープで表現するための補題や定理たち
ここはテクニカルなところなので，流れだけでもわかっておきましょう．補題3.3で無限和集合を分配して表現します．次に補題3.4をつかって無限和集合を有限和集合に変換します(仮定も追加？)．

**補題3.1**：無限和の集合を別の集合に変換
 $\pi$ と $P$ を考えます．全ての $s \in \mathcal{S}$ に対して，次が成り立ちます．

$$
\begin{align*}
& \bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{+}^{\pi_{s}, P_{s}} \cap \mathcal{H}_{-}^{\pi_{s}, P_{s}}= \\
& \quad\left[\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{+}^{\pi_{s}, P_{s}}\right] \cap\left[\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{-}^{\pi_{s}, P_{s}}\right] . \tag{22}
\end{align*}
$$
**補題3.2**：すべての方策に対する無限和の集合を$d_{s, a} \in \Delta_{\mathcal{A}}$の有限和の集合に変換
 $\pi$ と $P$ を考えます．全ての $s \in \mathcal{S}$ に対して，次が成り立ちます．
 $$
\begin{equation*}
\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{H}_{\delta}^{\pi_{s}, P_{s}}=\bigcup_{a \in \mathcal{A}} \mathcal{H}_{\delta}^{d_{s, a}, P_{s}}, \quad \forall \delta \in\{+,-\} . \tag{23}
\end{equation*}
$$
$U$が有限和に変換できるのは，方策が決定論的になったからです．

**定理3.5**：結論
 $P$ を考えます． $V^{P}$ は次のように表現できます．
$$
\begin{align*}
\mathcal{V}^{P} & =\bigcap_{s \in \mathcal{S}}\left[\left[\bigcup_{a \in \mathcal{A}} \mathcal{H}_{+}^{d_{s, a}, P_{s}}\right] \cap\left[\bigcup_{a \in \mathcal{A}} \mathcal{H}_{-}^{d_{s, a}, P_{s}}\right]\right]  \tag{24}\\
& =\bigcup_{\mathbf{a} \in \mathcal{A}^{S}} \bigcup_{\mathbf{a}^{\prime} \in \mathcal{A}^{s}} \bigcap_{s \in \mathcal{S}}\left[\mathcal{H}_{+}^{d_{s, \mathbf{a}_{s}}, P_{s}} \cap \mathcal{H}_{-}^{d_{s, \mathbf{a}_{s}^{\prime}}, P_{s}}\right]
\end{align*}
$$
 $\mathbf{a}=\left(\mathbf{a}_{s}\right)_{s \in \mathcal{S}}, \mathbf{a}^{\prime}=\left(\mathbf{a}_{s}^{\prime}\right)_{s \in \mathcal{S}}$, そして $\mathbf{a}_{s}, \mathbf{a}_{s}^{\prime} \in \mathcal{A}$ です．

## RMDPにおける価値関数空間(Final step)
ロバスト価値関数の写像先が，超曲面で表現することが分かりました．これを用いて非ロバスト価値関数の場合のように，ロバスト価値関数空間を作っていきましょう．以下の図にその過程を示します．

![](https://cdn.mathpix.com/snip/images/cKuZTU45bkVAgeF5Bs8ydORCGMkQXX3CeUI7UJxcjfk.original.fullsize.png)
ロバスト価値関数空間を示します．

**補題4.4**
s-Rectangular $\mathcal{P}$ ，方策 $\pi$ について，次が成り立ちます．
$$
\begin{equation*}
\left\{V^{\pi, \mathcal{P}}\right\}=\bigcap_{s \in \mathcal{S}} C^{\pi_{s}, \mathcal{P}_{s}} \tag{29}
\end{equation*}
$$
以下補題たちは，価値関数空間をポリトープで表現したいのロバスト版です．ロバスト版では，ポリトープ→円錐超曲面です．

**補題4.5**
s-Rectangular $\mathcal{P}$ について，ロバスト価値関数空間は以下のように表現できます．
$$
\begin{equation*}
\mathcal{V}^{\mathcal{P}}=\bigcup_{\pi \in \Pi} \bigcap_{s \in \mathcal{S}} C^{\pi_{s}, \mathcal{P}_{s}}=\bigcap_{s \in \mathcal{S}} \bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} C^{\pi_{s}, \mathcal{P}_{s}} . \tag{30}
\end{equation*}
$$

**補題4.6**
s-Rectangular $\mathcal{P}$ ，全ての状態 $s \in \mathcal{S}$ に対して、次が成り立ちます．
$$
\begin{equation*}
\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} C^{\pi_{s}, \mathcal{P}_{s}}=\left[\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{C}_{+}^{\pi_{s}, \mathcal{P}_{s}}\right] \cap\left[\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{C}_{-}^{\pi_{s}, \mathcal{P}_{s}}\right] \tag{31}
\end{equation*}
$$

補題4.7だけちょっと違います．下超平面は，(s,a)-Rectangularの場合のみ等式が成立します．

**補題4.7**
s-Rectangular $\mathcal{P}$ ，全ての状態 $s \in \mathcal{S}$ に対して、次が成り立ちます．
$$
\begin{align*}
\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathrm{C}_{+}^{\pi_{s}, \mathcal{P}_{s}} & =\bigcup_{a \in \mathcal{A}} \mathrm{C}_{+}^{d_{s, a}, \mathcal{P}_{s}}  \tag{32}\\
\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathrm{C}_{-}^{\pi_{s}, \mathcal{P}_{s}} & \supseteq \bigcup_{a \in \mathcal{A}} \mathcal{C}_{-}^{d_{s, a}, \mathcal{P}_{s}} \tag{33}
\end{align*}
$$
これらの補題をひっくるめて定理4.8を生成しましょう．

**補題4.8**
s-Rectangular $\mathcal{P}$ ，ロバスト価値関数空間 $\mathcal{V}^{\mathcal{P}}$ は次を満たします．

$$
\begin{align*}
\mathcal{V}^{\mathcal{P}} & =\bigcap_{s \in \mathcal{S}}\left[\left[\bigcup_{a \in \mathcal{A}} \mathcal{C}_{+}^{d_{s, a}, \mathcal{P}_{s}}\right] \cap\left[\bigcup_{\pi_{s} \in \Delta_{\mathcal{A}}} \mathcal{C}_{-}^{\pi_{s}, \mathcal{P}_{s}}\right]\right] \\
& \supseteq \bigcap_{s \in \mathcal{S}}\left[\left[\bigcup_{a \in \mathcal{A}} \mathcal{C}_{+}^{d_{s, a}, \mathcal{P}_{s}}\right] \cap\left[\bigcup_{a \in \mathcal{A}} \mathcal{C}_{-}^{d_{s, a}, \mathcal{P}_{s}}\right]\right] \tag{34}
\end{align*}
$$

ここで、2行目の等式は $\mathcal{P}$ が (s,a)-Rectangularの場合のみ等式が成立します．

## 有効不確実性部分集合
