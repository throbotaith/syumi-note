# 準備
**MDPの設定**
有限状態、有限行動のMDPを想定してます。目的関数は以下の通り、通常通りです。

$$
\begin{equation*}
\max _{\pi} V^{\pi}(s):=\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^{t} r_{s_{t} a_{t}} \mid i_{0}=s, a_{t} \sim \pi\left(\cdot \mid s_{t}\right), t=0,1, \ldots\right], \quad \forall s \in \mathcal{S} \tag{2.1}
\end{equation*}
$$
**報酬関数からOccupancy measure($\lambda^{\pi}$)へ**
方策を最大化する問題を、$\lambda^{\pi}$に変数変換して新しい関数を最大化する問題に変化させます。
$$
\begin{equation*}
\max _{\pi} R(\pi):=F\left(\lambda^{\pi}\right) \tag{2.2}
\end{equation*}
$$
$\lambda^{\pi}$の定義はいつもの感じです。$\Lambda: \Delta_{\mathcal{A}}^{\mathcal{S}} \mapsto \mathcal{L}$です。
$$
\begin{equation*}
\lambda_{s a}^{\pi}=\Lambda_{s a}(\pi):=\sum_{t=0}^{\infty} \gamma^{t} \cdot \mathbb{P}\left(s_{t}=s, a_{t}=a \mid \pi, s_{0} \sim \xi\right) \quad \text { for } \quad \forall a \in \mathcal{A}, \forall s \in \mathcal{S} \tag{2.3}
\end{equation*}
$$
具体例として、$F(\lambda)=\langle r, \lambda\rangle$とされることが多く、これは以下のように期待総報酬に対応します。
$$
F(\lambda)=\langle\lambda, r\rangle=\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^{t} r_{s_{t} a_{t}} \mid \pi, s_{0} \sim \xi\right]
$$


2.2では方策の最大化でしたが、等価な$\lambda$最大化問題に変換できます。
$$
\begin{equation*}
\max _{\lambda} F(\lambda) \quad \text { s.t. } \quad \sum_{a \in \mathcal{A}}\left(I-\gamma P_{a}^{\top}\right) \lambda_{a}=\xi, \lambda \geq 0 \tag{2.4}
\end{equation*}
$$
ここで、$\lambda_{a}=\left[\lambda_{1 a}, \cdots, \lambda_{S a}\right]^{\top} \in \mathbb{R}^{A}$は$\lambda$の$a$番目の列です。$\xi$は状態空間$\mathcal{S}$上の初期分布です。

# 変分方策勾配定理
状態行動の次元の呪いに対処する体で方策をパラメータ化するのがお決まりらしい。確かに状態*行動の表形式からパラメータ次元に下げられるので。$\pi=\pi_{\theta}$として考えます。$\theta \in \Theta \subset \mathbb{R}^{d}$はパラメータベクトルです。問題も以下のように置き換えられます。
$$
\begin{equation*}
\max _{\theta \in \Theta} R\left(\pi_{\theta}\right):=F\left(\lambda^{\pi_{\theta}}\right) \tag{3.1}
\end{equation*}
$$
