# 準備
**MDPの設定**
有限状態、有限行動のMDPを想定してます。目的関数は以下の通り、通常通りです。

$$
\begin{equation*}
\max _{\pi} V^{\pi}(s):=\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^{t} r_{s_{t} a_{t}} \mid i_{0}=s, a_{t} \sim \pi\left(\cdot \mid s_{t}\right), t=0,1, \ldots\right], \quad \forall s \in \mathcal{S} \tag{2.1}
\end{equation*}
$$
**報酬関数からOccupancy measure($\lambda^{\pi}$)へ**
方策を最大化する問題を、$\lambda^{\pi}$に変数変換して新しい関数を最大化する問題に変化させます。
$$
\begin{equation*}
\max _{\pi} R(\pi):=F\left(\lambda^{\pi}\right) \tag{2.2}
\end{equation*}
$$
$\lambda^{\pi}$の定義はいつもの感じです。$\Lambda: \Delta_{\mathcal{A}}^{\mathcal{S}} \mapsto \mathcal{L}$です。
$$
\begin{align*}
\lambda_{s a}^{\pi}=&\Lambda_{s a}(\pi) \\
&:= \sum_{t=0}^{\infty} \gamma^{t} \cdot \mathbb{P}\left(s_{t}=s, a_{t}=a \mid \pi, s_{0} \sim \xi\right) \quad \text { for } \quad \forall a \in \mathcal{A}, \forall s \in \mathcal{S} \tag{2.3}
\end{align*}
$$
具体例として、$F(\lambda)=\langle r, \lambda\rangle$とされることが多く、これは以下のように期待総報酬に対応します。
$$
F(\lambda)=\langle\lambda, r\rangle=\mathbb{E}\left[\sum_{t=0}^{\infty} \gamma^{t} r_{s_{t} a_{t}} \mid \pi, s_{0} \sim \xi\right]
$$


2.2では方策の最大化でしたが、**等価**な$\lambda$最大化問題に変換できます。
$$
\begin{equation*}
\max _{\lambda} F(\lambda) \quad \text { s.t. } \quad \sum_{a \in \mathcal{A}}\left(I-\gamma P_{a}^{\top}\right) \lambda_{a}=\xi, \lambda \geq 0 \tag{2.4}
\end{equation*}
$$
ここで、$\lambda_{a}=\left[\lambda_{1 a}, \cdots, \lambda_{S a}\right]^{\top} \in \mathbb{R}^{A}$は$\lambda$の$a$番目の列です。$\xi$は状態空間$\mathcal{S}$上の初期分布です。

# 変分方策勾配定理
状態行動の次元の呪いに対処する体で方策をパラメータ化するのがお決まりらしい。確かに状態*行動の表形式からパラメータ次元に下げられるので。$\pi=\pi_{\theta}$として考えます。$\theta \in \Theta \subset \mathbb{R}^{d}$はパラメータベクトルです。問題も以下のように置き換えられます。
$$
\begin{equation*}
\max _{\theta \in \Theta} R\left(\pi_{\theta}\right):=F\left(\lambda^{\pi_{\theta}}\right) \tag{3.1}
\end{equation*}
$$
ここで、$F$は$\lambda(\theta):=\lambda^{\pi_{\theta}}$に関して凹な関数とします。また、パラメータ空間$\Theta \subset \mathbb{R}^{d}$は閉じていて凸集合となります。

**累積報酬に対する方策勾配定理**
状態価値関数は累積報酬の期待値ですが、これは報酬関数と$\lambda$の内積で表現できます。つまり、以下のように表現できます。これはパラメータバージョンです。
$$V(\theta ; r):=\langle\lambda(\theta), r\rangle$$
と思ったんですが、あまり今後の話に関係なさそう。では、気を取り直して方策勾配定理は以下のように書くことが出来ます。$Q^{\pi}(s, a ; r):=\mathbb{E}^{\pi}\left[\sum_{t} \gamma^{t} r\left(s_{t}, a_{t}\right) \mid s_{0}=s, a_{0}=a, a_{t} \sim \pi\left(\cdot \mid s_{t}\right)\right]$とすると、
$$
\begin{equation*}
\nabla_{\theta} V(\theta ; r)=\mathbb{E}^{\pi_{\theta}}\left[\sum_{t=0}^{\infty} \gamma^{t} Q^{\pi_{\theta}}\left(s_{t}, a_{t} ; r\right) \cdot \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)\right], \tag{3.2}
\end{equation*}
$$
です。おなじみの構造です。対数の勾配トリックを使うとでます。

**累積報酬以外ではこの方策勾配定理を使うことが難しい？**
今回のように累積報酬の代わりに別の目的関数を設定する場合、この方策勾配定理を用いることは困難です。理由として目的関数が累積報酬でない他の関数の場合、多くが加法性を満たさないからというのがあげられます。加法性とは、ある$a,b \in X$について、$f:X\rightarrow Y$としたときに、$f(a+b)=f(a)+f(b)$が成り立つことです。

***累積報酬関数以外の一般的な関数の方策勾配**
ここで、そのような関数$F(\lambda(\theta)):=F\left(\lambda^{\pi_{\theta}}\right)$を一般化ユーティリティ関数ということにします。それでは、この関数の勾配を求めてみましょう。
 $$
\begin{equation*}
\nabla_{\theta} R\left(\pi_{\theta}\right)=\sum_{s \in \mathcal{S}} \sum_{a \in \mathcal{A}} \frac{\partial F(\lambda(\theta))}{\partial \lambda_{s a}} \cdot \nabla_{\theta} \lambda_{s a}(\theta) \tag{3.3}
\end{equation*}
$$
合成関数の微分の定義によりこんな感じになりますが、$\frac{\partial F(\lambda(\theta))}{\partial \lambda_{s a}}$と$\nabla_{\theta} \lambda_{s a}(\theta)$を推定することが難しそうです。Tsybakovによると、$\lambda^{\pi_{\theta}}$を推定してから導関数を出す試みは収束がとても遅くなります。

**ユーティリティ関数の勾配を求める方法1:双対性の活用**
$F$の勾配を求めることがムズい事はわかりました。ではこの関数に対する双対関数を考えてそこで勾配というか最適化をしましょうという方法です。$F$の双対関数は以下のように書くことが出来、もとの問題によらず必ず凸関数になります。とはいえ、しばしばこの双対関数は微分がムズいらしいです。

**変分方策勾配定理**
双対関数を使うことでユーティリティ関数の方策勾配を求めることができる、変分方策勾配定理を以下のように定義します。$z$の一例としては報酬関数、コスト関数などがあります。コレをいろいろな関数にしていくのが変分方策勾配なんですね。
$$
\begin{align*}
&\nabla_{\theta} R\left(\pi_{\theta}\right)\\
&=\lim _{\delta \rightarrow 0_{+}} \underset{x}{\operatorname{argmax}} \inf _{z} \\
&\left\{V(\theta ; z)+\delta \nabla_{\theta} V(\theta ; z)^{\top} x -F^{*}(z)- \\ \frac{\delta}{2}\|x\|^{2}\right\} \tag{3.4}
\end{align*}
$$

## 変分方策勾配の推定と推定誤差バウンド
定理がわかったとはいえ、実際に計算するには$V$やその勾配などを実際に求めないといけません。推定する計算式は以下のように書くことができます。軌跡を$\zeta_{i}=\left\{s_{k}^{(i)}, a_{k}^{(i)}\right\}_{k=1}^{K}$としましょう。そうなると、任意の関数$z$に対して以下に推定することができます。
$$
\begin{align*}
\tilde{V}(\theta ; z) & :=\frac{1}{n} \sum_{i=1}^{n} V\left(\theta ; z ; \zeta_{i}\right):=\frac{1}{n} \sum_{i=1}^{n} \sum_{k=1}^{K} \gamma^{k} \cdot z\left(s_{k}^{(i)}, a_{k}^{(i)}\right)  \tag{3.5}\\
\nabla \tilde{V}(\theta ; z) & :=\frac{1}{n} \sum_{i=1} \nabla_{\theta} V\left(\theta ; z ; \zeta_{i}\right):=\frac{1}{n} \sum_{i=1}^{n} \sum_{k=1}^{K} \sum_{a \in \mathcal{A}} \gamma^{k} \cdot Q\left(s_{k}^{(i)}, a ; z\right) \nabla_{\theta} \pi_{\theta}\left(a \mid s_{k}^{(i)}\right)
\end{align*}
$$
コレを使うと、以下のように勾配を書くことができます。

$$
\begin{align*}
&\hat{\nabla}_{\theta} R\left(\pi_{\theta} ; \delta\right) \\
&:=\underset{x}{\operatorname{argmax}} \inf _{\|z\|_{\infty} \leq \ell_{F}}\left\{-F^{*}(z)+\tilde{V}(\theta ; z)+\delta \nabla_{\theta} \tilde{V}(\theta ; z)^{\top} x-\frac{\delta}{2}\|x\|^{2}\right\}, \tag{3.6}
\end{align*}
$$

では、この推定値の誤差バウンドを考えてみましょう。

**定理3.2：方策勾配推定の誤差バウンド**
以下が成り立つと仮定します。
* $\operatorname{dom} F=\mathbb{R}^{S A}$であり、$\max \left\{\|\nabla F(\lambda)\|_{\infty}:\|\lambda\|_{1} \leq \frac{2}{1-\gamma}\right\} \leq \ell_{F}$となるような$\ell_{F}$が存在します。つまり有界です。
* $F$は$L_1$ノルムの下で$L_F$-平滑です。すなわち、以下が成立します。
$$\left\|\nabla F(\lambda)-\nabla F\left(\lambda^{\prime}\right)\right\|_{\infty} \leq L_{F}\left\|\lambda-\lambda^{\prime}\right\|_{1}$$
* $F^*$は、集合$\left\{z:\|z\|_{\infty} \leq 2 \ell_{F}, F^{*}(z)>-\infty\right\}$において、$L_{\infty}$ノルムに関して($\ell_{F^{*}}$)-リプシッツ連続です。
*  $\left\|\nabla_{\theta} \pi(\cdot \mid s)\right\|_{\infty, 2} \leq C$となるような$C$が存在するとします。
以下の定義を示します。
* $\nabla_{\theta} \pi(\cdot \mid s)=\left[\nabla_{\theta} \pi(1 \mid s), \cdots, \nabla_{\theta} \pi(A \mid s)\right]$
* $\hat{\nabla}_{\theta} R\left(\pi_{\theta}\right):=\lim _{\delta \rightarrow 0_{+}} \hat{\nabla}_{\theta} R\left(\pi_{\theta} ; \delta\right)$

このとき、以下が成立します。
$$
\mathbb{E}\left[\left\|\hat{\nabla}_{\theta} R\left(\pi_{\theta}\right)-\nabla_{\theta} R\left(\pi_{\theta}\right)\right\|^{2}\right] \leq \mathcal{O}\left(\frac{C^{2}\left(\ell_{F}^{2}+L_{F}^{2} \ell_{F^{*}}^{2}\right)}{n(1-\gamma)^{4}}+\frac{C^{2} L_{F}^{2}}{n(1-\gamma)^{6}}\right)+\mathcal{O}\left(\gamma^{K}\right)
$$

**定理3.2の証明**

**証明の全体戦略**

証明はいくつかのステップに分かれています。

**ステップ1**

真の勾配 $\nabla_{\theta} R(\pi_{\theta})$ とその推定量 $\hat{\nabla}_{\theta} R(\pi_{\theta})$ を直接比較するのは困難です。そこで、定理3.1で導入された変分表現を利用します。

*   **真の勾配の近似**:
    真の勾配は、次のように小さな正の数 $\delta$ を用いたmin-max問題の解の極限として表現されました。
$$ \nabla_{\theta} R(\pi_{\theta}) = \lim_{\delta \to 0_+} x^*(\delta) $$
    ここで、$(x^*(\delta), z^*(\delta))$ は以下の「真の」min-max問題の解です。
$$ (x^{*}(\delta), z^{*}(\delta)) := \underset{x}{\operatorname{argmax}} \underset{\|z\|_{\infty} \leq \ell_{F}}{\operatorname{argmin}} \left\{ V(\theta ; z)+\delta \nabla_{\theta} V(\theta ; z)^{\top} x-F^{*}(z)-\frac{\delta}{2}\|x\|^{2} \right\} \quad \text{(E.1)} $$
    この $x^*(\delta)$ を、真の勾配の「$\delta$-近似」と見なします。

*   **推定量の定義**:
    同様に、勾配の推定量 $\hat{\nabla}_{\theta} R(\pi_{\theta})$ は、$V$ や $\nabla_\theta V$ をサンプルから推定した $\tilde{V}$ や $\nabla_\theta \tilde{V}$ で置き換えた「経験的」min-max問題の解 $\hat{x}(\delta)$ の極限として定義されます。
    $$ \hat{\nabla}_{\theta} R(\pi_{\theta}) = \lim_{\delta \to 0_+} \hat{x}(\delta) $$
    ここで、$(\hat{x}(\delta), \hat{z}(\delta))$ は以下の問題の解です。
    $$ (\hat{x}(\delta), \hat{z}(\delta)) := \underset{x}{\operatorname{argmax}} \underset{\|z\|_{\infty} \leq \ell_{F}}{\operatorname{argmin}} \left\{ \tilde{V}(\theta ; z)+\delta \nabla_{\theta} \tilde{V}(\theta ; z)^{\top} x-F^{*}(z)-\frac{\delta}{2}\|x\|^{2} \right\} \quad \text{(E.2)} $$

**ステップ2**

評価したいのは $\mathbb{E}[\|\hat{\nabla}_{\theta} R - \nabla_{\theta} R\|^2]$ です。$\delta \to 0$ の極限を最後に取るとして、まずは固定された $\delta > 0$ に対する $\mathbb{E}[\|\hat{x}(\delta) - \nabla_{\theta} R\|^2]$ を評価します。ここで、中間目標である $x^*(\delta)$ を使って、三角不等式（より正確には、ベクトルのノルムの性質 $\|a+b\|^2 \le 2\|a\|^2 + 2\|b\|^2$）を用いて誤差を分解します。

$$
\begin{aligned}
\mathbb{E}[\|\hat{x}(\delta) - \nabla_{\theta} R\|^2] &= \mathbb{E}[\|(\hat{x}(\delta) - x^*(\delta)) + (x^*(\delta) - \nabla_{\theta} R)\|^2] \\
&\le 2 \underbrace{\mathbb{E}[\|\hat{x}(\delta) - x^*(\delta)\|^2]}_{\text{Term 1: 推定誤差}} + 2 \underbrace{\mathbb{E}[\|x^*(\delta) - \nabla_{\theta} R\|^2]}_{\text{Term 2: 近似誤差}}
\end{aligned}
$$

*   Term 1 (推定誤差): 真のmin-max問題の解 $x^*(\delta)$ と、経験的min-max問題の解 $\hat{x}(\delta)$ との差です。これは、真の期待値 $V, \nabla_\theta V$ を有限個のサンプルによる推定量 $\tilde{V}, \nabla_\theta \tilde{V}$ で置き換えたことによって生じる誤差、つまり**サンプリング誤差**です。

*   Term 2 (近似誤差): 真の勾配 $\nabla_{\theta} R$ を、$\delta$ を用いた近似 $x^*(\delta)$ で置き換えたことによる誤差です。これは、サンプリングとは無関係な、理論的な**近似の誤差**です。

**ステップ3**

証明は、これら2つの誤差項をそれぞれ評価することです。その結果をまとめたものが補題E.1です。

*   **補題E.1(i)** は、近似誤差 Term 2 を評価します。結果は $\|x^*(\delta) - \nabla_{\theta} R\|^2 = \mathcal{O}(\delta^2)$ となり、$\delta$ が小さいほど近似が良くなることを示します。

*   **補題E.1(ii), (iii)** は、推定誤差 Term 1 を評価するためのものです。
    *   補題E.1(ii)は、主変数 $x$ の推定誤差 $\|\hat{x}(\delta) - x^*(\delta)\|^2$ を、双対変数 $z$ の推定誤差 $\|\hat{z}(\delta) - z^*(\delta)\|^2$ で上から抑えます。これにより、問題は $z$ の誤差評価に帰着されます。
    *   補題E.1(iii)は、その双対変数 $z$ の推定誤差 $\mathbb{E}[\|\hat{z}(\delta) - z^*(\delta)\|^2]$ を評価します。この部分は最も技術的に高度で、**確率的鞍点問題の一般化理論**という外部の結果を利用します。

**ステップ4**
最後に、補題E.1で得られた各項の上界を元の分解式に代入し、式を整理します。そして、$\delta \to 0$ の極限を取ることで、最終的な定理3.2の主張を得ます。

この全体像を念頭に置きながら、各ステップの証明を見ていきましょう。

### 2. E.1 予備的な技術的事項の詳細解説

#### 2.1 線形性

**主張**: 関数 $Q^{\pi_\theta}(s, a; r)$, $V(\theta; r)$, $\nabla_\theta V(\theta; r)$ は、報酬関数 $r$ (ここでは一般化された報酬として $z \in \mathbb{R}^{|\mathcal{S}||\mathcal{A}|}$ を考えます) に関して線形です。

**詳細な説明**:
線形性とは、入力のスカラー倍と和が、出力のスカラー倍と和にそのまま対応することを意味します。数式で書くと、任意のスカラー $\alpha, \alpha'$ とベクトル $z, z'$ に対して、
$$ f(\alpha z + \alpha' z') = \alpha f(z) + \alpha' f(z') $$
が成り立つことです。

まず、$V(\theta; z)$ の定義を思い出しましょう。
$$ V(\theta; z) = \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t z(s_t, a_t) \right] $$
ここで、$z(s_t, a_t)$ は状態-行動ペア $(s_t, a_t)$ に対応する $z$ の成分 $z_{s_t a_t}$ を表します。期待値操作は線形なので、
$$
\begin{aligned}
V(\theta; \alpha z + \alpha' z') &= \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t (\alpha z + \alpha' z')(s_t, a_t) \right] \\
&= \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t (\alpha z(s_t, a_t) + \alpha' z'(s_t, a_t)) \right] \\
&= \mathbb{E}^{\pi_\theta} \left[ \alpha \sum_{t=0}^\infty \gamma^t z(s_t, a_t) + \alpha' \sum_{t=0}^\infty \gamma^t z'(s_t, a_t) \right] \\
&= \alpha \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t z(s_t, a_t) \right] + \alpha' \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t z'(s_t, a_t) \right] \\
&= \alpha V(\theta; z) + \alpha' V(\theta; z')
\end{aligned}
$$
となり、$V(\theta; z)$ の線形性が示されました。$Q^{\pi_\theta}(s, a; z)$ についても、期待値の線形性から同様に証明できます。

次に、勾配 $\nabla_\theta V(\theta; z)$ です。微分操作も線形なので、
$$ \nabla_\theta (\alpha f(\theta) + \alpha' g(\theta)) = \alpha \nabla_\theta f(\theta) + \alpha' \nabla_\theta g(\theta) $$
が成り立ちます。これと $V(\theta; z)$ の線形性を組み合わせると、
$$ \nabla_\theta V(\theta; \alpha z + \alpha' z') = \nabla_\theta (\alpha V(\theta; z) + \alpha' V(\theta; z')) = \alpha \nabla_\theta V(\theta; z) + \alpha' \nabla_\theta V(\theta; z') $$
となり、$\nabla_\theta V(\theta; z)$ の線形性も示されます。

**なぜ重要か？**: この性質は、補題E.1(ii)の証明の後半で役割を果たします。そこでは、$\nabla_\theta \tilde{V}(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; \hat{z})$ という差を考えますが、線形性によってこれを $\nabla_\theta \tilde{V}(\theta; z^* - \hat{z})$ と一つの項にまとめることができます。これにより、誤差の評価が非常に簡潔になります。

#### 2.2 $Q$と$V$の上界

**主張**: 任意の報酬関数 $r$ (または $z$) に対して、
$$ |Q^{\pi_\theta}(s, a; r)| \le \frac{\|r\|_\infty}{1-\gamma} \quad \text{and} \quad |V(\theta; r)| \le \frac{\|r\|_\infty}{1-\gamma} $$
が成り立ちます。ここで、$\|r\|_\infty = \max_{s,a} |r(s,a)|$ は報酬の絶対値の最大値です。

**詳細な説明**:
この上界は、割引報酬和が発散しないことを保証する基本的な性質です。証明は、三角不等式と等比級数の和の公式を用います。

まず、$V(\theta; r)$ から見てみましょう。
$$
\begin{aligned}
|V(\theta; r)| &= \left| \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t r(s_t, a_t) \right] \right| \\
&\le \mathbb{E}^{\pi_\theta} \left[ \left| \sum_{t=0}^\infty \gamma^t r(s_t, a_t) \right| \right] \quad (\text{イェンセンの不等式, } |\cdot| \text{は凸関数}) \\
&\le \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty |\gamma^t r(s_t, a_t)| \right] \quad (\text{三角不等式}) \\
&= \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t |r(s_t, a_t)| \right]
\end{aligned}
$$
ここで、任意の状態・行動 $(s_t, a_t)$ において、報酬の絶対値は最大値 $\|r\|_\infty$ で抑えられます。つまり、$|r(s_t, a_t)| \le \|r\|_\infty$ です。
$$
\begin{aligned}
|V(\theta; r)| &\le \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t \|r\|_\infty \right] \\
&= \|r\|_\infty \cdot \mathbb{E}^{\pi_\theta} \left[ \sum_{t=0}^\infty \gamma^t \right] \quad (\|r\|_\infty \text{は定数}) \\
&= \|r\|_\infty \cdot \sum_{t=0}^\infty \gamma^t \\
&= \|r\|_\infty \cdot \frac{1}{1-\gamma} \quad (\text{等比級数の和、} 0 \le \gamma < 1)
\end{aligned}
$$
$Q^{\pi_\theta}(s, a; r)$ についても、期待値の対象が $s_0=s, a_0=a$ から始まる軌道に限定されるだけで、全く同じ議論が成り立ちます。

**なぜ重要か？**: この上界は、証明の様々な箇所で、値が「爆発」しないことを保証するために使われます。特に、後述する推定量の上界を導出する際の基礎となります。

#### 2.3 推定量の一様な上界

**主張**: 有限長 $K$ の単一のサンプルパス $\zeta$ から計算される推定量は、次のように上界されます。
$$ |\tilde{V}(\theta ; z ; \zeta)| \leq \frac{\|z\|_{\infty}}{1-\gamma} \quad \text{and} \quad \left\|\nabla_{\theta} \tilde{V}(\theta ; z ; \zeta)\right\| \leq \frac{C\|z\|_{\infty}}{(1-\gamma)^{2}} \quad \text{(E.3)} $$
この結果、複数サンプル ($n$個) の平均である $\tilde{V}(\theta; z)$ と $\nabla_\theta \tilde{V}(\theta; z)$ も同じ上界を持ちます (E.4)。

**詳細な説明**:
まず、$\tilde{V}(\theta; z; \zeta)$ の上界を導出します。定義は以下の通りです。
$$ \tilde{V}(\theta; z; \zeta) = \sum_{k=0}^K \gamma^k z(s_k, a_k) $$
この絶対値を取ると、
$$
\begin{aligned}
|\tilde{V}(\theta; z; \zeta)| &= \left| \sum_{k=0}^K \gamma^k z(s_k, a_k) \right| \\
&\le \sum_{k=0}^K |\gamma^k z(s_k, a_k)| \quad (\text{三角不等式}) \\
&= \sum_{k=0}^K \gamma^k |z(s_k, a_k)| \\
&\le \sum_{k=0}^K \gamma^k \|z\|_\infty \quad (|z(s_k, a_k)| \le \|z\|_\infty) \\
&= \|z\|_\infty \sum_{k=0}^K \gamma^k
\end{aligned}
$$
ここで、有限等比級数の和 $\sum_{k=0}^K \gamma^k = \frac{1-\gamma^{K+1}}{1-\gamma} \le \frac{1}{1-\gamma}$ を使います。
$$ |\tilde{V}(\theta; z; \zeta)| \le \|z\|_\infty \cdot \frac{1}{1-\gamma} $$
となり、最初の不等式が示されました。

次に、より複雑な $\nabla_\theta \tilde{V}(\theta; z; \zeta)$ の上界を導出します。その定義は、
$$ \nabla_{\theta} \tilde{V}(\theta ; z ; \zeta) = \sum_{k=0}^{K} \gamma^{k} \cdot Q(s_{k}, a_{k} ; z) \nabla_{\theta} \log \pi_{\theta}(a_{k} \mid s_{k}) $$
という形式ではなく、論文では少し異なる（そしてより一般的な）形式を使っています。
$$ \nabla \tilde{V}(\theta ; z) := \frac{1}{n} \sum_{i=1}^{n} \sum_{k=1}^{K} \sum_{a \in \mathcal{A}} \gamma^{k} \cdot Q(s_{k}^{(i)}, a ; z) \nabla_{\theta} \pi_{\theta}(a \mid s_{k}^{(i)}) $$
この式は、方策勾配定理の「全行動」バージョンに基づいています。この形式で単一の軌道 $\zeta$ の場合を考えます。
$$ \nabla_\theta \tilde{V}(\theta; z; \zeta) = \sum_{k=1}^K \gamma^k \sum_{a \in \mathcal{A}} Q(s_k, a; z) \nabla_\theta \pi_\theta(a|s_k) $$
この2-ノルムを取ります。
$$
\begin{aligned}
\|\nabla_\theta \tilde{V}(\theta; z; \zeta)\| &= \left\| \sum_{k=1}^K \gamma^k \sum_{a \in \mathcal{A}} Q(s_k, a; z) \nabla_\theta \pi_\theta(a|s_k) \right\| \\
&\le \sum_{k=1}^K \gamma^k \left\| \sum_{a \in \mathcal{A}} Q(s_k, a; z) \nabla_\theta \pi_\theta(a|s_k) \right\| \quad (\text{三角不等式})
\end{aligned}
$$
ここで、内側の和に注目します。これは、ベクトル $u_a = Q(s_k, a; z)$ とベクトル $\nabla_\theta \pi_\theta(a|s_k)$ の線形結合です。
$s_k$ を固定して考えると、$\sum_{a \in \mathcal{A}} u_a \nabla_\theta \pi_\theta(a|s_k)$ という形をしています。
ここで、定理3.2の仮定(iv)を使います。
**(仮定(iv))**: $\left\|\nabla_{\theta} \pi(\cdot \mid s)\right\|_{\infty, 2} \leq C$
この $\|\cdot\|_{\infty, 2}$ は**誘導ノルム**であり、次のように定義されます。
$$ \|M\|_{\infty, 2} = \max_{\|u\|_\infty \le 1} \|Mu\|_2 $$
我々のケースでは、行列 $M$ は、各列が $\nabla_\theta \pi(a|s)$ ($a \in \mathcal{A}$) となる $d \times A$ 行列と見なせます。そして、この行列にベクトル $u = [u_1, \dots, u_A]^\top$ を右から掛けている形です。
$$ \left\| \sum_{a \in \mathcal{A}} u_a \nabla_\theta \pi_\theta(a|s_k) \right\| = \| (\nabla_\theta \pi(\cdot|s_k)) u \|_2 $$
ここで、$u_a = Q(s_k, a; z)$ です。$|Q(s_k, a; z)| \le \frac{\|z\|_\infty}{1-\gamma}$ でしたから、ベクトル $u$ の無限大ノルムは $\|u\|_\infty = \max_a |Q(s_k, a; z)| \le \frac{\|z\|_\infty}{1-\gamma}$ です。
誘導ノルムの定義から、
$$ \|(\nabla_\theta \pi(\cdot|s_k)) u \|_2 \le \| \nabla_\theta \pi(\cdot|s_k) \|_{\infty, 2} \cdot \|u\|_\infty $$
が成り立ちます。これに仮定(iv)と $Q$ の上界を代入すると、
$$ \left\| \sum_{a \in \mathcal{A}} Q(s_k, a; z) \nabla_\theta \pi_\theta(a|s_k) \right\| \le C \cdot \frac{\|z\|_\infty}{1-\gamma} $$
この結果を元の不等式に戻します。
$$
\begin{aligned}
\|\nabla_\theta \tilde{V}(\theta; z; \zeta)\| &\le \sum_{k=1}^K \gamma^k \left( C \cdot \frac{\|z\|_\infty}{1-\gamma} \right) \\
&= \frac{C \|z\|_\infty}{1-\gamma} \sum_{k=1}^K \gamma^k \\
&\le \frac{C \|z\|_\infty}{1-\gamma} \sum_{k=0}^\infty \gamma^k \quad (\text{和の範囲を広げても不等式は成り立つ}) \\
&= \frac{C \|z\|_\infty}{1-\gamma} \cdot \frac{1}{1-\gamma} = \frac{C \|z\|_\infty}{(1-\gamma)^2}
\end{aligned}
$$
これで、(E.3)の2番目の不等式も示されました。

最後に、(E.4)についてですが、$\tilde{V}(\theta; z) = \frac{1}{n} \sum_i \tilde{V}(\theta; z; \zeta_i)$ ですから、
$$ |\tilde{V}(\theta; z)| = \left| \frac{1}{n} \sum_i \tilde{V}(\theta; z; \zeta_i) \right| \le \frac{1}{n} \sum_i |\tilde{V}(\theta; z; \zeta_i)| \le \frac{1}{n} \sum_i \frac{\|z\|_\infty}{1-\gamma} = \frac{\|z\|_\infty}{1-\gamma} $$
となり、同じ上界が成り立ちます。$\nabla_\theta \tilde{V}(\theta; z)$ についても同様です。

**なぜ重要か？**: この上界は、サンプリングによって得られる推定量が、サンプルパスの具体的な内容によらず、一様に有界であることを保証します。これは、後の誤差評価において、最悪ケースのバウンドを導出する上で不可欠です。特に、補題E.1(ii)のバリアンス項の評価や、補題E.1(iii)で実行可能領域をコンパクトにする際に直接的に使われます。

### 3. E.2 補題E.1(i)の証明（近似誤差の評価）

**補題E.1(i)の主張**:
真の勾配 $\nabla_{\theta} R(\pi_{\theta})$ と、その$\delta$-近似である $x^*(\delta)$ との間の二乗誤差は、$\delta^2$ のオーダーである。
$$ \left\|x^{*}(\delta)-\nabla_{\theta} R\left(\pi_{\theta}\right)\right\|^{2}=\mathcal{O}\left(\delta^{2}\right) $$

**詳細な証明**:
この補題は、我々の近似戦略が数学的に妥当であることを示します。$\delta$ をゼロに近づけることで、いくらでも真の勾配に近づけるだけでなく、その収束が二次（$\delta^2$）の速さで起こることを保証します。

**ステップ1: 最適性条件の確認**

まず、$x^*(\delta)$ が満たすべき条件を導出します。$x^*(\delta)$ は、次のmin-max問題の $x$ に関する解です。
$$ x^*(\delta) = \underset{x}{\operatorname{argmax}} \left( \min_{\|z\|_\infty \le \ell_F} \left\{ V(\theta ; z)+\delta \nabla_{\theta} V(\theta ; z)^{\top} x-F^{*}(z)-\frac{\delta}{2}\|x\|^{2} \right\} \right) $$
内側の $z$ に関する最小化問題を先に解きます。この問題は、$z$ に関して凸です（$V, \nabla_\theta V$ は $z$ に線形、$-F^*$ は凸）。最適解は、勾配がゼロになる点（厳密には劣勾配がゼロを含む点）で与えられます。
$z$ に関する勾配（劣勾配）を計算すると、
$$ \nabla_z \left( V(\theta; z) + \delta \nabla_\theta V(\theta; z)^\top x - F^*(z) \right) = \nabla_z V(\theta; z) + \delta \nabla_z (\nabla_\theta V(\theta; z)^\top x) - \nabla_z F^*(z) $$
ここで、$V(\theta; z) = \langle \lambda(\theta), z \rangle$ でしたから、$\nabla_z V(\theta; z) = \lambda(\theta)$ です。同様に、$\nabla_\theta V(\theta; z) = \nabla_\theta \langle \lambda(\theta), z \rangle = (\nabla_\theta \lambda(\theta))^\top z$ ではないことに注意が必要です。正しくは、$\nabla_\theta V(\theta;z)$ は $z$ を引数に取る線形作用素です。$z$ に関する勾配を考えるには、$V(\theta; z) + \delta \langle \nabla_\theta V(\theta;z), x \rangle = \langle \lambda(\theta), z \rangle + \delta \langle \nabla_\theta \langle \lambda(\theta), z \rangle, x \rangle = \langle \lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x, z \rangle$ となります。
したがって、$z$に関する最小化問題の目的関数は $\langle \lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x, z \rangle - F^*(z)$ となります。
この最小化問題の解 $z^*(x, \delta)$ は、Fenchel双対の定義から、
$$ z^*(x, \delta) = \nabla F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x) $$
となります。これは、もし $G(y) = F(y)$ ならば、$G^*(z) = F^*(z)$ であり、$\nabla F(y) = z \iff \nabla F^*(z) = y$ という双対関係から導かれます。

この $z^*(x, \delta)$ を元の問題に代入すると、$x$ に関する最大化問題は次のようになります。
$$ \underset{x}{\operatorname{argmax}} \left\{ F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x) - \frac{\delta}{2}\|x\|^2 \right\} $$
この最大化問題の解が $x^*(\delta)$ です。この問題は $x$ に関して（一般には非凸ですが）微分可能なので、最適性条件として勾配がゼロになります。
$$ \nabla_x \left( F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x) - \frac{\delta}{2}\|x\|^2 \right) \bigg|_{x=x^*(\delta)} = 0 $$
連鎖律を用いて勾配を計算すると、
$$ (\nabla_\theta \lambda(\theta))^\top \nabla F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta)) \cdot \delta - \delta x^*(\delta) = 0 $$
$\delta > 0$ なので、両辺を $\delta$ で割ると、
$$ x^*(\delta) = (\nabla_\theta \lambda(\theta))^\top \nabla F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta)) $$
これが、$x^*(\delta)$ が満たす重要な方程式です。

**ステップ2: 制約 $\|z\|_\infty \le \ell_F$ が有効であることの確認**

上記の導出は、制約 $\|z\|_\infty \le \ell_F$ を無視して行いました。この制約が最適解に影響しない（非アクティブである）ことを示す必要があります。
最適解における $z$ の値は $z^*(\delta) = \nabla F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta))$ です。
我々は、この $z^*(\delta)$ が $\|z^*(\delta)\|_\infty \le \ell_F$ を満たすことを示したいのです。

まず、定理3.1から、$\lim_{\delta \to 0_+} x^*(\delta) = \nabla_\theta R(\pi_\theta)$ であることがわかっています。$\nabla_\theta R(\pi_\theta)$ は有界なベクトルなので、$x^*(\delta)$ も $\delta$ が小さい領域で有界です。
したがって、$\delta \to 0$ のとき、
$$ \lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta) \to \lambda(\theta) $$
となります。$\lambda(\theta)$ は有効な占有尺度なので、その $L_1$ ノルムは $\|\lambda(\theta)\|_1 = \sum_{s,a} \lambda_{sa}(\theta) = \sum_{t=0}^\infty \gamma^t = \frac{1}{1-\gamma}$ です。
連続性から、十分に小さい $\delta_0 > 0$ を選べば、全ての $\delta < \delta_0$ に対して、
$$ \|\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta)\|_1 \le \frac{2}{1-\gamma} $$
とすることができます。
ここで、定理3.2の**仮定(i)** を使います。
**(仮定(i))**: $\max\{\|\nabla F(\lambda)\|_\infty : \|\lambda\|_1 \le \frac{2}{1-\gamma}\} \le \ell_F$
この仮定は、占有尺度がある程度の範囲（元の占有尺度の近傍）にあれば、その点での効用関数の勾配の大きさ（無限大ノルム）が $\ell_F$ で抑えられることを要請しています。
上記の議論から、$\lambda' = \lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta)$ は $\|\lambda'\|_1 \le \frac{2}{1-\gamma}$ を満たすので、
$$ \|z^*(\delta)\|_\infty = \|\nabla F(\lambda')\|_\infty \le \ell_F $$
が成り立ちます。これは、制約なし問題の解が自動的に制約を満たすことを意味しており、したがって制約を追加しても解は変わらないことが正当化されました。

**ステップ3: 誤差の評価**

いよいよ誤差を評価します。真の勾配は、連鎖律から
$$ \nabla_\theta R(\pi_\theta) = \nabla_\theta F(\lambda(\theta)) = (\nabla_\theta \lambda(\theta))^\top \nabla F(\lambda(\theta)) $$
と書けます。
$x^*(\delta)$ の式と比較して、その差のノルム二乗を計算します。
$$
\begin{aligned}
\|x^*(\delta) - \nabla_\theta R(\pi_\theta)\|^2 &= \left\| (\nabla_\theta \lambda(\theta))^\top \nabla F(\lambda(\theta) + \delta (\nabla_\theta \lambda(\theta)) x^*(\delta)) - (\nabla_\theta \lambda(\theta))^\top \nabla F(\lambda(\theta)) \right\|^2 \\
&= \left\| (\nabla_\theta \lambda(\theta))^\top \left( \nabla F(\lambda(\theta) + \delta \nabla_\theta \lambda(\theta) x^*(\delta)) - \nabla F(\lambda(\theta)) \right) \right\|^2
\end{aligned}
$$
ここで、行列とベクトルの積のノルムに関する不等式 $\|AB\| \le \|A\| \|B\|$ を使います。ここでは、$\|A\|_{\infty, 2}$ のような誘導ノルムを使うのがより正確です。
$$ \|M v\|_2 \le \|M\|_{1,2} \|v\|_1 $$
あるいは、
$$ \|M^\top v\|_2 \le \|M\|_{\infty,2} \|v\|_\infty $$
という関係があります（行列 $M$ の転置 $M^\top$ の作用素ノルム）。論文の記法に従うと、
$$
\begin{aligned}
\|x^*(\delta) - \nabla_\theta R(\pi_\theta)\|^2 &\le \|\nabla_\theta \lambda(\theta)^\top\|_{\infty, 2}^2 \cdot \|\nabla F(\lambda(\theta) + \delta \nabla_\theta \lambda(\theta) x^*(\delta)) - \nabla F(\lambda(\theta))\|_\infty^2
\end{aligned}
$$
ここで、定理3.2の**仮定(ii)** を使います。
**(仮定(ii))**: $F$ は $L_1$ ノルムの下で $L_F$-平滑である。すなわち、
$$ \|\nabla F(\lambda) - \nabla F(\lambda')\|_\infty \le L_F \|\lambda - \lambda'\|_1 $$
この仮定を、$\lambda' = \lambda(\theta)$ と $\lambda = \lambda(\theta) + \delta \nabla_\theta \lambda(\theta) x^*(\delta)$ に適用すると、
$$ \|\nabla F(\lambda) - \nabla F(\lambda')\|_\infty \le L_F \|\delta \nabla_\theta \lambda(\theta) x^*(\delta)\|_1 $$
となります。これを上の不等式に代入すると、
$$
\begin{aligned}
\|x^*(\delta) - \nabla_\theta R(\pi_\theta)\|^2 &\le \|\nabla_\theta \lambda(\theta)^\top\|_{\infty, 2}^2 \cdot (L_F \|\delta \nabla_\theta \lambda(\theta) x^*(\delta)\|_1)^2 \\
&= L_F^2 \|\nabla_\theta \lambda(\theta)^\top\|_{\infty, 2}^2 \cdot \delta^2 \cdot \|\nabla_\theta \lambda(\theta) x^*(\delta)\|_1^2
\end{aligned}
$$
$x^*(\delta)$ は $\delta \to 0$ で有界な量に収束するため、$\delta$ が小さい領域では有界です。また、ヤコビ行列 $\nabla_\theta \lambda(\theta)$ も有界です。したがって、$\|\nabla_\theta \lambda(\theta)^\top\|_{\infty, 2}$ や $\|\nabla_\theta \lambda(\theta) x^*(\delta)\|_1$ も有界な定数です。
これらの定数部分をまとめて $C_{const}$ と書くと、
$$ \|x^*(\delta) - \nabla_\theta R(\pi_\theta)\|^2 \le C_{const} \cdot \delta^2 $$
これは、まさに $\mathcal{O}(\delta^2)$ であることを意味します。これで補題E.1(i)の証明が完了しました。

### 4. E.3 補題E.1(ii)の証明（主変数の推定誤差の評価）

**補題E.1(ii)の主張**:
主変数 $x$ の推定誤差は、双対変数 $z$ の推定誤差と、サンプリングに起因する項（打ち切り誤差と分散）で上から評価できる。
$$ \mathbb{E}\left[\left\|x^{*}-\hat{x}\right\|^{2}\right] \leq \frac{2 C^{2}\left\|z^{*}\right\|_{\infty}^{2}}{(1-\gamma)^{4}} \cdot\left(\frac{\gamma^{2 K}}{(1-\gamma)^{2}}+\frac{1}{n}\right)+\frac{2 C^{2}}{(1-\gamma)^{4}} \cdot \mathbb{E}\left[\left\|z^{*}-\hat{z}\right\|_{\infty}^{2}\right] $$

**詳細な証明**:
この補題は、証明戦略の中心的な部分です。直接評価するのが難しい $x$ の誤差を、より扱いやすい（と期待される）$z$ の誤差に問題をすり替える役割を果たします。

**ステップ1: 誤差の分解**

まず、評価したい量 $\mathbb{E}[\|x^* - \hat{x}\|^2]$ を考えます。ここで、$x^*$ と $\hat{x}$ の定義を思い出しましょう。
$$ x^{*} = \nabla_{\theta} V(\theta ; z^{*}) \quad \text{and} \quad \hat{x} = \nabla_{\theta} \tilde{V}(\theta ; \hat{z}) $$
これらは、それぞれのmin-max問題 (E.1) と (E.2) の最適性条件から得られる関係式です。
この差 $x^* - \hat{x}$ に、中間項として $\nabla_\theta \tilde{V}(\theta; z^*)$ を挿入します。
$$ x^* - \hat{x} = (\nabla_\theta V(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; z^*)) + (\nabla_\theta \tilde{V}(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; \hat{z})) $$
ノルム二乗の期待値を取り、不等式 $\|a+b\|^2 \le 2\|a\|^2 + 2\|b\|^2$ を適用すると、
$$ \mathbb{E}[\|x^* - \hat{x}\|^2] \le 2 \underbrace{\mathbb{E}[\|\nabla_\theta V(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; z^*)\|^2]}_{\text{Term A}} + 2 \underbrace{\mathbb{E}[\|\nabla_\theta \tilde{V}(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; \hat{z})\|^2]}_{\text{Term B}} $$
これで、評価すべき項が2つに分かれました。それぞれを個別に見ていきましょう。

**ステップ2: Term A の評価（固定された $z^*$ に対する推定誤差）**

Term A は、固定された（仮想的な）報酬関数 $z^*$ に対する方策勾配の真の値 $\nabla_\theta V(\theta; z^*)$ と、その推定量 $\nabla_\theta \tilde{V}(\theta; z^*)$ の間の平均二乗誤差です。これは、標準的なモンテカルロ推定の誤差評価に相当します。

ここで、統計学の基本的なツールである**バイアス-バリアンス分解**を用います。任意の推定量 $\hat{\theta}$ と真の値 $\theta$ に対して、平均二乗誤差は次のように分解できます。
$$ \mathbb{E}[\|\hat{\theta} - \theta\|^2] = \|\mathbb{E}[\hat{\theta}] - \theta\|^2 + \mathbb{E}[\|\hat{\theta} - \mathbb{E}[\hat{\theta}]\|^2] = (\text{Bias})^2 + \text{Variance} $$
我々のケースでは、$\theta = \nabla_\theta V(\theta; z^*)$、$\hat{\theta} = \nabla_\theta \tilde{V}(\theta; z^*)$ です。
$$ \text{Term A} = \underbrace{\|\mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)] - \nabla_\theta V(\theta; z^*)\|^2}_{\text{Squared Bias}} + \underbrace{\mathbb{E}[\|\nabla_\theta \tilde{V}(\theta; z^*) - \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)]\|^2]}_{\text{Variance}} $$

*   **バイアス項の評価**:
    バイアスは、推定量の期待値が真の値からどれだけずれているかを表します。この問題におけるバイアスの主な原因は、無限長の軌道を有限長 $K$ で**打ち切っている**ことです。
    まず、$\mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)]$ を計算します。
    $$ \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)] = \mathbb{E}\left[ \frac{1}{n} \sum_{i=1}^n \sum_{k=1}^K \sum_{a \in \mathcal{A}} \gamma^k Q(s_k^{(i)}, a; z^*) \nabla_\theta \pi_\theta(a|s_k^{(i)}) \right] $$
    期待値と和は交換できるので、
    $$ = \sum_{k=1}^K \gamma^k \mathbb{E}\left[ \sum_{a \in \mathcal{A}} Q(s_k, a; z^*) \nabla_\theta \pi_\theta(a|s_k) \right] $$
    状態 $s_k$ の分布は、初期状態分布 $\xi$ から方策 $\pi_\theta$ に従って $k$ ステップ遷移した後の分布です。これを $d_k(s) = \mathbb{P}(s_k=s | \pi_\theta, s_0 \sim \xi)$ と書くと、
    $$ \mathbb{E}[\dots] = \sum_{s \in \mathcal{S}} d_k(s) \sum_{a \in \mathcal{A}} Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) $$
    したがって、
    $$ \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)] = \sum_{k=1}^K \gamma^k \sum_{s \in \mathcal{S}} d_k(s) \sum_{a \in \mathcal{A}} Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) $$
    一方、真の値 $\nabla_\theta V(\theta; z^*)$ は、無限和で定義される定常分布 $d_\xi^\pi(s) = (1-\gamma) \sum_{t=0}^\infty \gamma^t d_t(s)$ を使って、
    $$ \nabla_\theta V(\theta; z^*) = \frac{1}{1-\gamma} \sum_{s \in \mathcal{S}} d_\xi^\pi(s) \sum_{a \in \mathcal{A}} Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) $$
    と書けます。
    ここで、$d_{\xi, K}^\pi(s) = (1-\gamma)\sum_{k=0}^K \gamma^k d_k(s)$ と定義すると、$\mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)]$ はこの $d_{\xi, K}^\pi(s)$ を使った形に近くなります。
    バイアスは、この $d_\xi^\pi(s)$ と $d_{\xi, K}^\pi(s)$ の差に起因します。
    よく知られた事実として、これらの分布の $L_1$ 距離は、
    $$ \sum_s |d_\xi^\pi(s) - d_{\xi, K}^\pi(s)| = (1-\gamma) \sum_s \left| \sum_{t=K+1}^\infty \gamma^t d_t(s) \right| \le (1-\gamma) \sum_{t=K+1}^\infty \gamma^t \sum_s d_t(s) = (1-\gamma) \frac{\gamma^{K+1}}{1-\gamma} = \gamma^{K+1} $$
    （論文では $\mathcal{O}(\frac{\gamma^K}{1-\gamma})$ となっていますが、これは少し緩い評価です。ここでは論文の導出を追います。）
    論文の(E.7)の導出は少し省略されていますが、詳細には以下のようになります。
    $$
    \begin{aligned}
    & \|\nabla_\theta V(\theta; z^*) - \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*)] \| \\
    &= \left\| \sum_{s} \left( \frac{d_\xi^\pi(s)}{1-\gamma} - \sum_{k=1}^K \gamma^k d_k(s) \right) \sum_a Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) \right\| \\
    &\approx \left\| \sum_{s} \frac{1}{1-\gamma}(d_\xi^\pi(s) - d_{\xi, K}^\pi(s)) \sum_a Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) \right\| \\
    &\le \frac{1}{1-\gamma} \sum_s |d_\xi^\pi(s) - d_{\xi, K}^\pi(s)| \left\| \sum_a Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) \right\| \\
    &\le \frac{1}{1-\gamma} \left(\sum_s |d_\xi^\pi(s) - d_{\xi, K}^\pi(s)|\right) \max_s \left\| \sum_a Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) \right\|
    \end{aligned}
    $$
    前のセクションで示したように、$\left\| \sum_a Q(s, a; z^*) \nabla_\theta \pi_\theta(a|s) \right\| \le \frac{C\|z^*\|_\infty}{1-\gamma}$ でした。
    また、分布の差は $\sum_s |d_\xi^\pi(s) - d_{\xi, K}^\pi(s)|$ が $\mathcal{O}(\gamma^K)$ で抑えられます。
    これらを組み合わせると、
    $$ \|\text{Bias}\| \le \frac{1}{1-\gamma} \cdot \mathcal{O}(\gamma^K) \cdot \frac{C\|z^*\|_\infty}{1-\gamma} = \mathcal{O}\left(\frac{C\|z^*\|_\infty \gamma^K}{(1-\gamma)^2}\right) $$
    したがって、二乗バイアスは $\mathcal{O}\left(\frac{C^2\|z^*\|_\infty^2 \gamma^{2K}}{(1-\gamma)^4}\right)$ となります。（論文の(E.7)の最後の式の $(1-\gamma)^6$ は、途中の係数の違いから来ている可能性がありますが、オーダーとしては同様です）

*   **バリアンス項の評価**:
    バリアンスは、推定量のサンプルごとのばらつきを表します。
    $$ \text{Variance} = \mathbb{E}\left[ \left\| \frac{1}{n}\sum_{i=1}^n \nabla_\theta \tilde{V}(\theta; z^*; \zeta_i) - \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*; \zeta_i)] \right\|^2 \right] $$
    ここで、各サンプル $\zeta_i$ は独立同分布(i.i.d.)です。i.i.d.な確率変数 $X_i$ の平均 $\bar{X} = \frac{1}{n}\sum_i X_i$ の分散は、$\mathrm{Var}(\bar{X}) = \frac{1}{n}\mathrm{Var}(X_i)$ となります。ベクトルでも同様で、
    $$ \text{Variance} = \frac{1}{n} \mathbb{E}\left[ \left\| \nabla_\theta \tilde{V}(\theta; z^*; \zeta_1) - \mathbb{E}[\nabla_\theta \tilde{V}(\theta; z^*; \zeta_1)] \right\|^2 \right] $$
    ここで、$\mathbb{E}[\|X - \mathbb{E}[X]\|^2] \le \mathbb{E}[\|X\|^2]$ という簡単な不等式を使います。
    $$ \text{Variance} \le \frac{1}{n} \mathbb{E}\left[ \left\| \nabla_\theta \tilde{V}(\theta; z^*; \zeta_1) \right\|^2 \right] $$
    ここで、(E.3)で示した推定量の一様な上界を使います。
    $$ \|\nabla_\theta \tilde{V}(\theta; z^*; \zeta_1)\| \le \frac{C\|z^*\|_\infty}{(1-\gamma)^2} $$
    この不等式は確率変数である $\zeta_1$ に依存しない deterministic な上界なので、期待値を取っても変わりません。
    $$ \mathbb{E}\left[ \left\| \nabla_\theta \tilde{V}(\theta; z^*; \zeta_1) \right\|^2 \right] \le \left( \frac{C\|z^*\|_\infty}{(1-\gamma)^2} \right)^2 = \frac{C^2\|z^*\|_\infty^2}{(1-\gamma)^4} $$
    したがって、
    $$ \text{Variance} \le \frac{C^2\|z^*\|_\infty^2}{n(1-\gamma)^4} $$
    となります。

バイアスとバリアンスの評価をまとめると、Term A は次のように評価できます。
$$ \text{Term A} \le \mathcal{O}\left(\frac{C^2\|z^*\|_\infty^2 \gamma^{2K}}{(1-\gamma)^6}\right) + \frac{C^2\|z^*\|_\infty^2}{n(1-\gamma)^4} $$
これは論文の式と整合しています。（係数の違いは無視できる範囲です）

**ステップ3: Term B の評価（$z$ の変動による誤差）**

Term B は、$\mathbb{E}[\|\nabla_\theta \tilde{V}(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; \hat{z})\|^2]$ です。この項は、$z$ が真の値 $z^*$ から推定値 $\hat{z}$ に変わったことで、勾配の推定量がどれだけ変化するかを表しています。

ここで、予備事項で確認した**線形性**が威力を発揮します。
$$ \nabla_\theta \tilde{V}(\theta; z^*) - \nabla_\theta \tilde{V}(\theta; \hat{z}) = \nabla_\theta \tilde{V}(\theta; z^* - \hat{z}) $$
したがって、
$$ \text{Term B} = \mathbb{E}[\|\nabla_\theta \tilde{V}(\theta; z^* - \hat{z})\|^2] $$
これに、(E.4)で示した推定量の上界を適用します。引数が $z$ から $z^* - \hat{z}$ に変わっただけなので、
$$ \|\nabla_\theta \tilde{V}(\theta; z^* - \hat{z})\| \le \frac{C\|z^* - \hat{z}\|_\infty}{(1-\gamma)^2} $$
この不等式は、確率変数であるサンプル $\{\zeta_i\}$ と、それから決まる $\hat{z}$ に依存します。両辺を二乗して期待値を取ると、
$$ \mathbb{E}[\|\nabla_\theta \tilde{V}(\theta; z^* - \hat{z})\|^2] \le \mathbb{E}\left[ \left( \frac{C\|z^* - \hat{z}\|_\infty}{(1-\gamma)^2} \right)^2 \right] = \frac{C^2}{(1-\gamma)^4} \mathbb{E}[\|z^* - \hat{z}\|_\infty^2] $$
これで Term B が双対変数 $z$ の推定誤差で評価できました。

**ステップ4: 全ての項を結合**

ステップ2と3の結果を、ステップ1の分解式に代入します。
$$
\begin{aligned}
\mathbb{E}[\|x^* - \hat{x}\|^2] &\le 2 \cdot \text{Term A} + 2 \cdot \text{Term B} \\
&\le 2 \left( \mathcal{O}\left(\frac{C^2\|z^*\|_\infty^2 \gamma^{2K}}{(1-\gamma)^6}\right) + \frac{C^2\|z^*\|_\infty^2}{n(1-\gamma)^4} \right) + 2 \left( \frac{C^2}{(1-\gamma)^4} \mathbb{E}[\|z^* - \hat{z}\|_\infty^2] \right)
\end{aligned}
$$
$\mathcal{O}$記法を使い、係数をまとめると、補題E.1(ii)の主張そのものが得られます。
$$ \mathbb{E}\left[\left\|x^{*}-\hat{x}\right\|^{2}\right] \leq \frac{2 C^{2}\left\|z^{*}\right\|_{\infty}^{2}}{(1-\gamma)^{4}} \cdot\left(\frac{\gamma^{2 K}}{(1-\gamma)^{2}}+\frac{1}{n}\right)+\frac{2 C^{2}}{(1-\gamma)^{4}} \cdot \mathbb{E}\left[\left\|z^{*}-\hat{z}\right\|_{\infty}^{2}\right] $$
これで補題E.1(ii)の証明が完了しました。

### 5. E.4 補題E.1(iii)の証明（双対変数の推定誤差の評価）

**補題E.1(iii)の主張**:
双対変数 $z$ の推定誤差は、サンプル数 $n$ に反比例する項で上から評価できる。
$$ \mathbb{E}\left[\left\|\hat{z}(\delta)-z^{*}(\delta)\right\|_{\infty}^{2}\right] \leq \mathcal{O}\left(\frac{L_{F}^{2}}{n(1-\gamma)^{2}}+\frac{L_{F}^{2} \ell_{F^{*}}^{2}}{n}+\frac{L_{F}^{2} \delta^{2}+L_{F} \delta}{n}\right) $$

**詳細な証明**:
この証明は、本論文で最も技術的に高度な部分です。鞍点問題 (E.1) と (E.2) を、それぞれ「真の（期待値上の）問題」と「経験的な（サンプル平均上の）問題」とみなし、両者の解の間の距離を評価します。これは、機械学習における**一般化誤差**の解析と密接に関連しており、ここでは確率的鞍点問題に関する専門的な結果（Zhang et al., 2020b）を「ブラックボックス」として利用します。ここでは、その理論を適用するための準備と、パラメータのマッピングを丁寧に解説します。

**ステップ1: 問題設定の確認と理論適用の準備**

我々が比較したいのは、以下の二つの問題の解です。
*   **真の問題 (E.1)**: 解 $(x^*, z^*)$
    $$ \max_x \min_z \left\{ \mathbb{E}_\zeta[\Psi_\zeta(x, z)] \right\} $$
    （ここで $\Psi_\zeta$ はサンプル $\zeta$ 上の目的関数）
*   **経験的問題 (E.2)**: 解 $(\hat{x}, \hat{z})$
    $$ \max_x \min_z \left\{ \frac{1}{n} \sum_{i=1}^n \Psi_{\zeta_i}(x, z) \right\} $$

Zhang et al. (2020b) の理論は、このような状況で解の間の距離、特に $\mathbb{E}[\|\hat{z} - z^*\|^2]$ を評価するためのツールを提供します。しかし、この理論を適用するには、いくつかの技術的な条件を満たす必要があります。その一つが、最適化変数の定義域が**コンパクト（有界かつ閉）**であることです。

我々の問題 (E.1), (E.2) では、$z$ の定義域 $\mathcal{Z} = \{z: \|z\|_\infty \le \ell_F\}$ は既にコンパクトです。しかし、$x$ の定義域は $\mathbb{R}^d$ であり、コンパクトではありません。そこで、解の性質を利用して、最適化の範囲を実質的にコンパクトな集合に限定できることを示します。

予備事項(E.4)から、
$$ \|\nabla_\theta \tilde{V}(\theta; z)\| \le \frac{C\|z\|_\infty}{(1-\gamma)^2} $$
でした。$\hat{x}$ の最適性条件は $\hat{x} = \nabla_\theta \tilde{V}(\theta; \hat{z})$ であり、$\hat{z} \in \mathcal{Z}$ なので $\|\hat{z}\|_\infty \le \ell_F$ です。したがって、
$$ \|\hat{x}\| \le \frac{C\|\hat{z}\|_\infty}{(1-\gamma)^2} \le \frac{C \ell_F}{(1-\gamma)^2} $$
となります。これは、経験的問題の解 $\hat{x}$ が、サンプルに依らず、必ずこの半径の球内に収まることを意味します。同様の議論が真の問題の解 $x^*$ にも成り立ちます。
$$ \|x^*\| \le \frac{C \ell_F}{(1-\gamma)^2} $$
したがって、最初から $x$ の探索範囲を $\mathcal{X} = \{x: \|x\| \le \frac{C \ell_F}{(1-\gamma)^2}\}$ に限定しても、解は全く変わりません。これで、$x, z$ 両方の定義域がコンパクトになり、理論を適用する準備が整いました。

**ステップ2: Zhang et al. (2020b)の理論の要求事項の確認**

Zhang et al. (2020b)の定理1を適用するには、目的関数 $\Psi_\zeta(x, z)$ が以下の2つの主要な性質を持つことを確認する必要があります。

*   **性質1: 強凹性-強凸性**
    *   $\Psi_\zeta(x, z)$ は、固定された $z$ に対して、$x$ の**強凹関数**である必要があります。
    *   $\Psi_\zeta(x, z)$ は、固定された $x$ に対して、$z$ の**強凸関数**である必要があります。

    我々の目的関数は、
    $$ \Psi_{\zeta}(x, z) = \tilde{V}(\theta ; z ; \zeta)+\delta \nabla_{\theta} \tilde{V}(\theta ; z ; \zeta)^{\top} x - F^{*}(z) - \frac{\delta}{2}\|x\|^{2} $$
    です。
    *   **$x$ に関する強凹性**: $x$ を含む項は $\delta \nabla_{\theta} \tilde{V}(\theta ; z ; \zeta)^{\top} x - \frac{\delta}{2}\|x\|^{2}$ です。これは $x$ の二次関数であり、ヘシアンは $-\delta I$（$I$は単位行列）です。ヘシアンの固有値がすべて $-\delta$ なので、この関数は $\mu_x = \delta$ の**強凹性**を持ちます。
    *   **$z$ に関する強凸性**: $z$ を含む項は $\tilde{V}(\theta ; z ; \zeta)+\delta \nabla_{\theta} \tilde{V}(\theta ; z ; \zeta)^{\top} x - F^{*}(z)$ です。最初の2項は $z$ に関して線形なので、凸性にも凹性にも寄与しません。したがって、強凸性は $-F^*(z)$ の項から来ます。
        ここで、凸解析における重要な双対関係を使います。関数 $F$ が $L_F$-smooth（勾配がリプシッツ連続）であることと、そのFenchel共役 $F^*$ が $(1/L_F)$-strongly convex であることは等価です（参考文献: Kakade et al., 2012）。
        我々は定理3.2の**仮定(ii)** で $F$ が $L_F$-smooth であることを仮定しているので、その双対 $F^*$ は $(1/L_F)$-強凸です。したがって、$-F^*(z)$ は $(1/L_F)$-強凸性を持ちます。よって、$\mu_z = 1/L_F$ となります。（論文では $\mu_z = L_F^{-1}$ と表記）

*   **性質2: リプシッツ性**
    *   $\Psi_\zeta(x, z)$ の $x$ に関する勾配が、ある量 $\ell_x(\zeta, z)$ で抑えられる必要があります。
    *   $\Psi_\zeta(x, z)$ の $z$ に関する勾配（劣勾配）が、ある量 $\ell_z(\zeta, x)$ で抑えられる必要があります。

    *   **$x$ に関するリプシッツ性**: $\nabla_x \Psi_\zeta(x, z) = \delta \nabla_\theta \tilde{V}(\theta; z; \zeta) - \delta x$。
        $\|\nabla_x \Psi_\zeta(x, z)\| \le \delta \|\nabla_\theta \tilde{V}(\theta; z; \zeta)\| + \delta \|x\|$。
        $\|z\|_\infty \le \ell_F, \|x\| \le \frac{C\ell_F}{(1-\gamma)^2}$ なので、この勾配は有界です。論文では $\ell_x(\zeta, z) = \mathcal{O}(\delta)$ と簡潔にまとめています。
    *   **$z$ に関するリプシッツ性**: $\partial_z \Psi_\zeta(x, z) = \partial_z(\tilde{V}(\theta; z; \zeta) + \dots) - \partial F^*(z)$。
        $\tilde{V}(\theta; z; \zeta) = \sum_k \gamma^k z_{s_k a_k}$ なので、その勾配は特定の成分が $\gamma^k$ で他がゼロのベクトルです。その $L_1$ ノルムは $\sum_k \gamma^k \le \frac{1}{1-\gamma}$ です。
        $\nabla_\theta \tilde{V}$ の項も同様に $z$ に線形であり、その勾配も評価できます。
        $\partial F^*(z)$ は、仮定(iii)の $F^*$ が $\ell_{F^*}$-Lipschitz であることから、その劣勾配のノルムは $\ell_{F^*}$ で抑えられます。
        これらをまとめると、劣勾配のノルム（$L_1$ノルム）は $\ell_z(\zeta, x) = \frac{1}{1-\gamma} + \ell_{F^*} + \mathcal{O}(\delta)$ で抑えられます。

**ステップ3: 理論の適用と結果の導出**

Zhang et al. (2020b) の定理1は、これらのパラメータを用いて、次のような一般化誤差の上界を与えます。
$$ \frac{\mu_{z}}{2} \mathbb{E}\left[\left\|\hat{z}-z^{*}\right\|_{\infty}^{2}\right] \leq \frac{C'}{n} \cdot\left(\frac{\mathbb{E}[\ell_{x}^2]}{\mu_{x}}+\frac{\mathbb{E}[\ell_{z}^2]}{\mu_{z}}\right) $$
（論文の式は少し異なりますが、本質は同じです）。
ここで、$\mathbb{E}[\ell_x^2]$ と $\mathbb{E}[\ell_z^2]$ は、それぞれ $\ell_x(\zeta, z)^2$ と $\ell_z(\zeta, x)^2$ の期待値の上界です。
*   $\mathbb{E}[\ell_x^2] = \mathcal{O}(\delta^2)$
*   $\mathbb{E}[\ell_z^2] = \mathcal{O}(\frac{1}{(1-\gamma)^2} + \ell_{F^*}^2 + \delta^2)$

これらのパラメータを代入します。
$$
\begin{aligned}
\frac{1/L_F}{2} \mathbb{E}[\|\hat{z} - z^*\|_\infty^2] &\le \frac{C'}{n} \left( \frac{\mathcal{O}(\delta^2)}{\delta} + \frac{\mathcal{O}(\frac{1}{(1-\gamma)^2} + \ell_{F^*}^2)}{1/L_F} \right) \\
&= \frac{C'}{n} \left( \mathcal{O}(\delta) + L_F \cdot \mathcal{O}(\frac{1}{(1-\gamma)^2} + \ell_{F^*}^2) \right)
\end{aligned}
$$
両辺に $2L_F$ を掛けると、
$$ \mathbb{E}[\|\hat{z} - z^*\|_\infty^2] \le \frac{2C'}{n} \left( \mathcal{O}(\delta L_F) + L_F^2 \cdot \mathcal{O}(\frac{1}{(1-\gamma)^2} + \ell_{F^*}^2) \right) $$
これを $\mathcal{O}$ 記法でまとめると、補題E.1(iii)の主張が得られます。
$$ \mathbb{E}\left[\left\|\hat{z}-z^{*}\right\|_{\infty}^{2}\right] \leq \mathcal{O}\left(\frac{L_{F}^{2}}{n(1-\gamma)^{2}}+\frac{L_{F}^{2} \ell_{F^{*}}^{2}}{n}+\frac{L_{F} \delta}{n}\right) $$
（論文の式にある $L_F^2 \delta^2$ の項は、より精密な評価から来ていますが、オーダーとしては $\delta$ の項に含まれます）。
これで補題E.1(iii)の証明が完了しました。

### 6. 最終的な結合と定理3.2の証明の完成

すべての部品が揃いました。これらを組み合わせて、定理3.2の最終的な誤差限界を導出します。

**ステップ1: 誤差の分解式への代入**

証明の初めに立てた誤差分解式を再掲します。
$$ \mathbb{E}[\|\hat{x}(\delta) - \nabla_{\theta} R\|^2] \le 2 \mathbb{E}[\|\hat{x}(\delta) - x^*(\delta)\|^2] + 2 \|x^*(\delta) - \nabla_{\theta} R\|^2 $$
（期待値は確率的な量 $\hat{x}$ にのみかかるので、第2項からは外れます）

この式に、補題E.1の結果を順に代入していきます。

1.  **補題E.1(i)を代入**: 第2項に代入します。
    $$ \|x^*(\delta) - \nabla_{\theta} R\|^2 = \mathcal{O}(\delta^2) $$

2.  **補題E.1(ii)を代入**: 第1項に代入します。
    $$ \mathbb{E}[\|\hat{x}(\delta) - x^*(\delta)\|^2] \le \frac{2 C^{2}\|z^{*}\|_{\infty}^{2}}{(1-\gamma)^{4}} \left(\frac{\gamma^{2 K}}{(1-\gamma)^{2}}+\frac{1}{n}\right)+\frac{2 C^{2}}{(1-\gamma)^{4}} \mathbb{E}\left[\left\|z^{*}-\hat{z}\right\|_{\infty}^{2}\right] $$

3.  **補題E.1(iii)を代入**: 上の式の最後の項 $\mathbb{E}[\|z^* - \hat{z}\|_\infty^2]$ に代入します。
    $$ \mathbb{E}[\|z^* - \hat{z}\|_\infty^2] \le \mathcal{O}\left(\frac{L_{F}^{2}}{n(1-\gamma)^{2}}+\frac{L_{F}^{2} \ell_{F^{*}}^{2}}{n}+\frac{L_{F} \delta + L_F^2 \delta^2}{n}\right) $$

これらをすべて元の分解式にまとめると、非常に長い式になります。
$$
\begin{aligned}
\mathbb{E}[\|\hat{x}(\delta) - \nabla_{\theta} R\|^2] \le & 2 \left[ \frac{2 C^{2}\|z^{*}\|_{\infty}^{2}}{(1-\gamma)^{4}} \left(\frac{\gamma^{2 K}}{(1-\gamma)^{2}}+\frac{1}{n}\right) \right. \\
& \left. + \frac{2 C^{2}}{(1-\gamma)^{4}} \cdot \mathcal{O}\left(\frac{L_{F}^{2}}{n(1-\gamma)^{2}}+\frac{L_{F}^{2} \ell_{F^{*}}^{2}}{n}+\frac{L_{F} \delta + L_F^2 \delta^2}{n}\right) \right] \\
& + 2 \cdot \mathcal{O}(\delta^2)
\end{aligned}
$$

**ステップ2: 式の整理**

この式を $\mathcal{O}$ 記法を用いて整理します。
まず、$\|z^*\|_\infty$ は、$\delta$ が小さいとき $\|z^*\|_\infty \le \ell_F$ でした。これを代入します。
主要な項を特定していきます。
*   **$1/n$ に比例する項 (サンプリング誤差)**:
    *   $\frac{C^2 \ell_F^2}{n(1-\gamma)^4}$
    *   $\frac{C^2}{(1-\gamma)^4} \cdot \frac{L_F^2}{n(1-\gamma)^2} = \frac{C^2 L_F^2}{n(1-\gamma)^6}$
    *   $\frac{C^2}{(1-\gamma)^4} \cdot \frac{L_F^2 \ell_{F^*}^2}{n} = \frac{C^2 L_F^2 \ell_{F^*}^2}{n(1-\gamma)^4}$
*   **$\gamma^K$ に比例する項 (打ち切り誤差)**:
    *   $\frac{C^2 \ell_F^2 \gamma^{2K}}{(1-\gamma)^6}$
*   **$\delta$ に依存する項 (近似誤差と、そのサンプリングへの影響)**:
    *   $\mathcal{O}(\delta^2)$
    *   $\frac{C^2}{(1-\gamma)^4} \cdot \mathcal{O}(\frac{\delta}{n}) = \mathcal{O}(\delta/n)$

これらの項をまとめると、
$$ \mathbb{E}[\|\hat{x}(\delta) - \nabla_{\theta} R\|^2] \le \mathcal{O}\left(\frac{C^{2}(\ell_{F}^{2}+L_{F}^{2} \ell_{F^{*}}^{2})}{n(1-\gamma)^{4}}+\frac{C^{2} L_{F}^{2}}{n(1-\gamma)^{6}}\right) + \mathcal{O}(\gamma^{2K}) + \mathcal{O}(\delta^2 + \delta/n) $$
となります。$\gamma^{2K}$ は $\gamma^K$ のオーダーなので、論文の表記と一致します。
$$ \mathbb{E}\left[\left\|\hat{x}(\delta)-\nabla_{\theta} R(\theta)\right\|^{2}\right] \leq \mathcal{O}\left(\frac{C^{2}\left(\ell_{F}^{2}+L_{F}^{2} \ell_{F^{*}}^{2}\right)}{n(1-\gamma)^{4}}+\frac{C^{2} L_{F}^{2}}{n(1-\gamma)^{6}}\right)+\mathcal{O}\left(\delta^{2}+\delta / n+\gamma^{K}\right) $$

**ステップ3: 極限操作**

最後に、我々が本当に興味があるのは、$\hat{\nabla}_\theta R(\pi_\theta) = \lim_{\delta \to 0_+} \hat{x}(\delta)$ と $\nabla_\theta R(\pi_\theta)$ の差です。
上の不等式で、$\delta \to 0$ の極限を取ることを考えます。
極限を取ると、$\delta$ に依存する項 $\mathcal{O}(\delta^2 + \delta/n)$ はゼロになります。
したがって、
$$ \lim_{\delta \to 0_+} \mathbb{E}[\|\hat{x}(\delta) - \nabla_\theta R\|^2] \le \mathcal{O}\left(\frac{C^{2}(\ell_{F}^{2}+L_{F}^{2} \ell_{F^{*}}^{2})}{n(1-\gamma)^{4}}+\frac{C^{2} L_{F}^{2}}{n(1-\gamma)^{6}}\right) + \mathcal{O}(\gamma^K) $$
期待値と極限の交換が可能であるという仮定のもと（これは優収束定理などにより正当化されます）、
$$ \mathbb{E}[\|\hat{\nabla}_\theta R - \nabla_\theta R\|^2] \le \mathcal{O}\left(\frac{C^{2}(\ell_{F}^{2}+L_{F}^{2} \ell_{F^{*}}^{2})}{n(1-\gamma)^{4}}+\frac{C^{2} L_{F}^{2}}{n(1-\gamma)^{6}}\right) + \mathcal{O}(\gamma^K) $$
となり、定理3.2の最終的な主張が証明されました。
