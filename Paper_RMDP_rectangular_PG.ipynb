

## 準備 
### MDPタプル

$$
(\mathcal{S}, \mathcal{A}, \gamma, \mu, P, R)
$$
### 方策と価値関数など

方策：$\boldsymbol{\pi}：\mathcal{S} \to \Delta(\mathcal{A})$
状態価値関数：$V_P^{\boldsymbol{\pi}}(s)$
行動価値関数：$Q_P^{\boldsymbol{\pi}}(s,a)$
状態確率：$P^{\boldsymbol{\pi}}(s'|s) = \sum_a \pi_s(a) P(s'|s,a)$
Occupancy Measure：$d_{P, \nu}^{\boldsymbol{\pi}} = \nu^T (I - \gamma P^{\boldsymbol{\pi}})^{-1}$

### ロバストMDP

不確実集合：$\mathcal{U}$
(s,a)-rectangular：$\mathcal{U}^{sa} = \times_{(s,a) \in \mathcal{X}} (\mathcal{P}_{(s,a)}, \mathcal{R}_{(s,a)})$
s-rectangular：$\mathcal{U}^s = \times_{s \in \mathcal{S}} (\mathcal{P}_s, \mathcal{R}_s)$

最悪ケース収益関数：$\rho^{\mathcal{U}}(\boldsymbol{\pi}) = \min_{(P,R) \in \mathcal{U}} J_{P,R}(\boldsymbol{\pi})$
ロバスト価値関数：$v_{\mathcal{U}}^{\boldsymbol{\pi}}(s) = \min_{(P,R) \in \mathcal{U}} V_{P,R}^{\boldsymbol{\pi}}(s)$
最悪ケース遷移確率，報酬関数：$(P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}) \in \arg\min_{(P,R) \in \mathcal{U}} J_{P,R}(\boldsymbol{\pi})$

### 不確実集合
 $l_p$-ball-constrained不確実集合：
     $\mathcal{R}_{(s,a)} = \{r \in \mathbb{R} \mid |r| \le \alpha_{s,a}\}$
     $\mathcal{P}_{(s,a)} = \{p \in \mathbb{R}^{\mathcal{S}} \mid \langle p, \mathbf{1} \rangle_S = 0, ||p||_p \le \beta_{s,a}\}$ 

 q-variance ($\kappa_q(v)$) と q-mean ($\omega_q(v)$)：
     $\kappa_q(v) = \min_{w \in \mathbb{R}} ||v - w\mathbf{1}||_q$
     $\omega_q(v) = \arg\min_{w \in \mathbb{R}} ||v - w\mathbf{1}||_q$
     $p, q$ はヘルダーの共役指数 ($1/p + 1/q = 1$)

 Normalized and Balanced (Robust) Value Function ($\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}$ または $\mathbf{u}_q(v)$)：
     $\mathbf{u}_q(v)(s) := \frac{\text{sign}(v(s) - \omega_q(v)) ||v(s) - \omega_q(v)||_q^{q-1}}{\kappa_q(v)^{q-1}}$
     $\langle \mathbf{u}_q(v), \mathbf{1} \rangle_S = 0$ かつ $||\mathbf{u}_q(v)||_p = 1$ 
     $\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}$ は $v=v_{\mathcal{U}}^{\boldsymbol{\pi}}$ 


## 定理

For any rectangular $l_p$-ball-constrained uncertainty, the robust policy gradient is given by:
$$ \partial_{\boldsymbol{\pi}} \rho^{\mathcal{U}} = \sum_{(s,a) \in \mathcal{X}} (d_{P_0, \mu}^{\boldsymbol{\pi}}(s) - c^{\boldsymbol{\pi}}(s)) Q_{\mathcal{U}}^{\boldsymbol{\pi}}(s,a) \nabla_{\boldsymbol{\pi}} \pi_s(a) $$
where
$$ c^{\boldsymbol{\pi}}(s) := \frac{\gamma \langle d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}, \beta^{\boldsymbol{\pi}} \rangle_S}{1 + \gamma \langle d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}, \beta^{\boldsymbol{\pi}} \rangle_S} d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}(s), \quad \forall s \in \mathcal{S}. $$

Recutangularを仮定した不確実集合を用いたロバストMDPにおける方策勾配は，通常のMDPにおける方策勾配と似た形で解けるということを示しています．

## 証明


まず，$\rho^{\mathcal{U}}(\boldsymbol{\pi})$ は，$\boldsymbol{\pi}$ に対して最悪のモデル $(P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}})$ を選んだときの性能 $J_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}}(\boldsymbol{\pi})$ を示していました．
$$ \rho^{\mathcal{U}}(\boldsymbol{\pi}) = J_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}}(\boldsymbol{\pi}) $$

方策勾配を考える際，この最悪ケースモデルを固定したものとして扱います（これはダンスキンの定理に近い考え？）．さらに標準的な方策勾配定理を，この固定された最悪ケースの下で適用します．

標準的な方策勾配定理によれば，
$$ \partial_{\boldsymbol{\pi}} J_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}}(\boldsymbol{\pi}) = \sum_{(s,a) \in \mathcal{X}} d_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, \mu}^{\boldsymbol{\pi}}(s) Q_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}(s,a) \nabla_{\boldsymbol{\pi}} \pi_s(a) $$
で表されます．この論文に合わせて$d_{\mathcal{U}, \mu}^{\boldsymbol{\pi}}(s) := d_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, \mu}^{\boldsymbol{\pi}}(s)$ および $Q_{\mathcal{U}}^{\boldsymbol{\pi}}(s,a) := Q_{P_{\mathcal{U}}^{\boldsymbol{\pi}}, R_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}(s,a)$ と書くと，
$$ \partial_{\boldsymbol{\pi}} \rho^{\mathcal{U}} = \sum_{(s,a) \in \mathcal{X}} d_{\mathcal{U}, \mu}^{\boldsymbol{\pi}}(s) Q_{\mathcal{U}}^{\boldsymbol{\pi}}(s,a) \nabla_{\boldsymbol{\pi}} \pi_s(a) $$
となります．次に，定理4.6(**導出TODO**)を用いて，$d_{\mathcal{U}, \mu}^{\boldsymbol{\pi}}(s)$ をNominal model$P_0$のOccupancy measureで表現します．Rectangularな $l_p$-ball constrained不確実性集合において，
$$
d_{\mathcal{U}, \mu}^{\boldsymbol{\pi}}(s) = d_{P_0, \mu}^{\boldsymbol{\pi}}(s) - \frac{\gamma \langle d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}, \beta^{\boldsymbol{\pi}} \rangle_S}{1 + \gamma \langle d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}, \beta^{\boldsymbol{\pi}} \rangle_S} d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}(s) 
$$
となります．  $d_{P_0, \mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}}^{\boldsymbol{\pi}}(s)$ は，Nominal model $P_0$ と初期ベクトルとして $\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}$を用いた場合の Occupancy measureです． $\beta^{\boldsymbol{\pi}}$ は，期待遷移不確実性の半径のベクトルです．

定理4.6 の導出は，最悪ケースの遷移確率 $P_{\mathcal{U}}^{\boldsymbol{\pi}}$ が Nominal model $P_0$ のランク1摂動であるという事実と，ランク1摂動下での Occupancy measure の変化に関する Lemma 4.4 に基づいています．
## 証明一時停止

### ランク1摂動とは？
ランク1摂動とは，ある行列や線形作用素に，**ランクが1の行列を加える（または引く）操作**のことを指します．
行列のランクとは，その行列の線形独立な行，列の最大数のことです．
**行列ランクの例**
* ランク0の行列は，全ての要素が0のゼロ行列だけです．
* ランク1の行列は，あるゼロでない列ベクトル $\mathbf{u}$ と，あるゼロでない行ベクトル $\mathbf{v}^T$ を用いて，$\mathbf{u} \mathbf{v}^T$ と書ける行列です．
	例えば，$\mathbf{u} = \begin{pmatrix} u_1 \\ u_2 \end{pmatrix}$, $\mathbf{v}^T = \begin{pmatrix} v_1 & v_2 & v_3 \end{pmatrix}$ のとき，$\mathbf{u} \mathbf{v}^T = \begin{pmatrix} u_1 \\ u_2 \end{pmatrix} \begin{pmatrix} v_1 & v_2 & v_3 \end{pmatrix} = \begin{pmatrix} u_1 v_1 & u_1 v_2 & u_1 v_3 \\ u_2 v_1 & u_2 v_2 & u_2 v_3 \end{pmatrix}$

**ランク1摂動の定義**

最初の行列を $A_0$ とし，ランク1の行列を $B$ とすると，ランク1摂動を受けた行列 $A_1$ は以下のように書けます．

$A_1 = A_0 + B$

ここで，$B$ はランク1なので，$B = \mathbf{u} \mathbf{v}^T$（$\mathbf{u}$ は列ベクトル，$\mathbf{v}^T$ は行ベクトル）と書けます．したがって，

$A_1 = A_0 + \mathbf{u} \mathbf{v}^T$　または，　$A_1 = A_0 - \mathbf{u} \mathbf{v}^T$

と表現されます．論文では，$P_0$に対して$P_{\mathcal{U}}^{\boldsymbol{\pi}}$が，
$$
P_{\mathcal{U}}^{\boldsymbol{\pi}} = P_0 - \mathbf{b} (\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}})^T
$$

のような形で書けることが示されています．ここで，$\mathbf{b}$ はあるベクトル，$\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}}$ もベクトルなので，$\mathbf{b} (\mathbf{u}_{\mathcal{U}}^{\boldsymbol{\pi}})^T$ はランク1の行列となります．

ランク1摂動は，元の行列の構造を比較的単純な形で変化させるため，数学的に扱いやすい性質をく持ちます．
**シャーマン・モリソンの公式**
元の行列 $A_0$ の逆行列 $A_0^{-1}$ が分かっている場合，ランク1摂動を受けた行列 $A_1 = A_0 + \mathbf{u} \mathbf{v}^T$ の逆行列 $A_1^{-1}$ は，シャーマン・モリソンの公式を用いて効率的計算できます．
$$
(A_0 + \mathbf{u} \mathbf{v}^T)^{-1} = A_0^{-1} - \frac{A_0^{-1} \mathbf{u} \mathbf{v}^T A_0^{-1}}{1 + \mathbf{v}^T A_0^{-1} \mathbf{u}}
$$
論文は，このシャーマン・モリソンの公式を，MDPのOccupancy measure $(I - \gamma P)^{-1}$ の計算に適用したものです．
