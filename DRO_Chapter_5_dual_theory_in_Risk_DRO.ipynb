# 最悪ケースリスク問題の双対理論
分布ロバスト最適化を以下のように再掲します．
$$
\begin{equation*}
\inf _{x \in \mathcal{X}} \sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)] \tag{1.2}
\end{equation*}
$$
ここで，$x$は決定変数(意思決定者によって$\mathcal{X} \subseteq \mathbb{R}^{n}$から選ばれるもの)，$Z$は不確実なパラメータであり，確率分布 $\mathbb{P}$ に支配される確率ベクトルです．
このページでは，本題に入るために，意思決定者が**リスク中立**(**risk-neutral**)であり，**曖昧さを回避**(**ambiguity-averse**) すると仮定します．

**リスク中立性とは？**：
$Z$をサンプルする$\mathbb{P}$が既知の場合，以下のような$\ell$の期待損失最小化問題である確率的計画法と同義です．
$$
\begin{equation*}
\inf _{x \in \mathcal{X}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)] \tag{1.1}
\end{equation*}
$$
意思決定者は，各$x$に対して期待損失が最も低いものを選びます．これをリスク中立性といいます．

**曖昧さ回避とは？**：
$\mathbb{P}$が未知の場合，$(1.2)$を解きます．ここで曖昧な部分とはどこを指すでしょうか？未知の部分ですね．つまり，$\mathbb{P}$がどんな分布かわからないわけです．曖昧さを回避するために，意思決定者は彼らにとって**最も不利な分布**の下での期待損失で評価されることになります．これを曖昧さ回避という，，，と思います．

## Law-invariant Risk尺度
意思決定者は**期待損失**をもとに意思決定を行います．例えば，とても確率は低いが非常に大きな損失をもたらすよくない事象が存在した場合，期待損失は期待値計算ですので，それも合わせて和をとります．そうなると，よくない事象に引っ張られて適切な意思決定ができない場合があります．そこで，本章では，**Law-invariant Risk尺度**($\varrho$)を用いて意思決定を行う場合を考えます．$\varrho$は実数空間上の確率分布(ボレル確率測度)の集合$\mathcal{P}(\mathbb{R})$ に対してリスキーさ指標を割り当てます．

### Law-invariantリスク尺度(定義5.1)
**前提**
* $\mathbb{P} \circ \ell^{-1}$ ：像測度(push-forward分布) ある確率変数を入力とする関数の出力値が従う新しい確率分布を言います．ここでは，$\mathbb{P}$に従った$Z$という確率変数を，損失関数$\ell: \mathbb{R}^{d} \rightarrow \mathbb{R}$という関数に入力したときに，$\ell$が従う確率分布とします．
* $\mathcal{L}\left(\mathbb{R}^{d}\right)$ ：全ての$\ell$の集合
**本文**
 $\varrho: \mathcal{P}(\mathbb{R}) \rightarrow \overline{\mathbb{R}}$ です．任意の $\ell \in \mathcal{L}\left(\mathbb{R}^{d}\right)$，$\mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right)$ および$d \in \mathbb{N}$ に対して，$\varrho\left(\mathbb{P} \circ \ell^{-1}\right)$ を $\varrho_{\mathbb{P}}[\ell(Z)]$ と示すことにします．

**どこが，Law-invariantなのか**
Lawとは，**確率法則,分布**のことです．invariantとは，不変という意味です．つまり，Law-invariantとは**確率分布がリスク尺度に関与しない**という意味になります．一方で，確率分布が生成する**像測度**には関与します．例えば，異なる $\ell_{1}$ と $\ell_{2}$，および2つの異なる $\mathbb{P}_{1}$ と $\mathbb{P}_{2}$に対して $\mathbb{P}_{1} \circ \ell_{1}^{-1}=\mathbb{P}_{2} \circ \ell_{2}^{-1}$ であれば，$\varrho_{\mathbb{P}_{1}}\left[\ell_{1}\left(Z_{1}\right)\right]=\varrho_{\mathbb{P}_{2}}\left[\ell_{2}\left(Z_{2}\right)\right]$ になります．

次に，Law-invariant Risk尺度の具体的な例をみていきましょう．

### 具体例：Value at Risk(VaR)
$\ell \in \mathcal{L}\left(\mathbb{R}^{d}\right)$ かつ $Z \sim \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right)$ である不確実な損失 $\ell(Z)$ のレベル $\beta \in(0,1)$ におけるVaRは次のように与えられます．
$$
\begin{equation*}
\beta-\operatorname{VaR}_{\mathbb{P}}[\ell(Z)]=\inf \{\tau \in \mathbb{R}: \mathbb{P}(\ell(Z) \leq \tau) \geq 1-\beta\} \tag{5.1}
\end{equation*}
$$
レベル $\beta$ におけるVaRは，確率 $1-\beta$ で損失を上回る最小$\tau^{\star}$ を表します．逆をとると，
$$
\begin{equation*}
\beta-\operatorname{VaR}_{\mathbb{P}}[\ell(Z)]=\inf \{\tau \in \mathbb{R}: \mathbb{P}(\ell(Z) \geq \tau) \leq \beta\} \tag{5.2}
\end{equation*}
$$
と表現できます．この尺度のいいところは値が有限であるということです．ほかのLaw-invariant Risk尺度の多くは特定の関数などに対してのみ有限です．


### Law-invariant Risk尺度の性質
いくつかの性質を示し，コヒーレントRisk尺度を定義しましょう．
$\varrho$は，
* 以下の場合，並進不変 (translation-invariant) です．

$$
\varrho_{\mathbb{P}}[\ell(Z)+c]=\varrho_{\mathbb{P}}[\ell(Z)]+c \quad \forall \ell \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall c \in \mathbb{R}, \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
$$

* 以下の場合，スケール不変 (scale-invariant) です．

$$
\varrho_{\mathbb{P}}[c \ell(Z)]=c \varrho_{\mathbb{P}}[\ell(Z)] \quad \forall \ell \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall c \in \mathbb{R}_{+}, \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
$$

* 以下の場合，単調 (monotone) です．

$$
\begin{aligned}
& \varrho_{\mathbb{P}}\left[\ell_{1}(Z)\right] \leq \varrho_{\mathbb{P}}\left[\ell_{2}(Z)\right] \\
& \quad \forall \ell_{1}, \ell_{2} \in \mathcal{L}\left(\mathbb{R}^{d}\right) \text { with } \ell_{1}(Z) \leq \ell_{2}(Z) \mathbb{P} \text {-a.s., } \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
\end{aligned}
$$

* 以下の場合，凸 (convex) です．

$$
\begin{aligned}
\varrho_{\mathbb{P}}\left[\theta \ell_{1}(Z)+(1-\theta) \ell_{2}(Z)\right] & \leq \theta \varrho_{\mathbb{P}}\left[\ell_{1}(Z)\right]+(1-\theta) \varrho_{\mathbb{P}}\left[\ell_{2}(Z)\right] \\
\forall \ell_{1}, \ell_{2} & \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall \theta \in[0,1], \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) .
\end{aligned}
$$

コヒーレントなRisk尺度は，**並進不変，スケール不変，単調かつ凸**です．凸Risk尺度は，並進不変，単調かつ凸です．

### Law-invariant Risk尺度が引き起こすDRO問題
期待損失に対するDROではなく．Law-invariant Risk尺度に対するDROを考慮します．

\end{equation*}
$$
