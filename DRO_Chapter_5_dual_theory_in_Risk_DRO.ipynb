# このページについて
このページでは，Distributionally Robust Optimization(https://arxiv.org/abs/2411.02549)の第5章の勉強内容をまとめていこうと思います．

# 最悪ケースリスク問題の双対理論
分布ロバスト最適化を以下のように再掲します．
$$
\begin{equation*}
\inf _{x \in \mathcal{X}} \sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)] \tag{1.2}
\end{equation*}
$$
ここで，$x$は決定変数(意思決定者によって$\mathcal{X} \subseteq \mathbb{R}^{n}$から選ばれるもの)，$Z$は不確実なパラメータであり，確率分布 $\mathbb{P}$ に支配される確率ベクトルです．
このページでは，本題に入るために，意思決定者が**リスク中立**(**risk-neutral**)であり，**曖昧さを回避**(**ambiguity-averse**) すると仮定します．

**リスク中立性とは？**：
$Z$をサンプルする$\mathbb{P}$が既知の場合，以下のような$\ell$の期待損失最小化問題である確率的計画法と同義です．
$$
\begin{equation*}
\inf _{x \in \mathcal{X}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)] \tag{1.1}
\end{equation*}
$$
意思決定者は，各$x$に対して期待損失が最も低いものを選びます．これをリスク中立性といいます．

**曖昧さ回避とは？**：
$\mathbb{P}$が未知の場合，$(1.2)$を解きます．ここで曖昧な部分とはどこを指すでしょうか？未知の部分ですね．つまり，$\mathbb{P}$がどんな分布かわからないわけです．曖昧さを回避するために，意思決定者は彼らにとって**最も不利な分布**の下での期待損失で評価されることになります．これを曖昧さ回避という，，，と思います．

## Law-invariant Risk尺度
意思決定者は**期待損失**をもとに意思決定を行います．例えば，とても確率は低いが非常に大きな損失をもたらすよくない事象が存在した場合，期待損失は期待値計算ですので，それも合わせて和をとります．そうなると，よくない事象に引っ張られて適切な意思決定ができない場合があります．そこで，本章では，**Law-invariant Risk尺度**($\varrho$)を用いて意思決定を行う場合を考えます．$\varrho$は実数空間上の確率分布(ボレル確率測度)の集合$\mathcal{P}(\mathbb{R})$ に対してリスキーさ指標を割り当てます．

### Law-invariantRisk尺度(定義5.1)
**前提**
* $\mathbb{P} \circ \ell^{-1}$ ：像測度(push-forward分布) ある確率変数を入力とする関数の出力値が従う新しい確率分布を言います．ここでは，$\mathbb{P}$に従った$Z$という確率変数を，損失関数$\ell: \mathbb{R}^{d} \rightarrow \mathbb{R}$という関数に入力したときに，$\ell$が従う確率分布とします．
* $\mathcal{L}\left(\mathbb{R}^{d}\right)$ ：全ての$\ell$の集合
**本文**
 $\varrho: \mathcal{P}(\mathbb{R}) \rightarrow \overline{\mathbb{R}}$ です．任意の $\ell \in \mathcal{L}\left(\mathbb{R}^{d}\right)$，$\mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right)$ および$d \in \mathbb{N}$ に対して，$\varrho\left(\mathbb{P} \circ \ell^{-1}\right)$ を $\varrho_{\mathbb{P}}[\ell(Z)]$ と示すことにします．

**どこが，Law-invariantなのか？**
Lawとは，**確率法則,分布**のことです．invariantとは，不変という意味です．つまり，Law-invariantとは**確率分布がRisk尺度に関与しない**という意味になります．一方で，確率分布が生成する**像測度**には関与します．例えば，異なる $\ell_{1}$ と $\ell_{2}$，および2つの異なる $\mathbb{P}_{1}$ と $\mathbb{P}_{2}$に対して $\mathbb{P}_{1} \circ \ell_{1}^{-1}=\mathbb{P}_{2} \circ \ell_{2}^{-1}$ であれば，$\varrho_{\mathbb{P}_{1}}\left[\ell_{1}\left(Z_{1}\right)\right]=\varrho_{\mathbb{P}_{2}}\left[\ell_{2}\left(Z_{2}\right)\right]$ になります．

次に，Law-invariant Risk尺度の具体的な例をみていきましょう．

### 具体例：Value at Risk(VaR)
$\ell \in \mathcal{L}\left(\mathbb{R}^{d}\right)$ かつ $Z \sim \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right)$ である不確実な損失 $\ell(Z)$ のレベル $\beta \in(0,1)$ におけるVaRは次のように与えられます．
$$
\begin{equation*}
\beta-\operatorname{VaR}_{\mathbb{P}}[\ell(Z)]=\inf \{\tau \in \mathbb{R}: \mathbb{P}(\ell(Z) \leq \tau) \geq 1-\beta\} \tag{5.1}
\end{equation*}
$$
レベル $\beta$ におけるVaRは，確率 $1-\beta$ で損失を上回る最小$\tau^{\star}$ を表します．逆をとると，
$$
\begin{equation*}
\beta-\operatorname{VaR}_{\mathbb{P}}[\ell(Z)]=\inf \{\tau \in \mathbb{R}: \mathbb{P}(\ell(Z) \geq \tau) \leq \beta\} \tag{5.2}
\end{equation*}
$$
と表現できます．この尺度のいいところは値が有限であるということです．ほかのLaw-invariant Risk尺度の多くは特定の関数などに対してのみ有限です．


### Law-invariant Risk尺度の性質
いくつかの性質を示し，コヒーレントRisk尺度を定義しましょう．
$\varrho$は，
* 以下の場合，並進不変 (translation-invariant) です．

$$
\varrho_{\mathbb{P}}[\ell(Z)+c]=\varrho_{\mathbb{P}}[\ell(Z)]+c \quad \forall \ell \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall c \in \mathbb{R}, \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
$$

* 以下の場合，スケール不変 (scale-invariant) です．

$$
\varrho_{\mathbb{P}}[c \ell(Z)]=c \varrho_{\mathbb{P}}[\ell(Z)] \quad \forall \ell \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall c \in \mathbb{R}_{+}, \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
$$

* 以下の場合，単調 (monotone) です．

$$
\begin{aligned}
& \varrho_{\mathbb{P}}\left[\ell_{1}(Z)\right] \leq \varrho_{\mathbb{P}}\left[\ell_{2}(Z)\right] \\
& \quad \forall \ell_{1}, \ell_{2} \in \mathcal{L}\left(\mathbb{R}^{d}\right) \text { with } \ell_{1}(Z) \leq \ell_{2}(Z) \mathbb{P} \text {-a.s., } \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) ;
\end{aligned}
$$

* 以下の場合，凸 (convex) です．

$$
\begin{aligned}
\varrho_{\mathbb{P}}\left[\theta \ell_{1}(Z)+(1-\theta) \ell_{2}(Z)\right] & \leq \theta \varrho_{\mathbb{P}}\left[\ell_{1}(Z)\right]+(1-\theta) \varrho_{\mathbb{P}}\left[\ell_{2}(Z)\right] \\
\forall \ell_{1}, \ell_{2} & \in \mathcal{L}\left(\mathbb{R}^{d}\right), \forall \theta \in[0,1], \forall \mathbb{P} \in \mathcal{P}\left(\mathbb{R}^{d}\right) .
\end{aligned}
$$

コヒーレントなRisk尺度は，**並進不変，スケール不変，単調かつ凸**です．凸Risk尺度は，並進不変，単調かつ凸です．

## Law-invariant Risk尺度が引き起こすDRO問題

### 確実性等価
期待損失に対するDROではなく．Law-invariant Risk尺度に対するDROを考慮します．
$$
\begin{equation*}
\inf _{x \in \mathcal{X}} \sup _{\mathbb{P} \in \mathcal{P}} \varrho_{\mathbb{P}}[\ell(x, Z)] \tag{5.3}
\end{equation*}
$$
(5.3)は，Risk尺度を最小化するため，リスク回避的なDRO問題といえます．この問題が，じつは(5.3)は，リスク中立なDRO(1.2)に等価に変換できます．よってリスク回避とリスク中立間の関係性を導くことができます．つまり，特定条件でリスクを考える問題がリスクを考えなくてもよい問題と等価(**確実性等価**)ということを示します．

最初に，非効用関数を定義します．これは損失に対する何らかの指標を示すRisk尺度の一種で凸で連続な関数です．

**定義5.4(非効用関数)**
非効用関数 $g: \mathbb{R} \rightarrow \mathbb{R}$ は，$g(0)=0$ かつ全ての $\tau \neq 0$ に対して $g(\tau)>\tau$ を満たす凸（したがって連続）関数とします．

非効用関数を用いると，Law-invariant Risk尺度の損失に対する期待値と等価になります．等価の性質を持つ$\varrho$を，最適化された確実性等価といいます．

**定義 5.5 (最適化された確実性等価)**
 $g$ によって誘導される最適化された確実性等価は，
$$
\varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left(\underbrace{\tau}_{\text{現在の損失}} + \underbrace{\mathbb{E}_{\mathbb{P}}[g(\ell(Z)-\tau)]}_{\text{将来の損失の期待非効用}}\right) \tag{5.5}
$$
を持つ $\varrho$ です．$\tau$は現在の損失であり，右辺第二項は，将来現在の損失を引いた将来の損失の期待非効用です．$\tau$を増やせば，現在の損失は増えますが将来の損失は減ります．$\tau$をどうやって調整するか，(5.5)右辺は期待損失を最小化する問題を示しています．この解(最適化されたもの)こそが， $\varrho$ であるのです．

### 最適化された確実性等価と$\phi$ダイバージェンス
最適化された確実性等価は，あるDROの最適値と一致します．見ていきましょう．
**定義**
* $\phi: \mathbb{R} \rightarrow \overline{\mathbb{R}}$ ：エントロピー関数(定義2.4)
	* $\phi(1)=0$ 
	* すべての $s<0$ に対して $\phi(s)=+\infty$ 
	* 下半連続凸関数
	* 1の近傍で2回連続微分可能
	* $\phi^{\prime}(1)=0$ 
	* $\phi^{\prime \prime}(1)>0$
	* $\phi^{\infty}(1)=\infty$ 
*  $\phi^{*}$：$\phi$の凸共役 
	* 全ての $t \in \mathbb{R}$ に対して $\phi^{*}(t) \geq t$ 
	*  $\phi^{*}(0)=0$
	* 全ての $t \neq 0$ に対して $\phi^{*}(t)>t$
もし，$\mathbb{E}_{\hat{p}}[\ell(Z)]>-\infty$ であれば， $g=\phi^{*}$ によって作られる最適化された確実性等価は，
$$
\begin{equation*}
\inf _{\lambda_{0} \in \mathbb{R}}\left( \lambda_{0}+\mathbb{E}_{\hat{\mathbb{P}}}\left[\phi^{*}\left(\ell(Z)-\lambda_{0}\right)\right]\right)=\sup _{\mathbb{P} \in \mathcal{P}(\mathcal{Z})}\left( \mathbb{E}_{\mathbb{P}}[\ell(Z)]-\mathrm{D}_{\phi}(\mathbb{P}, \hat{\mathbb{P}})\right) \tag{5.6}
\end{equation*}
$$
となります．右辺は，ある距離$D$によって正則化されたペナルティベースのDRO問題の最適解です．infとsupが一致するところが，強双対性っぽいですね．実際に4章の制約付き $\phi$-ダイバージェンス不確実性集合上の最悪ケース期待値問題に対する強双対定理と綱が英があります．

## 様々な最適化された確実性等価
最適化された確実性等価は，色々なところで使われている有名なRisk尺度を含みます．それらについて，いくつか説明します．

### 平均分散Risk尺度
**定義 5.6 (平均分散Risk尺度)**
リスク回避係数 $\beta \in(0, \infty)$ を持つ平均分散Risk尺度は，
*  $\mathbb{V}_{\mathbb{P}}[\ell(Z)]$：$\ell(Z)$ の分散
$$
\varrho_{\mathbb{P}}[\ell(Z)]=\mathbb{E}_{\mathbb{P}}[\ell(z)]+\beta \cdot \mathbb{V}_{\mathbb{P}}[\ell(Z)],
$$

を持つLaw-invariant Risk尺度 $\varrho$ となります．

**用法(強制的：coercive)**
 $f: \mathbb{R} \rightarrow \overline{\mathbb{R}}$ が強制的 (coercive) であるとは，$\lim _{i \rightarrow \infty}\left|\tau_{i}\right|=\infty$ である全てのシーケンス $\left\{\tau_{i}\right\}_{i \in \mathbb{N}}$ に対して $\lim _{i \rightarrow \infty} f\left(\tau_{i}\right)=\infty$ である場合をいいます．
* 最適化された確実性等価→最悪ケース期待値と変換するときに使うらしいです．


**命題 5.7 (平均分散Risk尺度)**
$\beta \in(0, \infty)$ を持つ平均分散Risk尺度 $\varrho$ は，非効用関数 $g(\tau)=\tau+\beta \tau^{2}$ によって誘導される最適化された確実性等価となります．(5.5) の目的関数は $\tau$ に関して強制的であり，$\tau^{\star}=\mathbb{E}_{\mathbb{P}}[\ell(Z)]$ によって一意に最小化されます． $\tau$ が強制的である場合，最小値が存在します．

**証明**
まず，5.5を再掲します．
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left( \tau+\mathbb{E}_{\mathbb{P}}[g(\ell(Z)-\tau)] \right) $$
ここに， $g(\tau)=\tau+\beta \tau^{2}$ を代入します．
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left( \tau+\mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau)+\beta (\ell(Z)-\tau)^{2}] \right) $$
期待値の線形性を用いて，式を変形します．
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left( \tau+\mathbb{E}_{\mathbb{P}}[\ell(Z)-\tau]+\mathbb{E}_{\mathbb{P}}[\beta (\ell(Z)-\tau)^{2}] \right) $$
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left( \tau+\mathbb{E}_{\mathbb{P}}[\ell(Z)]-\tau+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau)^{2}] \right) $$
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \left( \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau)^{2}] \right) $$
最小化したい関数を $H(\tau)$ とおきます．
$$ H(\tau) = \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau)^{2}] $$
$H(\tau)$ を $\tau$ について最小化する問題は $\inf_{\tau \in \mathbb{R}} H(\tau)$ となります．最小化する対象の関数 $H(\tau)$ は $\tau$ の関数とみなせます．$H(\tau)$ を $\tau$ で微分して，最小値を与える $\tau$ を探します．
$$ \frac{d H(\tau)}{d \tau} = \frac{d}{d \tau} \left( \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau)^{2}] \right) $$
微分と期待値の交換をして，
$$ \frac{d H(\tau)}{d \tau} = \beta \mathbb{E}_{\mathbb{P}}\left[ \frac{d}{d \tau} (\ell(Z)-\tau)^{2} \right] $$
$$ \frac{d H(\tau)}{d \tau} = \beta \mathbb{E}_{\mathbb{P}}\left[ 2(\ell(Z)-\tau) \cdot (-1) \right] $$
$$ \frac{d H(\tau)}{d \tau} = -2\beta \mathbb{E}_{\mathbb{P}}[\ell(Z)-\tau] $$
$$ \frac{d H(\tau)}{d \tau} = -2\beta (\mathbb{E}_{\mathbb{P}}[\ell(Z)]-\tau) $$
最小になるような $\tau$ は勾配が 0 になるので，$\frac{d H(\tau)}{d \tau} = 0$ とおきます．
$$ -2\beta (\mathbb{E}_{\mathbb{P}}[\ell(Z)]-\tau) = 0 $$
$\beta > 0$ なので，これが成り立つのは
$$ \mathbb{E}_{\mathbb{P}}[\ell(Z)]-\tau = 0 $$
$$ \tau = \mathbb{E}_{\mathbb{P}}[\ell(Z)] $$
のときです．この値を $\tau^*$ とします．
$$ \tau^{\star}=\mathbb{E}_{\mathbb{P}}[\ell(Z)] $$
次に，これが最小値を与えることを確認します．二階微分を計算します．
$$ \frac{d^2 H(\tau)}{d \tau^2} = \frac{d}{d \tau} \left( -2\beta (\mathbb{E}_{\mathbb{P}}[\ell(Z)]-\tau) \right) = -2\beta (-1) = 2\beta $$
$\beta > 0$ なので二階微分は正であり，$H(\tau)$ は $\tau$ に関して下に凸な関数です．したがって，勾配が 0 になる $\tau = \tau^{\star}$ で最小値を取ります．

また，$H(\tau)$ は $\tau$ の2次関数であり，$\tau^2$ の係数（$\beta$）が正なので，$\tau \to \pm \infty$ で $H(\tau) \to +\infty$ となります．したがって，$H(\tau)$ は $\tau$ に関して**強制的 (coercive)** です．

最小値は，$H(\tau)$ に $\tau = \tau^{\star} = \mathbb{E}_{\mathbb{P}}[\ell(Z)]$ を代入することで得られます．
$$ \inf_{\tau \in \mathbb{R}} H(\tau) = H(\tau^{\star}) = \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\tau^{\star})^{2}] $$
$$ H(\tau^{\star}) = \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{E}_{\mathbb{P}}[(\ell(Z)-\mathbb{E}_{\mathbb{P}}[\ell(Z)])^{2}] $$
ここで，$\mathbb{E}_{\mathbb{P}}[(\ell(Z)-\mathbb{E}_{\mathbb{P}}[\ell(Z)])^{2}]$ は，損失 $\ell(Z)$ の分散 $\mathbb{V}_{\mathbb{P}}[\ell(Z)]$ の定義です．
したがって，
$$ \inf_{\tau \in \mathbb{R}} H(\tau) = \mathbb{E}_{\mathbb{P}}[\ell(Z)]+\beta \mathbb{V}_{\mathbb{P}}[\ell(Z)] $$
これは，リスク回避係数 $\beta$ を持つ平均分散Risk尺度の定義と一致します．

### 平均-MAD Risk尺度
平均分散の次は，平均中央値絶対偏差Risk尺度です．
**定義 5.8 (平均-MAD Risk尺度)**
 $\beta \in(0, \infty)$ を持つ平均-中央値絶対偏差 (MAD) Risk尺度は，
* $\mathbb{M}_{\mathbb{P}}[\ell(Z)]$： $\mathbb{P}$ の下での $\ell(Z)$ の中央値
$$
\varrho_{\mathbb{P}}[\ell(Z)]=\mathbb{E}_{\mathbb{P}}[\ell(z)]+\beta \cdot \mathbb{E}_{\mathbb{P}}\left[\left|\ell(Z)-\mathbb{M}_{\mathbb{P}}[\ell(Z)]\right|\right],
$$

を持つLaw-invariant Risk尺度 $\varrho$ となります．

**命題 5.9 (平均-MAD Risk尺度)**
 $\beta \in(0, \infty)$ を持つ平均-MAD Risk尺度 $\varrho$ は， $g(\tau)=\tau+\beta|\tau|$ によって作られる最適化された確実性等価です． (5.5) の目的関数は $\tau$ に関して強制的であり，$\tau^{\star}=\mathbb{M}_{\mathbb{P}}[\ell(Z)]$ によって最小化されます．

証明は前と同じ感じでやれば行けると思うのでスキップします．


### 条件付きVaR(CVaR)
VaRに対して条件付きVaR(CVaR)を定義します．制約付きMDPなどでもたまに出てくる指標です．

**定義 5.10 (条件付きVaR：CVaR)**
 $\beta \in(0,1)$ における条件付きVaR (CVaR) は，
$$
\begin{equation*}
\beta-\mathrm{CVaR}_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \tau+\frac{1}{\beta} \mathbb{E}_{\mathbb{P}}[\max \{\ell(Z)-\tau, 0\}] \tag{5.7}
\end{equation*}
$$
を持つLaw-invariant Risk尺度であり，$\beta-\mathrm{CVaR}$ と示されます．
* $\beta$ が1に近づくにつれて $\beta$ - $\mathrm{CVaR}_{\mathbb{P}}[\ell(Z)]$ は $\mathbb{E}_{\mathbb{P}}[\ell(Z)]$ に収束します．

 $\beta \in(0,1)$ におけるCVaRが最適化された確実性等価であることを示します．


**命題 5.11 (CVaR)**
$\beta \in(0,1)$ におけるCVaRは，$g(\tau)=\beta^{-1} \max \{\tau, 0\}$ によって誘導される最適化された確実性等価です．(5.5) の目的関数は $\tau$ に関して強制的であり，$\tau^{\star}=\beta-\mathrm{VaR}_{\mathbb{P}}[\ell(Z)]$ によって最小化されます．

**証明**
前のように，5.5を出します．
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \tau+\mathbb{E}_{\mathbb{P}}[g(\ell(Z)-\tau)] $$
次に，$g(\tau)=\beta^{-1} \max \{\tau, 0\}$ をこの式に代入します．
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \tau+\mathbb{E}_{\mathbb{P}}[\beta^{-1} \max \{\ell(Z)-\tau, 0\}] $$
期待値の線形性を用いると，
$$ \varrho_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \tau+\frac{1}{\beta} \mathbb{E}_{\mathbb{P}}[\max \{\ell(Z)-\tau, 0\}] $$
となります．
ここで，定義 5.10 で与えられた $\beta$-CVaR の定義式は以下のようになります．
$$ \beta-\mathrm{CVaR}_{\mathbb{P}}[\ell(Z)]=\inf _{\tau \in \mathbb{R}} \tau+\frac{1}{\beta} \mathbb{E}_{\mathbb{P}}[\max \{\ell(Z)-\tau, 0\}] $$
両者が一致していることが分かります．
****

Rockafellarの研究により，$G(\tau)$ を最小化する $\tau$ の集合は， $\ell(Z)$ の $(1-\beta)$-分位点の集合と一致し，特に $\beta-\mathrm{VaR}_{\mathbb{P}}[\ell(Z)]$ が最小化子の一つであることが知られています.
したがって，$\tau^{\star}=\beta-\mathrm{VaR}_{\mathbb{P}}[\ell(Z)]$ は $G(\tau)$ を最小化します．
$(1-\beta)$-分位点とは，損失がこの値以下になる確率が $(1-\beta)$ 以上となる**最小の値，或いは区間**を示します．$\beta$-VaR と同じです．つまり，CVaRを定義する最小化問題の解 $\tau^*$ は，この**分位点の集合の中**にあるということです．

### エントロピーRisk尺度
双対表現を持つRisk尺度を紹介します．
**定義 5.12(エントロピーリスク尺度)**
 $\beta \in(0, \infty)$ を持つエントロピーRisk尺度は，

$$
\begin{equation*}
\beta-\mathrm{ERM}_{\mathbb{P}}[\ell(Z)]=\frac{1}{\beta} \log \mathbb{E}_{\mathbb{P}}[\exp (\beta \ell(Z))] \tag{5.8}
\end{equation*}
$$

を持つLaw-invariant Risk尺度であり，$\beta$-ERM と示されます．これも最適化された確実性等価であり，，$\tau^{\star}=\beta^{-1} \log \left(\mathbb{E}_{\mathbb{P}}[\exp (\beta \ell(Z))]\right)$ によって最小化されます．


**命題 5.14 (エントロピーリスク尺度の双対表現)**
$\mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)]>-\infty$ と仮定します．エントロピーリスク尺度は双対表現
$$
\beta-\operatorname{ERM}_{\hat{\mathbb{P}}}[\ell(Z)]=\sup _{\mathbb{P} \in \mathcal{P}(\mathcal{Z})} \mathbb{E}_{\mathbb{P}}[\ell(Z)]-\frac{1}{\beta} \cdot \operatorname{KL}(\mathbb{P}, \hat{\mathbb{P}})
$$

を持ちます．

**証明**
Donsker-Varadhan の公式 (命題 2.9) を利用して証明します．
命題 2.9 より，
$$ KL(\mathbb{P}, \hat{\mathbb{P}}) = \sup_{f} \mathbb{E}_{\mathbb{P}}[f(Z)] - \log(\mathbb{E}_{\hat{\mathbb{P}}}[e^{f(Z)}]) $$
これを変形すると，
$$ \sup_{\mathbb{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] - \frac{1}{\beta} KL(\mathbb{P}, \hat{\mathbb{P}}) = \sup_{\mathbb{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] - \frac{1}{\beta} \sup_{f} \left( \mathbb{E}_{\mathbb{P}}[f(Z)] - \log(\mathbb{E}_{\hat{\mathbb{P}}}[e^{f(Z)}]) \right) $$
となります．ここで $\ell(Z) = f(Z)/\beta$ となるような $f$ を考えると，右辺が $\beta$-ERM の定義式 ( に帰着することを示すことができるみたいです．
