# このページについて
このページでは、Distributionally Robust Optimization(https://arxiv.org/abs/2411.02549)の第4章の勉強内容をまとめていこうと思います。
## 内側最大化と双対理論

DROにおける内側最大化問題は次のように表されます。

$$
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] \quad (*) \tag{1}
$$

3章であったように、$(1)$は簡単に解けない場合があります。4章では、双対理論を使ってこの問題を解いていくことを考えます。

### 双対問題の構築
**1：問題を値関数の形で表現**

元の最適化問題の最適値を、ある**値関数 $h$** を使って表現することを考えます。$\mathcal{P}$ のパラメータを $u$ として元の最適化問題を $\sup -h(u)$ のような形に書き換えます。ここで、$h(u)$ は、$u$ をパラメータとして持つ**最小化**問題の最適値として定義します。
モーメント不確実性集合の場合、以下のように変換します。
$$
\begin{equation*}
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)]=\sup _{u \in \mathcal{F}} \sup _{\mathbb{P} \in \mathcal{P}_{f}(\mathcal{Z})}\left\{\mathbb{E}_{\mathbb{P}}[\ell(Z)]: \mathbb{E}_{\mathbb{P}}[f(Z)]=u\right\}=\sup _{u \in \mathcal{F}}-h(1, u) \tag{2}
\end{equation*}
$$
 $u$ を指定したときの最小化問題(の負の値)として $h(1, u)$ を考え、元の問題は $\sup_{u \in \mathcal{F}} -h(1, u)$ となっています。

**2：値関数の凸性を証明(補題 4.1)**

次に値関数が凸関数であることを示します。補題 4.1 (最適値関数の凸性) を用います。

*   **補題 4.1** $H(u, v)$ が $(u, v)$ に関して凸な関数であるとき、$h(u) = \inf_{v} H(u, v)$ で定義される関数 $h(u)$ も $u$ に関して凸になります。

元の問題は $\mathbb{P}$ に関する複雑な最適化問題ですが、$h(u)$ の最小化問題は$v$と $(u_0, u)$ に関して凸構造を持っていれば、$h(u)$ 自体の凸性が保証されるため、解きやすい問題になります。

**3：双対問題を構築し、強い双対性を得る(補題 4.2)**

$h(u)$ が凸関数であるならば、双対問題を構築することが可能です。ここでフェンシェル・モローの定理が登場します。

*   前提：凸共役と双共役
    *   凸関数 $h(u)$ に対して、その**凸共役** 関数 $h^*(\lambda)$ は $h^*(\lambda) = \sup_{u} \{ \lambda^{\top} u - h(u) \}$ で定義されます。 $h(u)$ を、双対空間の変数 $\lambda$ で表現したものです。
    *   $h^*$ の凸共役 $(h^*)^*(\cdot)$ を $h^{**}(\cdot)$ と書き、これを $h$ の双共役関数といいます。
*   **補題 4.2**
    1.  常に $h^{**}(u) \le h(u)$ が成り立つ(弱い双対性)。
    2.  $h$ が閉じた (closed)凸関数であれば、$h^{**}(u) = h(u)$ が成り立つ(少なくとも $h$ の定義域の内部 $\operatorname{rint}(\operatorname{dom}(h))$ では $h$ は連続であり、$h^{**}=h$ となる)。

$h(u)$ が凸関数であれば、一度双対空間に$h^*$として移し、元の空間に戻しても元の $h$ が復元される、ということです。

**4：双対問題の最終構築**

元の問題が $\sup -h(u)$ とします。フェンシェル・モローの定理から $h(u) \ge h^{**}(u)$ なので、$\sup -h(u) \le \sup -h^{**}(u)$ が成り立ちます。ここで、$h^{**}(u)$ の定義は、

$$
h^{**}(u) = \sup_{\lambda} \{ u^{\top} \lambda - h^*(\lambda) \}
$$

なので、元の問題は

$$
\sup -h(u) \le \sup \left\{ - \sup_{\lambda} \{ u^{\top} \lambda - h^*(\lambda) \} \right\} = \sup \left\{ \inf_{\lambda} \{ -u^{\top} \lambda + h^*(\lambda) \} \right\}
$$

となります。この右辺 $\inf_{\lambda} \{ -u^{\top} \lambda + h^*(\lambda) \}$ が、元の問題 $\sup -h(u)$ の**双対問題**として現れるわけです。

そして、フェンシェル・モローの定理の2は、元の問題と双対問題の最適値が一致する、すなわち**強い双対性**が成り立つための条件を示します。つまり閉じた凸関数かどうかが重要ですね。

それでは、モーメント不確実性集合(定理4.5)と$\phi$ダイバージェンス不確実性集合(定理4.14)について、双対問題を見ていきましょう。


### モーメント不確実性集合に対する双対理論

#### 問題設定
モーメント不確実性集合 $\mathcal{P}=\left\{\mathbb{P} \in \mathcal{P}_{f}(\mathcal{Z}): \mathbb{E}_{\mathbb{P}}[f(Z)] \in \mathcal{F}\right\}$ 上での最悪ケース期待値 $\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)]$ を考えます。

#### 補助関数 $h$ の導入とその凸性

問題を以下のように書き換えます。

$$
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)]=\sup _{u \in \mathcal{F}} \underbrace{\sup _{\mathbb{P} \in \mathcal{P}_{f}(\mathcal{Z})}\left\{\mathbb{E}_{\mathbb{P}}[\ell(Z)]: \mathbb{E}_{\mathbb{P}}[f(Z)]=u\right\}}_{\text{期待値が } u \text{ となる最悪ケース}}
$$

ここで、補助関数 $h(u_0, u)$ を以下のように定義します (式 (4.3))。

$$
h\left(u_{0}, u\right)=\inf _{v \in \mathcal{M}_{f,+}(\mathcal{Z})}\left\{-\int_{\mathcal{Z}} \ell(z) \mathrm{d} v(z): \int_{\mathcal{Z}} \mathrm{d} v(z)=u_{0}, \int_{\mathcal{Z}} f(z) \mathrm{d} v(z)=u\right\}
$$


#### 凸共役，双共役 $h^{**}$ の計算

$h$ の凸共役 $h^*$ を計算します。定義より、
$h^*(-\lambda_0, -\lambda) = \sup_{u_0, u} \{ -\lambda_0 u_0 - \lambda^{\top} u - h(u_0, u) \}$ です。これに $h(u_0, u)$ の定義を代入し、sup と inf を入れ替えるような計算(ラグランジュ双対のような計算)を行うと、

$$
h^{*}\left(-\lambda_{0},-\lambda\right) = \begin{cases} 0 & \text { if } \ell(z)-\lambda_{0}-f(z)^{\top} \lambda \leq 0 \quad \forall z \in \mathcal{Z}, \\ \infty & \text { otherwise }, \end{cases}
$$

が得られます。

次に、双共役$h^{**}(u_0, u) = \sup_{\lambda_0, \lambda} \{ -u_0 \lambda_0 - u^{\top} \lambda - h^*(-\lambda_0, -\lambda) \}$ を計算します。$h^*$ が上記の指示関数であることから、sup を取る対象は $h^*(-\lambda_0, -\lambda)=0$ となる $(\lambda_0, \lambda)$、すなわち $\lambda_0 + f(z)^{\top} \lambda \ge \ell(z)$ を満たす $(\lambda_0, \lambda)$ に限定されます。これにより、

$$
h^{* *}\left(u_{0}, u\right)=\sup _{\lambda_{0} \in \mathbb{R}, \lambda \in \mathbb{R}^{m}}\left\{-u_{0} \lambda_{0}-u^{\top} \lambda: \lambda_{0}+f(z)^{\top} \lambda \geq \ell(z) \forall z \in \mathcal{Z}\right\}
$$

が得られます。これが $h$ の双共役です。

#### 双対定理の導出

元の問題は $\sup_{u \in \mathcal{F}} -h(1, u)$ です。

弱い双対性: 常に $h(1, u) \ge h^{**}(1, u)$ なので、
$$
\sup_{u \in \mathcal{F}} -h(1, u) \le \sup_{u \in \mathcal{F}} -h^{**}(1, u) = \sup_{u \in \mathcal{F}} \inf_{\lambda_0, \lambda: \text{制約}} \{ \lambda_0 + u^{\top} \lambda \}
$$
ここで $\lambda_0, \lambda$ は $h^{**}$ の定義にあった制約 $\lambda_0 + f(z)^{\top} \lambda \ge \ell(z)$ を満たすものです。最大最小不等式 ($\sup \inf \le \inf \sup$) を使うと、
$$
\le \inf_{\lambda_0, \lambda: \text{制約}} \sup_{u \in \mathcal{F}} \{ \lambda_0 + u^{\top} \lambda \} = \inf_{\lambda_0, \lambda: \text{制約}} \{ \lambda_0 + \underbrace{\sup_{u \in \mathcal{F}} u^{\top} \lambda}_{\delta_{\mathcal{F}}^{*}(\lambda)} \}
$$これが弱い双対性の式です。右辺が双対問題となります。


### $\phi$-ダイバージェンス不確実性集合に対する双対理論
#### 問題設定
$\phi$-ダイバージェンス不確実性集合 $\mathcal{P}=\left\{\mathbb{P} \in \mathcal{P}(\mathcal{Z}): \mathrm{D}_{\phi}(\mathbb{P}, \hat{\mathbb{P}}) \leq r\right\}$ 上での最悪ケース期待値 $\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)]$ を考えます。

#### 補助関数 $h$ の導入とその凸性

 $h(u_0, u)$ を以下のように定義します (式 (4.7))。

$$
h\left(u_{0}, u\right)=\inf _{v \in \mathcal{M}_{+}(\mathcal{Z})}\left\{-\int_{\mathcal{Z}} \ell(z) \mathrm{d} v(z): \int_{\mathcal{Z}} \mathrm{d} v(z)=u_{0}, \mathrm{D}_{\phi}(v, \hat{\mathbb{P}}) \leq u\right\}
$$

#### 凸共役，双共役の計算
$$
h^{*}\left(-\lambda_{0},-\lambda\right) = \begin{cases}\mathbb{E}_{\hat{\mathbb{P}}}\left[\left(\phi^{*}\right)^{\pi}\left(\ell(Z)-\lambda_{0}, \lambda\right)\right] & \text { if } \lambda \geq 0 \text { and } \lambda_{0}+\lambda \phi^{\infty}(1) \geq \sup _{z \in \mathcal{Z}} \ell(z), \\ \infty & \text { otherwise },\end{cases}
$$
