# このページについて
弱凸関数が満たす，勾配支配条件についてその定理の確認と証明を書きます．弱凸関数は，ロバストMDPにおける，max演算子を含む目的関数が満たす性質の一つであり，勾配支配条件は，目的関数の勾配停留点が大域的な最適解であると示すために必要なものになっています．なお，表記などはすべて[Near-Optimal Policy Identification in Robust Constrained Markov Decision Processes via Epigraph Form](https://arxiv.org/abs/2408.16286)に従っています．

## 定理4：弱凸関数の勾配支配条件
 下の仮定が成り立つ場合，$(\pi, b_0) \in \Pi \times \mathbb{R}$,
$$ \Delta_{b_0}(\pi) - \min_{\pi' \in \Pi} \Delta_{b_0}(\pi') \leq DH \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle \quad \forall g \in \partial \Delta_{b_0}(\pi) $$
ここで，$D = \max_{n, P} \|d_P^{\pi_{n,P}^*} / \mu\|_{\infty}$


## 定理4成立のための準備
### ロバスト制約付き目的関数 
$$
\Delta_{b_0}(\pi) = \max_{n \in [[ 0, N ]]} \max_{P \in \mathcal{U}} J_{c_n, P}(\pi) - b_n 
$$
### 不確実集合の仮定
 $\mathcal{U}$ は以下のいずれかを満たします．
(i) 有限集合
(ii) コンパクト集合かつ，任意の$\pi \in \Pi$ と任意の$n$ に対して，$J_{c_n, P}(\pi)$ の $\pi$ に関する勾配 $\nabla J_{c_n, P}(\pi)$ が、$P \in \mathcal{U}$ に関して連続

### 初期状態分布のカバレッジ
初期状態分布$\mu$は，全ての状態$s \in \mathcal{S}$ に対して $\mu(s) > 0$ を満たします．

### 劣勾配
不確実集合の仮定の下で，$\pi$における$\Delta_{b_0}(\cdot)$の劣勾配は，

$$
\partial \Delta_{b_0}(\pi) = \text{conv} \left\{ \nabla_{\pi} J_{c_n, P}(\pi) \mid (n,P) \in \mathcal{W}_{b_0}(\pi) \right\} 
$$
ここで，$\mathcal{W}_{b_0}(\pi) = \arg\max_{(n',P') \in [[ 0,N ]] \times \mathcal{U}} (J_{c_{n'},P'}(\pi) - b_{n'})$です．

### 通常の(Pが固定された)MDPにおける勾配支配
初期状態分布のカバレッジの仮定の下で，任意の$\pi \in \Pi, P: \mathcal{S} \times \mathcal{A} \rightarrow \mathscr{P}(\mathcal{S})$および，コスト$c \in [0,1]^{SA}$に対して，
$$
J_{c,P}(\pi) - \min_{\pi' \in \Pi} J_{c,P}(\pi') \leq H \left\| \frac{d_P^{\pi_{c,P}^* }}{\mu} \right\|_{\infty} \max_{\pi'' \in \Pi} \langle \pi - \pi'', \nabla J_{c,P}(\pi) \rangle 
$$
ここで，$\pi_{c,P}^* \in \arg\min_{\pi' \in \Pi} J_{c,P}(\pi')$です．

### シオンのミニマックス定理
コンパクトな凸集合 $\mathcal{X}$ と凸集合 $\mathcal{Y}$ 上の関数 $f(x,y)$ が、$x$ に関して準凸，$y$ に関して準凹（かつ適切な連続性を持つ）ならば，
$$
\min_{x \in \mathcal{X}} \sup_{y \in \mathcal{Y}} f(x,y) = \sup_{y \in \mathcal{Y}} \min_{x \in \mathcal{X}} f(x,y) 
$$
minisupをsupminで表現できます．

### 凸包と線形関数の最小化
コンパクト集合 $\mathcal{X}$ 上の線形関数の最小化は，その凸包 $\text{conv}\{\mathcal{X}\}$ 上での最小化と同じ値を取ります．
$$
\min_{x \in \mathcal{X}} \langle c,x \rangle = \min_{x \in \text{conv}\{\mathcal{X}\}} \langle c,x \rangle 
$$

## 定理4の証明

$\mathcal{G}$と$\mathcal{W}$を以下のような略記として導入します．
$$ \mathcal{G} = \left\{ \nabla J_{c_n, P}(\pi) \mid (n,P) \in \mathcal{W}_{b_0}(\pi) \right\} \quad \text{and} \quad \mathcal{W} = \mathcal{W}_{b_0}(\pi) = \underset{(n',P') \in [[ 0,N ]] \times \mathcal{U}}{\arg\max} (J_{c_{n'},P'}(\pi) - b_{n'}) $$
$\mathcal{W}$は，$\Delta_{b_0}(\pi)$の定義における最大値を達成するインデックス$(n,P)$のペアの集合です．$\mathcal{G}$は，それらのペアに対応する勾配の集合である．ここで，$\pi_{b_0}^* \in \arg\min_{\pi \in \Pi} \Delta_{b_0}(\pi)$とします．任意の$\pi \in \Pi$と$b_0 \in \mathbb{R}$に対して，次のように評価を行います．
$$
\begin{align*}
\Delta_{b_0}(\pi) - \Delta_{b_0}(\pi_{b_0}^*)
&= \left( \max_{(n,P) \in [[ 0,N ]] \times \mathcal{U}} (J_{c_n,P}(\pi) - b_n) \right) - \left( \max_{(n',P') \in [[ 0,N ]] \times \mathcal{U}} (J_{c_{n'},P'}(\pi_{b_0}^*) - b_{n'}) \right) \\
&= \left( J_{c_{\hat{n}},\hat{P}}(\pi) - b_{\hat{n}} \right) - \left( \max_{(n',P')} (J_{c_{n'},P'}(\pi_{b_0}^*) - b_{n'}) \right) \quad (\text{ある } (\hat{n},\hat{P}) \in \mathcal{W} \text{ に対して}) \\
&\leq (J_{c_{\hat{n}},\hat{P}}(\pi) - b_{\hat{n}}) - (J_{c_{\hat{n}},\hat{P}}(\pi_{b_0}^*) - b_{\hat{n}}) \quad \\
&= J_{c_{\hat{n}},\hat{P}}(\pi) - J_{c_{\hat{n}},\hat{P}}(\pi_{b_0}^*) \\
&\leq J_{c_{\hat{n}},\hat{P}}(\pi) - \min_{\pi' \in \Pi} J_{c_{\hat{n}},\hat{P}}(\pi') \quad \quad \cdots (36)
\end{align*}
$$
つぎに，(36)の右辺に通常の(Pが固定された)MDPにおける勾配支配を適用します．
$$
J_{c_{\hat{n}},\hat{P}}(\pi) - \min_{\pi' \in \Pi} J_{c_{\hat{n}},\hat{P}}(\pi') \leq H \left\| \frac{d_{\hat{P}}^{\pi_{c_{\hat{n}},\hat{P}}^*}}{\mu} \right\|_{\infty} \max_{\pi'' \in \Pi} \langle \pi - \pi'', \nabla J_{c_{\hat{n}},\hat{P}}(\pi) \rangle
$$
$D = \max_{n,P} \|d_P^{\pi_{n,P}^*} / \mu\|_{\infty}$であるため，$\|d_{\hat{P}}^{\pi_{c_{\hat{n}},\hat{P}}^*} / \mu\|_{\infty} \leq D$です．したがって，
$$ \Delta_{b_0}(\pi) - \Delta_{b_0}(\pi_{b_0}^*) \leq DH \max_{\pi'' \in \Pi} \langle \pi - \pi'', \nabla J_{c_{\hat{n}},\hat{P}}(\pi) \rangle $$
この不等式は，$\Delta_{b_0}(\pi)$の最大値を達成する特定の$(\hat{n},\hat{P})$に対応する勾配$\nabla J_{c_{\hat{n}},\hat{P}}(\pi) \in \mathcal{G}$に対して成り立ちます．よって，以下のような勾配支配が成り立ちます．
$$ \Delta_{b_0}(\pi) - \Delta_{b_0}(\pi_{b_0}^*) \leq DH \min_{g \in \mathcal{G}} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle $$
しかし，主張のために，ミニマックス問題に対する以下の等式を示すことが必要です．
$$ \min_{g \in \mathcal{G}} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle = \min_{g \in \partial \Delta_{b_0}(\pi)} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle \quad \cdots (37) $$
劣勾配の定義より$\partial \Delta_{b_0}(\pi) = \text{conv}\{\mathcal{G}\}$であるので，(37)が成り立つのは，$g^* \in \arg\min_{g \in \text{conv}\{\mathcal{G}\}} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle$となる$g^*$が$\mathcal{G}$の要素として存在する場合です．

$z^* \in \arg\max_{\pi' \in \Pi} \min_{g \in \text{conv}\{\mathcal{G}\}} \langle \pi - \pi', g \rangle$とします．任意の$g^* \in \arg\min_{g \in \text{conv}\{\mathcal{G}\}} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle$に対して，次が成り立ちます．
$$
\begin{align*}
\max_{\pi' \in \Pi} \langle \pi - \pi', g^* \rangle
&= \min_{g \in \text{conv}\{\mathcal{G}\}} \max_{\pi' \in \Pi} \langle \pi - \pi', g \rangle \\
&\stackrel{(a)}{=} \max_{\pi' \in \Pi} \min_{g \in \text{conv}\{\mathcal{G}\}} \langle \pi - \pi', g \rangle \\
&= \min_{g \in \text{conv}\{\mathcal{G}\}} \langle \pi - z^*, g \rangle \quad \cdots (38) \\
&\stackrel{(b)}{=} \min_{g \in \mathcal{G}} \langle \pi - z^*, g \rangle
\end{align*}
$$
(a)ではシオンのミニマックス定理を，$\Pi$および$\text{conv}\{\mathcal{G}\}$の凸性と共に用いています．(b) は凸包と線形関数の最小化を用いています．
最後に，
$$ \langle \pi - z^*, g^* \rangle \leq \max_{\pi' \in \Pi} \langle \pi - \pi', g^* \rangle \stackrel{(a)}{=} \min_{g \in \text{conv}\{\mathcal{G}\}} \langle \pi - z^*, g \rangle \leq \langle \pi - z^*, g^* \rangle \quad \cdots (39) $$
ここの不等式は等号でなければならないので，したがって，
$$ \langle \pi - z^*, g^* \rangle \stackrel{(a)}{=} \max_{\pi' \in \Pi} \langle \pi - \pi', g^* \rangle \stackrel{(b)}{=} \min_{g \in \mathcal{G}} \langle \pi - z^*, g \rangle $$
したがって，$g^* \in \mathcal{G}$であり，式(37)が成り立つ．これで証明完了です．お疲れ様でした．
