# このページについて
このページでは，Distributionally Robust Optimization(https://arxiv.org/abs/2411.02549)の第8章の勉強内容をまとめていこうと思います．

# ロバスト化による正則化
この章のテーマは，DROにおけるロバスト化という考え方と，正則化の間に存在するいつながりを見つけることです．

背景として，**確率最適化**$\inf_{x \in \mathcal{X}} \mathbb{E}_{\hat{\mathbb{P}}}[\ell(x, Z)]$を考えます．しかし，推定された$\hat{\mathbb{P}}$は真の$\mathbb{P}$とは異なる可能性があり，特にデータが少ない場合，$\hat{\mathbb{P}}$に最適化された$x$が，真の$\mathbb{P}$の下ではうまく機能しなくなるような過学習が発生する可能性があります．
一方，DROの定式化$\inf_{x \in \mathcal{X}} \sup_{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)]$は，最悪ケースの分布を考慮して最適化を行います．このようなロバスト化は，$\hat{\mathbb{P}}$の推定誤差の影響を受けにくくし，過学習を防ぐ効果があります．つまり，DROにおけるロバスト化は一種の暗黙的(Implicit)な正則化と見なすことができます．一方で，明示的(Explicit)な正則化とは，目的関数にペナルティ項(正則化項)を追加する手法です．
この章では，DROによる暗黙的な正則化と，ペナルティ項を加える明示的な正則化が，実は数学的に関連しているということ，場合によっては等価になることを示します．
特に，以下に焦点を当てます．
* $\phi$-ダイバージェンス不確実性集合上のDROと**分散正則化**
* Wasserstein不確実性集合上のDROと**変動・リプシッツ正則化**
* モーメント不確実集合に関しては明らかになっていないらしい

## 8.1 $\phi$-ダイバージェンス不確実性集合

まず，具体的な$\phi$-ダイバージェンスであるピアソン$\chi^2$-ダイバージェンスの場合を考えます．不確実性集合は次のように定義されます．
$$
\mathcal{P} = \left\{ \mathbb{P} \in \mathcal{P}(\mathcal{Z}): \chi^2(\mathbb{P}, \hat{\mathbb{P}}) \le r \right\}
$$
$\chi^2$-ダイバージェンスは，命題2.13により以下の双対形式を持ちます．
$$
\chi^2(\mathbb{P}, \hat{\mathbb{P}}) = \sup_{f: \text{有界ボレル}} \frac{\left( \mathbb{E}_{\mathbb{P}}[f(Z)] - \mathbb{E}_{\hat{\mathbb{P}}}[f(Z)] \right)^2}{\mathbb{V}_{\hat{\mathbb{P}}}[f(Z)]}
$$
$\mathbb{V}_{\hat{\mathbb{P}}}[f(Z)]$は$f(Z)$の$\hat{\mathbb{P}}$の下での分散です．この式から，$\mathcal{P}$に属する任意の分布$\mathbb{P}$は，任意の有界ボレル関数$\ell$に対して，
* $\mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)] > 0$
$$
\frac{\left( \mathbb{E}_{\mathbb{P}}[\ell(Z)] - \mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)] \right)^2}{\mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]} \le \chi^2(\mathbb{P}, \hat{\mathbb{P}}) \le r
$$
を満たす必要があります．式を変形させると，
$$
\mathbb{E}_{\mathbb{P}}[\ell(Z)] - \mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)] \le \sqrt{r} \sqrt{\mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]}
$$
が得られます．これは$\mathcal{P}$内の全ての$\mathbb{P}$で成り立つため，最悪ケース期待値をとっても不等式は保存されます．
$$
\sup_{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] \le \mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)] + r^{\frac{1}{2}} \mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]^{\frac{1}{2}}
$$
$r^{\frac{1}{2}} \mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]^{\frac{1}{2}}$は$\hat{\mathbb{P}}$における損失の標準偏差に比例する項です．この項は，損失の分散が大きいほどペナルティが大きくなるため，**分散正則化項**と解釈されます．

上記の議論は$\ell$が有界であることを仮定していましたが，定理8.2は，$\ell$が積分可能 ($\mathbb{E}_{\hat{\mathbb{P}}}[|\ell(Z)|] < \infty$) であれば同じ形の不等式が成り立つことを示します．

### 定理8.2
$\mathcal{P}$がピアソン$\chi^2$-ダイバージェンス不確実性集合 (2.17) であり，$\mathbb{E}_{\hat{\mathbb{P}}}[|\ell(Z)|] < \infty$ならば，
$$
\sup_{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] \le \mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)] + r^{\frac{1}{2}} \mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]^{\frac{1}{2}}
$$

定理8.2はピアソン$\chi^2$の場合の上界を示しましたが，他の滑らかな$\phi$-ダイバージェンスについても，半径$r$が小さいときには同様の関係が成り立ちます．

### 仮定8.3 
エントロピー関数$\phi$は$s=1$の周りで滑らかで，$\phi(1)=0, \phi'(1)=0, \phi''(1)=2$を満たすとします．

この仮定は，$\phi(s)$を$s=1$の周りでテイラー展開すると，
$$
\phi(s) = \phi(1) + \phi'(1)(s-1) + \frac{\phi''(1)}{2}(s-1)^2 + o((s-1)^2) = (s-1)^2 + o((s-1)^2)
$$
となることを意味します．つまり，$s \approx 1$では$\phi(s)$は$(s-1)^2$(ピアソン$\chi^2$のエントロピー関数)として表すことができます．

### 定理8.4(最悪ケース期待値のテイラー展開)
$\mathcal{P}$を**制限付き**$\phi$-ダイバージェンス不確実性集合 (2.11) とし，$\phi$が仮定8.3を満たし，損失$\ell(Z)$が$\hat{\mathbb{P}}$-ほとんど確実に有界であるとします．このとき，$r \to 0$で，
$$
\sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(Z)] = \mathbb{E}_{\hat{\mathbb{P}}}[\ell(Z)] + r^{\frac{1}{2}} \mathbb{V}_{\hat{\mathbb{P}}}[\ell(Z)]^{\frac{1}{2}} + o(r^{\frac{1}{2}})
$$

$r$がとても小さい場合，仮定8.3を満たす多くの一般的な$\phi$-ダイバージェンスを用いたDROの最悪ケース期待値は，$\hat{\mathbb{P}}$に従う損失関数の期待値に分散正則化項を加えたものと一致することを示しています．
