## 第3章：不確実性集合のトポロジー性質

分布ロバスト最適化において、不確実性集合は重要な役割をもっています。不確実性集合を適切に設計しなければ、計算がNP困難になったり最適解にたどり着けないなどの問題が発生する可能性があります。では、そのような問題を回避するためには不確実性集合どのような構造を持つべきでしょうか？

  
分布ロバスト最適化問題をもう一度以下のように示します。

$$

\begin{equation*}

\inf _{x \in \mathcal{X}} \sup _{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)] \tag{1.2}

\end{equation*}

$$

$\mathcal{X}:$ 決定変数 $x$ の取りうる 決定領域
$\mathcal{P}:$ 確率分布 $\mathbb{P}$ を要素とする不確実性集合
$Z$ ：確率変数（または確率ベクトル）
$\ell(x, Z):$  $x$ と $Z$ を入力とする目的関数

この最適化問題が解けるかどうかは、内側最大化が計算可能か、つまり可解かどうかが重要です。例えば分布ロバスト最適化の1つのインスタンスであるロバストMDPにおける強化学習問題を考えます。この問題設定において、不確実性集合が矩形(状態行動ペア又は状態ペアごとに遷移確率が独立)でなければ、**内側最大化の計算が困難になる**ことが言われています。(2005年くらいの論文)  関連研究では内側最大化がほぼ最適に求まるという仮定を置いたり、矩形性を不確実性集合に仮定したりしてこの問題に対処しています。分布ロバスト最適化の問題に戻します。**内側最大化を可解にするために、不確実性集合に必要な性質は何か**？を第3章では示します。

が、結論から言うと、

確率分布の集合 $\mathcal{P}$ が
- **弱位相において定義**されており（＝有界連続関数に対する期待値写像が連続）、
- **弱位相で閉**じていて、
- かつ**タイト**であるとき、
**$\mathcal{P}$ は弱位相においてコンパクト**になります(Prokhorovの定理)。 このとき、期待損失が $\mathbb{P}$ に関して**弱下半連続**であれば
$$\sup_{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)]$$
は、**可解**であることが保証されます。

#### Weierstrass の極値定理
もう少し形式的な話をしましょう。関数の$\text{sup}$問題において重要なのは、ある関数への入力の集合(定義域)内にその関数を最大化する入力が存在しているかどうかでです。都合の良いことに、関数が**連続**関数であり、かつ入力の集合が**コンパクト集合**の場合、必ず最大値が入力の集合内に存在するという定理があります。この定理をWeierstrass の極値定理といいます。以下のように定義します。
***

**Weierstrass の極値定理**

連続関数 $f: K \rightarrow \mathbb{R}$ が，コンパクト集合 $K \subset \mathbb{R}^n$ 上で定義されているとき，

- 最大値 $\max _{x \in K} f(x)$
- 最小値 $\min _{x \in K} f(x)$

が必ず存在する。
