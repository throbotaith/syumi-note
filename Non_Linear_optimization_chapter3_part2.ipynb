# 不等式制約に対する最適性条件
パート1では制約条件をSに属するかどうかでみてきました．ここでは具体的な数式による不等式制約を持つ以下の問題を考えます．
$$
\begin{aligned}
\text{目的関数:} f(x) &→ \text{最小} \\ 
\text{制約条件:} gᵢ(x)& ≤ 0 (i = 1, ..., m)
\end{aligned}
$$
## 有効制約
$gᵢ(x̄) = 0$となる$x̄$を有効制約と言います．なぜ重要かというと，$x̄$を少し動かすと制約条件を破る可能性があるからです．つまり，有効制約はパート1におけるSの境界のようなものを表します．また，$x̄$ において $gᵢ(x̄) = 0$ となる集まりを $I(x̄)$ と書き，**有効制約集合 (Active Set)** $I(x̄) = (i | gᵢ(x̄) = 0)$と呼びます．

## 線形化錐と接錐
パート1のように錐を用いて幾何学的な理解を得ましょう．接錐はSから定義されました．これの不等式制約バージョンを考えましょう．それをを表現する一つとして線形化錐があります．以下の図と一緒に理解しましょう．

**定義**：点$x̄$における線形化錐$Cs(x̄)$とは，すべての有効制約$i ∈ I(x̄)$に対して，その勾配$∇gᵢ(x̄)$との**内積が0以下**となるような方向ベクトル$y$の集まりです．
$$
C_s(\bar{x})=\left\{y \in \mathbb{R}^x \mid\left(\nabla g_i(\bar{x}), y\right) \leq 0 \text { for all } i \in I(\bar{x})\right\}
$$

![](https://cdn.mathpix.com/snip/images/S8LOAKTeY_FvaObKMvsQJbevpQ3JGT0IBS1v6VuqQEY.original.fullsize.png)
## 接錐との関係性：
常に線形化錐は接錐を含みます($Ts(x̄) ⊆ Cs(x̄)$，補題3.3)，が等価とは限りません．

## 最適性条件
不等式制約問題における最適性の必要条件を考えましょう．何を満たせば最適と言えるでしょうか？最初に先ほどの錐と絡めて**制約想定**というものを考えましょう．

## 制約想定
制約想定とは，線形化錐が接錐を適切に近似できているかどうかを保証する条件です．つまり，以下のような場合をいいます．
$$
C_S(\bar{x}) \subseteq \operatorname{co} T_S(\bar{x})
$$
これがないと後のKKT条件における最適性の必要条件をいえません．実例としてはスレーター制約想定が有名です．これは，これは後のパートで詳しく説明します．

## 最適性の必要条件
最適性の必要条件であるKarush-Kuhn-Tucker条件(KKT条件)を示します．最初にラグランジュ関数$Lo(x, λ)$を定義します．($λᵢ ≥ 0$)
$$
Lo(x, λ) = f(x) + \Sigma_{i} λᵢgᵢ(x)
$$
ラグランジュ関数は，制約付き最適化問題を制約なしの最適化問題に変換することができます．$\lambda$はラグランジュ乗数といって，これを最適化します．

## KKT条件
$x̄$が局所最適解であり，$f$と$gᵢ$が$x̄$で微分可能，適切な制約想定が$x̄$で成り立っているならば，以下の条件をすべて満たすラグランジュ乗数のベクトル$\bar{\lambda} = (\bar{\lambda_{1}}, ...,)$が存在します．

(1)停留条件：
$$
\nabla f(\bar{x})+\sum_i \bar{\lambda}_i \nabla g_i(\bar{x})=0
$$
(2)**相補性条件**：
$$
\bar{\lambda}_i g_i(\bar{x})=0 \text { for all } i=1, \ldots, m
$$
* めっちゃ重要です．$λᵢ ≥ 0$かつ$gᵢ(x̄) ≤ 0$nなので，この積が0になるということは，以下のいずれかが成立することと同義です．
	* $\bar{λᵢ} = 0$
	* $gᵢ(x̄) = 0$
* 1つ目は，有効制約で無いつまり不等式のみの制約に対応するラグランジュ乗数は必ず0になります．逆に，ラグランジュ乗数が正であるならば，その制約は必ず有効制約です．
(3)主問題の実行可能性：
$$
g_i(\bar{x}) \leq 0 \text { for all } i=1, \ldots, m
$$
(4)双対問題の実行可能性：
$$
\bar{\lambda}_i \geq 0 \text { for all } i=1, \ldots, m
$$
KKT条件における停留条件を式変形してみましょう．
$$
\begin{aligned}
&\nabla f(\bar{x})+\sum_i \bar{\lambda}_i \nabla g_i(\bar{x})=0\\
&-\nabla f(\bar{x})=\sum_i \bar{\lambda}_i \nabla g_i(\bar{x})
\end{aligned}
$$
ここで，右辺$\sum_i \bar{\lambda}_i \nabla g_i(\bar{x})$は，点$\bar{x}$での各制約の勾配によって張られる凸多面体です．それに，$\nabla f(\bar{x})$が存在しているということということを言います．これはつまり，目的関数$f(x)$の値をこれ以上小さくできるような実行可能な方向が、点 $x̄$ の近傍には存在しないということを示しています．以下の図が参考になります！

![](https://cdn.mathpix.com/snip/images/mOMZHWj6BwyYeruIN4Qh7LXzuhjYre2L96Sdygoi1xg.original.fullsize.png)
KKT条件は
$$
\begin{aligned}
\text{目的関数:} f(x) &→ \text{最小} \\ 
\text{制約条件:} gᵢ(x)& ≤ 0 (i = 1, ..., m)
\end{aligned}
$$
におけいて$\bar{x}$局所最適解であるための必要条件(局所解になるための条件の一つだけど，KKT条件が成り立ったからと言って無条件に局所最適解といえるわけじゃない)です．
### KKT条件が大域的最適性を満たす場合
一方で，目的関数と制約関数が凸関数ならば，KKT条件が最適性の十分条件を満たします．つまり，以下が言えます．

---
$f: \mathbb{R}^n \rightarrow \mathbb{R}$と制約関数$g_i: \mathbb{R}^n \rightarrow \mathbb{R} \quad(i=1, \ldots, m)$ は微分可能な凸関数とします．$\bar{x} \in \mathbb{R}^n$ と $\bar{\lambda} \in \mathbb{R}^m$ がKKT条件を満たすならば，$\bar{x}$ は大域的最適解となります．

---
つまり，凸最適化問題の場合制約想定がいい感じに言えていればKKT条件が大域的最適性を得るための必要十分条件と言えるでしょう．


# 制約想定
前回までは以下のような制約想定を考えていました．ちなみにこの制約想定は，Guignard 制約想定と言われています．ほかにも制約想定があるので見ていきましょう．
$$
C_S(\bar{x}) \subseteq \operatorname{co} T_S(\bar{x})
$$

**一次独立制約想定**

ベクトル$\nabla g_i(\bar{x})(i \in \mathcal{I}(\bar{x}))$が，一次独立であるという制約想定です．

**スレーター制約想定**

$g_i(i \in \mathcal{I}(\bar{x}))$ は凸関数であり，$g_i\left(\boldsymbol{x}^0\right)<0(i=1, \ldots, m)$ を満たす $\boldsymbol{x}^0$ が存在するという制約想定です．

**Cottle制約想定**

$\left\langle\nabla g_i(\overline{\boldsymbol{x}}), \boldsymbol{y}\right\rangle<0$ $(i \in \mathcal{I}(\bar{x}))$ を満たすべクトル $y \in \mathbb{R}^n$ が存在するという制約想定です．

**Abadie制約想定**

$$
C_S(\bar{x}) \subseteq T_S(\bar{x})
$$

**Guignard制約想定**
先ほどの節で確認した制約想定そのもです．
$$
C_S(\bar{x}) \subseteq \operatorname{co} T_S(\bar{x})
$$
## 制約想定の性質
### **成立の連鎖**
 * 一次独立制約想定または，スレーター制約想定成立
$$
\downarrow
$$
* Cottle制約想定が成立
$$
\downarrow
$$
* Abadie制約想定が成立
$$
\downarrow
$$
* Guignard制約想定が成立

### **成り立つ定理**
 * いずれかの制約想定が成り立てば，KKT条件を満たすLagrange乗数$\bar{\lambda} \in \mathbb{R}^m$が存在します．
 * 1次独立制約想定が成立すれば、KKT条件を満たすLagrange乗数$\bar{\lambda} \in \mathbb{R}^m$ はただ一つになります．
 * Cottle 制約想定が成立すれば、 KKT 条件を満たすLagrange乗数$\bar{\lambda} \in \mathbb{R}^m$ の集合は有界になります．
# 鞍点定理
この節では、最適化問題
$$ \min_{x \in \mathbb{R}^n} f(x) \quad \text{s.t.} \quad g_i(x) \le 0 \quad (i=1, \ldots, m) $$
における目的関数$f$と制約関数$g_i$に対して、これまで仮定してきた微分可能性の代わりに**凸性**を仮定します。そして、ラグランジュ関数の**鞍点**という概念を用いて、この問題の最適性条件を導きます。この鞍点を用いた条件は、もし関数が微分可能であれば、3.2節で学んだKKT条件と等価になるという性質を持ちます．

### 鞍点
ラグランジュ関数を考えます．
$$ L_0(x, \lambda) = f(x) + \sum_{i=1}^m \lambda_i g_i(x) $$
 $L_0: \mathbb{R}^{n+m} \rightarrow [-\infty, +\infty]$ に対して、ある点 $(\bar{x}, \bar{\lambda})^{\top} \in \mathbb{R}^{n+m}$が以下の式を満たすとき、その点を$L_0$ の**鞍点**と呼びます。
 
$$
L_0(\bar{x}, \lambda) \le L_0(\bar{x}, \bar{\lambda}) \le L_0(x, \bar{\lambda}) \quad (x \in \mathbb{R}^n, \lambda \in \mathbb{R}^m, \lambda \ge 0) 
$$

鞍点 $(\bar{x}, \bar{\lambda})$ は、一方の変数については最大化を達成し、もう一方の変数については最小化を達成するような点と言えます。馬の鞍のように、ある方向から見れば谷の底、別の方向から見れば山の頂上のようなイメージです。

### 鞍点定理

最適化問題において、以下の条件が満たされていると仮定します。

1.  目的関数 $f: \mathbb{R}^n \rightarrow \mathbb{R}$ は**凸関数**である。
2.  制約関数 $g_i: \mathbb{R}^n \rightarrow \mathbb{R}$ $(i=1, \ldots, m)$ はすべて**凸関数**である。
3.  Slater制約想定: ある点 $x^0 \in \mathbb{R}^n$ が存在し、すべての制約を厳密に満たす。すなわち、$g_i(x^0) < 0$ $(i=1, \ldots, m)$ が成り立つ。

このとき、点 $\bar{x} \in \mathbb{R}^n$ が最適化問題の最適解であるための必要十分条件は、あるベクトル $\bar{\lambda} \in \mathbb{R}^m$ で $\bar{\lambda} \ge 0$ を満たすものが存在し、$(\bar{x}, \bar{\lambda})^{\top}$ がラグランジュ関数 $L_0$ の鞍点となることです．
