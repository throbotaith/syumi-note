# このページについて
このページでは，Distributionally Robust Optimization(https://arxiv.org/abs/2411.02549)の第6章の勉強内容をまとめていこうと思います．ただしこのページでは，8章における正則化におけるロバスト化を読むための準備の位置づけとします．よって6章の概説をしつつ，8章とつながりの強い節の解説を主たる内容とします．

## Nature’s Subproblem(自然のサブ問題)の解析解
### 自然のサブ問題とは
DROは，与えられた$x$に対して最悪ケースの確率分布$\mathbb{P}$見つけ，その中での期待損失$\sup_{\mathbb{P} \in \mathcal{P}} \mathbb{E}_{\mathbb{P}}[\ell(x, Z)]$の評価をする必要があります．この一連の計算を自然のサブ問題や内部問題と言います．しかし，この問題は確率分布を扱うため，これが無限次元の可能性があることから，この内部問題の最大化は一般的には解くことは難しいです．既存研究では，内部問題がε-最適に解けているなどの仮定を置いて解析することが多いです．そんな内部問題ですが，特定の状況下においては，この内部問題を解析的に解けるということを示すことができます．

### アプローチ
解析解を得るために，主問題と双対問題を構成します．さらにそれらに対して実行可能解を導出し，それらの目的関数が一致することを示します．これにより，弱双対性が成り立ち，両問題にとって解が最適解であるということを示すことができます．

### 自然のサブ問題を解析的に解くための具体例
では，特定の状況というのはどういったものでしょうか？それを6章では詳しく見ていきます．なお，8章と関連のあるもののみ今後詳しく取り上げております．なにとぞご了承くださいませ．

* 平均値のみが既知(**マルコフ不確実性集合**)で，損失関数が**凹or凸**関数(6.1，6.2節)
* 平均値と平均絶対偏差が既知で損失関数が**凸**関数(6.4節)
* 平均と分散が既知(**チェビシェフ不確実性集合**)で，損失関数がランプ関数などの**特定の形**(6.5節)
* KLダイバージェンスで測定した距離が一定以下(6.9節)
* 総変動距離で測定した距離が一定以下(6.10節)
* Wasserstein距離で測定した距離が一定以下(6.12～6.15節)

このページでは，特に**KLダイバージェンスで測定した距離が一定以下**(6.9節)と，**Wasserstein距離で測定した距離が一定以下**(6.12～6.15節)の部分について詳しく説明いたします．

## KLダイバージェンスで測定した距離が一定以下(6.9節)
この節では，$\mathbb{P}$が完全にわかっていない状況で，ある損失関数$\ell(Z)$の期待値$\mathbb{E}_{\mathbb{P}}[\ell(Z)]$を考えます．特に，「真の分布$\mathbb{P}$は，**基準分布$\hat{\mathbb{P}}$** に**近い**だろう」という情報を持っている場合を考えます．

### 主問題
ここで，距離的指標として$\operatorname{KL}(\mathbb{P}, \hat{\mathbb{P}})$を用います．この節で解きたいのは，以下の問題です．

$$
\sup _{\mathbb{P} \in \mathcal{P}(\mathcal{Z})}\left\{\mathbb{E}_{\mathbb{P}}[\ell(Z)]: \operatorname{KL}(\mathbb{P}, \hat{\mathbb{P}}) \leq r\right\} 
$$
制約を満たすKLダイバージェンス不確実性集合(KL球ともいう)を，$\mathcal{P} = \{\mathbb{P} \in \mathcal{P}(\mathcal{Z}): \operatorname{KL}(\mathbb{P}, \hat{\mathbb{P}}) \leq r\}$と定義します．

**仮定：**$\operatorname{KL}(\mathbb{P}, \hat{\mathbb{P}})$が有限の値になるためには，$\mathbb{P}$は$\hat{\mathbb{P}}$に対して**絶対連続** ($\mathbb{P} \ll \hat{\mathbb{P}}$) である必要があります．

今後の話では，以下の具体的な状況設定を考えます．
*   サポート:$\mathcal{Z} = \mathbb{R}^d$（$d$次元実数空間全体）
* $\hat{\mathbb{P}}$: 平均$\hat{\mu}$，共分散行列$\hat{\Sigma}$（$\hat{\Sigma}$は正定値）を持つ**正規分布$N(\hat{\mu}, \hat{\Sigma})$
* $\ell(Z)$: 線形関数$\ell(Z) = \theta^\top Z$（$\theta \in \mathbb{R}^d$はゼロでない定数ベクトル）
*   KLダイバージェンスの閾値:$r > 0$

### 双対問題
主問題には対応する**双対問題**が存在します．主問題との間には弱双対性があります．

$$
\inf _{\lambda_{0} \in \mathbb{R}, \lambda \in \mathbb{R}_{+}} \lambda_{0}+\lambda(r-1)+\mathbb{E}_{\hat{\mathbb{P}}}\left[\lambda \exp \left(\frac{\ell(Z)-\lambda_{0}}{\lambda}\right)\right] 
$$

この節の最も面白い点は，上記の具体的な設定の下で，主問題と双対問題の**両方の解が具体的に求まり，かつそれらの最適値が一致する（強双対性が成り立つ）** ということです．(命題6.12)

**主問題の解（最悪ケース分布$\mathbb{P}^{\star}$）:**

最も期待損失を大きくする分布$\mathbb{P}^{\star}$は，**正規分布**$N(\mu^{\star}, \Sigma^{\star})$です．
*   **共分散行列:**$\Sigma^{\star} = \hat{\Sigma}$（基準分布と同じ）．
*   **平均:**$\mu^{\star} = \hat{\mu} + \frac{\sqrt{2r}}{\sqrt{\theta^{\top} \hat{\Sigma} \theta}} \hat{\Sigma} \theta$．
	*   基準分布の平均$\hat{\mu}$を，$\ell(Z)=\theta^\top Z$が大きくなる方向に（$\hat{\Sigma} \theta$の方向）動かしたものです．
	*   動かす大きさは$r$が大きいほど，また損失の分散$\theta^\top \hat{\Sigma} \theta$が小さいほど，大きくなります．

**双対問題の解 ($\lambda_0^{\star}, \lambda^{\star}$):**
最適な双対変数$\lambda^{\star}$と$\lambda_0^{\star}$も以下のように具体的に求まります．
* $\lambda^{\star} = \frac{\sqrt{\theta^{\top} \hat{\Sigma} \theta}}{\sqrt{2r}}$
* $\lambda_0^{\star} = \lambda^{\star} \log \mathbb{E}_{\hat{\mathbb{P}}}[\exp(\theta^\top Z / \lambda^{\star})]$

**強双対性:**
主問題の最大値と双対問題の最小値は一致し，その値は以下のようになります．
$$
\hat{\mu}^{\top} \theta + \sqrt{2r} \sqrt{\theta^{\top} \hat{\Sigma} \theta}
$$

## Wasserstein距離で測定した距離が一定以下(6.12節)
### 主問題
1-Wasserstein距離($W_1(\mathbb{P}, \hat{\mathbb{P}})$)は，最適輸送の概念における距離です．例えば$\hat{\mathbb{P}}$をもう一方の$\mathbb{P}$に輸送するために必要な最小のコスト（距離×輸送量）を表します．この節で解きたいのは以下の問題になります．

$$
\sup _{\mathbb{P} \in \mathcal{P}(\mathcal{Z})}\left\{\mathbb{E}_{\mathbb{P}}[\ell(Z)]: \mathrm{W}_{1}(\mathbb{P}, \hat{\mathbb{P}}) \leq r\right\} \tag{6.18a}
$$

制約を満たす1-Wasserstein不確実性集合(1-Wasserstein球ともいう)を，$\mathcal{P} = \{\mathbb{P} \in \mathcal{P}(\mathcal{Z}): \mathrm{W}_{1}(\mathbb{P}, \hat{\mathbb{P}}) \leq r\}$と定義します．
今後の話では，以下の具体的な状況設定を考えます．
* サポート：$\mathcal{Z} = \mathbb{R}^d$（$d$次元実数空間全体）
* $\hat{\mathbb{P}}$：$\mathcal{P}(\mathbb{R}^d)$内の任意の分布（特定の形は仮定しないが，$\mathbb{E}_{\hat{\mathbb{P}}}[\ell(\hat{Z})]$が有限であることを仮定）
* $\ell(Z)$：凸関数であり，かつリプシッツ連続
* $r \ge 0$．

### 双対問題
$$
\inf _{\lambda \geq 0} \lambda r+\mathbb{E}_{\hat{\mathbb{P}}}\left[\sup _{z \in \mathcal{Z}} \ell(z)-\lambda\|z-\hat{Z}\|\right] \tag{6.18b}
$$
