## PD法の収束解析
凹CMDPにおけるPrimal Dual Policy Gradient(PDPG)法の収束解析をわかりやすくまとめます。最初に、収束解析の証明(定理4.4)を示すための定義、仮定、補題を示します。補題の証明は別途行います。


### 定義


### 仮定
強そうな仮定たちですが、仮定2と3は標準的なモノです。

#### 仮定1：スレーター条件
次が成り立つような$\widetilde{\theta}$と$\xi>0$が存在する。
$$
g(\lambda(\widetilde{\theta})) \geq \xi
$$
* パラメータの中に何かしら制約を満たすモノが存在するという仮定です。

#### 仮定2：パラメータ化における設定
softmaxパラメータ化を次のように行い、$\psi(\cdot ; s, a)$としてパラメータ化された方策を表現する。
$$
\begin{equation*}
\pi_{\theta}(a \mid s)=\frac{\exp (\psi(\theta ; s, a))}{\sum_{a^{\prime} \in \mathcal{A}} \exp \left(\psi\left(\theta ; s, a^{\prime}\right)\right)}, \forall(s, a) \in \mathcal{S} \times \mathcal{A} \tag{2.10}
\end{equation*}
$$
**仮定2.1(微分可能性)**：$\forall (s, a) \in \mathcal{S} \times \mathcal{A}$に対して、$\psi(\cdot ; s, a)$は$C^2$級(二回微分可能)である。

**仮定2.2(勾配の有界性)**：次を満たす$M_{\pi, 1}, M_{\pi, 2}>0$が存在する。
$$
\begin{equation*}
\max _{(s, a) \in \mathcal{S} \times \mathcal{A}} \sup _{\theta \in \mathbb{R}^{K}}\left\|\nabla_{\theta} \psi(\theta ; s, a)\right\| \leq M_{\pi, 1}, \quad \max _{(s, a) \in \mathcal{S} \times \mathcal{A}} \sup _{\theta \in \mathbb{R}^{K}}\left\|\nabla_{\theta}^{2} \psi(\theta ; s, a)\right\| \leq M_{\pi, 2} 
\end{equation*}
$$
* $\psi$に対して、その一次勾配、二次勾配が最も大きくなるように$s,a,\theta$を選んできても$M_{\pi, 1}$、$M_{\pi, 2}$で上から抑えられるということです。勾配が有界なのは結構ある仮定です。

**仮定2.3(隠れ凸性のための$\lambda(\cdot)$の設定)**：

---
$\forall \theta \in \mathbb{R}^{K}$に対して、$\mathcal{U}_{\theta}$ と $\mathcal{V}_{\lambda(\theta)}:=\lambda\left(\mathcal{U}_{\theta}\right)$の関係が全単射であるような近傍$\mathcal{U}_{\theta} \ni \theta$が存在する。
* 局所的に見れば、$\lambda(\cdot)$が可逆という仮定です。

---
$\lambda_{\mathcal{V}_{\lambda(\theta)}}^{-1}: \mathcal{V}_{\lambda(\theta)} \rightarrow \mathcal{U}_{\theta}$を$\lambda(\cdot)$の局所的な逆写像とする。$\forall \theta \in \mathbb{R}^{K}$に対して$\lambda_{\mathcal{V}_{\lambda(\theta)}}^{-1}(\cdot)$が$L_{\lambda}$-リプシッツ連続であるような普遍的な定数$L_{\lambda}$が存在する
* 逆写像$\lambda_{\mathcal{V}_{\lambda(\theta)}}^{-1}(\cdot)$はリプシッツ連続、つまり微分可能という仮定です。コレは多分、のちの証明で使うのだろうと思われます。

---
$(1-\varepsilon) \lambda(\theta)+\varepsilon \lambda\left(\theta^{\star}\right) \in \mathcal{V}_{\lambda(\theta)}, \forall \varepsilon \leq \bar{\varepsilon}, \forall \theta \in \mathbb{R}^{K}$であるような$\bar{\varepsilon}>0$が存在する。
* コレよくわからんですね。後の証明で使う用？分かり次第更新します。


#### 仮定3：目的関数・制約関数の特性
次を満たす$M_{\lambda}, \ell_{\lambda}>0$が存在する。
* $f(\cdot)$と$g(\cdot)$は$\forall \lambda$に対して次が成り立つ($M_{\lambda}$-リプシッツ連続)。
$$
\begin{align*}
    \left\|\nabla_{\lambda} f(\lambda)\right\| \leq M_{\lambda},\left\|\nabla_{\lambda} g(\lambda)\right\| \leq M_{\lambda}
\end{align*}
$$

*  $f(\cdot)$と$g(\cdot)$は、$\forall \lambda_{1}, \lambda_{2}$に対して$\ell_{\lambda}$-滑らかである。つまり次が成立する。

$$
\begin{align*}
  & \left\|\nabla_{\lambda} f\left(\lambda_{1}\right)-\nabla_{\lambda} f\left(\lambda_{2}\right)\right\|_{\infty} \leq \ell_{\lambda}\left\|\lambda_{1}-\lambda_{2}\right\|\\
& \left\|\nabla_{\lambda} g\left(\lambda_{1}\right)-\nabla_{\lambda} g\left(\lambda_{2}\right)\right\|_{\infty} \leq \ell_{\lambda}\left\|\lambda_{1}-\lambda_{2}\right\|  
\end{align*}
$$
