## 凸解析の基礎
超平面から劣勾配まで、勉強した内容をまとめています。


このページでは、非線形最適化の基礎という本について、分からなかった部分をまとめて、自分に向けた解説として書いていきます。本から引用するものは、定理や補題などの事実のみであり、それらに対する解釈は著作権侵害に当たるため書いていません。

### 超平面
#### 部分空間
$W \subseteq \mathbb{R}^n$とします．$W$が部分空間(subspace)であるとは，$W$に属する任意の要素の和が$W$の中にいるということ，任意の要素を任意の実数で書けたとしても$W$の中に存在すること，原点つまり$0$が$W$に属しているということが言えることです．

二次元平面の部分空間の例です．斜めになってる平面が部分空間です．

#### 部分空間を平行移動したものがアフィン集合
アフィンとは何でしょうか？アフィンとは疑似という意味です．疑似ということは何かに対する模倣なわけです．空間論における疑似とは空間に対する模倣です．疑似なのでユークリッド空間などとは異なる部分があります．例えば原点，長さ，角度などの概念がありません．これを疑似空間（アフィン空間，アフィン集合？）と言います．

ここでは，アフィン集合について考えましょう．部分空間はユークリッド空間の性質を持っているので必ず原点を含んでいなければいけませんでした．アフィン集合は部分空間をより制約少なく動かしたものと言えます．アフィン集合は疑似空間なので原点を含まないといけないという制約がありません．例えば$z$軸方向に原点から部分空間を平行移動させたりすることが可能です．

**三次元での例**

アフィン集合は以下のような形で表現されます．

$$
\left\{x \in \mathbb{R}^3 \mid x=x^0+\alpha_1 x^1+\alpha_2 x^2+\alpha_3 x^3, \alpha_1, \alpha_2, \alpha_3 \in \mathbb{R}\right\}
$$

$x$は原点$x^0$も和に加わっています．一方で部分空間は以下のような形で表現されます．

$$
\left\{x \in \mathbb{R}^3\left|x=\alpha_1 x^1+\alpha_2 x^2+\alpha_3 x^3\right| \alpha_1, \alpha_2, \alpha_3 \in \mathbb{R}\right\}
$$

#### 分離超平面

超平面が分ける2つの空間(超平面によって定義される半空間)を考えます。
$$
\begin{aligned}
H^{+} & =\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle a, x\rangle \leqq \alpha\right\} \\
H^{-} & =\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle a, \boldsymbol{x}\rangle \leqq \alpha\right\}
\end{aligned}
$$
半空間にそれぞれ属する集合 $S, T \subseteq \mathbb{R}^n$ が $S \subseteq H^{+}$かつ $T \subseteq H^{-}$を満たすとき，すなわち
$$
\begin{array}{ll}
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \leqq \alpha & (\boldsymbol{z} \in S) \\
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \leqq \alpha & (\boldsymbol{z} \in T)
\end{array}
$$
超平面は$S$、$T$を分離するといい、分離超平面と呼ばれます。

**定理**
凸集合$S$および、その凸包に属さない点$x$について、以下の式を満たす$S$ と $x$ の分離超平面 $H=\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle\boldsymbol{a}, \boldsymbol{x}\rangle=\alpha\right\}$が存在します
。
$$
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \geqq \alpha>\langle\boldsymbol{a}, \boldsymbol{x}\rangle \quad(\boldsymbol{z} \in S)
$$

* $z$の内積 $\geq \alpha$
*  $x$の内積$< \alpha$
 $S$の点（$z$）は $H$ より上
 ↑
 ｜   ● $z ∈ S$
 ｜   
 ｜――――――――――――――――――― 超平面 $H$（内積 $= α$）
 ｜   
 ｜          ● $x$（外の点）は$H$より下
 ↓

#### 支持超平面
凸包に属する$x$が超平面にも属している場合、それを支持超平面といいます。
****
凸集合 $S \subseteq \mathbb{R}^n$ が超平面 $H$ によって定義される半空間 $H^{+}$または $H^{-}$ に含まれ，さらに点 $\boldsymbol{x} \in \mathrm{cl} S$ が $H$ に含まれるとき，$H$ を $\boldsymbol{x}$ における $S$ の支持超平面（supporting hyperplane）といいます。

#### 支持超平面定理3連発

**定理**
凸集合の境界の点はどこの点でも支持超平面を持ちます。←劣勾配(超平面を勾配をとりたい点に近づけていく)が存在するために凸性が必要な理由です。
****
空でない凸集合 $S \subseteq \mathbb{R}^n$ は，その境界の任意の点において支持超平面をもちます．すなわち， $\boldsymbol{x} \in \operatorname{bd} S$ ならば

$$
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \geqq \alpha=\langle\boldsymbol{a}, \boldsymbol{x}\rangle \quad(\boldsymbol{z} \in S)
$$
を満たす $a \in \mathbb{R}^n(a \neq 0)$ と $\alpha \in \mathbb{R}$ が存在します．
****

**定理**
任意の凸集合と、それに属さない点に対してどの点でも凸集合と分離できる分離超平面があります。
****
任意の空でない凸集合 $S \subseteq \mathbb{R}^n$ と任意の点 $\boldsymbol{x} \notin S$ に対して，$S$ と $x$ の分離超平面が存在します．
****

**定理**
凸集合同士が交わっていない場合、その凸集合同士を分離できる超平面があります。

****
空でない二つの凸集合 $S, T \subseteq \mathbb{R}^n$ の共通集合 $S \cap T$ が空集合 であれば，$S$ と $T$ の分離超平面が存在します．
****

### 平均値、テイラー、陰関数定理

**平均値の定理**
関数の差分を内積で表現できます。これをつかうと、関数の勾配に関する一階条件を導出できます。
****
［平均値定理（mean value theorem）］関数 $f: \mathbb{R}^n \rightarrow \mathbb{R}$ は凸集合 $S \subseteq \mathbb{R}^n$ において微分可能とすします．そのとき，任意の $\boldsymbol{x}, \boldsymbol{y} \in S$ に対して

$$
f(\boldsymbol{x})-f(\boldsymbol{y})=\langle\nabla f(\tau \boldsymbol{x}+(1-\tau) \boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\rangle
$$


を満足する実数 $\tau \in(0,1)$ が存在します．
****
**テイラーの定理**
関数の差分を内積で表現できます。対象が2回微分可能かどうかが平均値の定理との違いです。
****
関数 $f: \mathbb{R}^n \rightarrow \mathbb{R}$ は凸集合 $S \subseteq \mathbb{R}^n$ において 2 回連続的微分可能とすします．そのとき，任意の $\boldsymbol{x}, \boldsymbol{y} \in S$ に対して次式を満足する実数 $\tau \in(0,1)$ が存在します。
$$
f(\boldsymbol{x})=f(\boldsymbol{y})+\langle\nabla f(\boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\rangle+\frac{1}{2}\left\langle\nabla^2 f(\tau \boldsymbol{x}+(1-\tau) \boldsymbol{y})(\boldsymbol{x}-\boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\right\rangle
$$
****


**陰関数定理**
関数の一部の定義域をみると、そこだけ別の関数で表現できることがあります。ラグランジュ、最適性条件、双対性とかにでてきます。
****

関数 $f_i: \mathbb{R}^{n+p} \rightarrow \mathbb{R}(i=1, \ldots, n)$ が与えられており，
ある点 $\bar{x} \in \mathbb{R}^n, \bar{u} \in \mathbb{R}^p$ において，

$$
f_i(\bar{x}, \bar{u})=0 \quad(i=1, \ldots, n)
$$


が成り立っているとします。
また，各 $f_i$ は点 $(\bar{x}, \bar{u})$ のある近傍で連続的に微分可能であり，
次のヤコビ行列（ $\boldsymbol{x}$ に関する偏導関数からなる行列）：

$$
\nabla_x \boldsymbol{f}(\boldsymbol{x}, \boldsymbol{u})=\left[\begin{array}{ccc}
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_n}{\partial x_1} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_1}{\partial x_n} & \cdots & \frac{\partial f_n}{\partial x_n}
\end{array}\right] \in \mathbb{R}^{n \times n}
$$


が点 $(\bar{x}, \bar{u})$ において正則（逆行列を持つ）であるとします。

そのとき ：
点 $\bar{u}$ の適当な近傍 $U \subset \mathbb{R}^p$ において，
連続的微分可能な関数

$$
\phi_i: U \rightarrow \mathbb{R} \quad(i=1, \ldots, n)
$$


が存在して，

$$
f_i\left(\phi_1(u), \ldots, \phi_n(u), u\right)=0 \quad \text { for all } u \in U
$$


が成り立ちます。
また，$\phi_i(\bar{u})=\bar{x}_i$ となります。
****

### 凸関数とその性質



グラフ、エピグラフ、定義域を以下のように定義します。

$$
\text { graph } f=\left\{(\boldsymbol{x}, \beta)^{\top} \in \mathbb{R}^{n+1} \mid \beta=f(\boldsymbol{x})\right\}
$$
$$
\text { epi } f=\left\{(\boldsymbol{x}, \beta)^{\top} \in \mathbb{R}^{n+1} \mid \beta \geqq f(\boldsymbol{x})\right\}
$$
$$
\operatorname{dom} f=\left\{x \in \mathbb{R}^n \mid f(x)<+\infty\right\}
$$

エピグラフが凸集合になる$f: \mathbb{R}^n \rightarrow[-\infty,+\infty]$を**凸関数**といいます。$f: \mathbb{R}^n \rightarrow(-\infty,+\infty]$ かつ $\operatorname{dom} f \neq \emptyset$ であるとき，**真凸関数**と言います。
* 真凸関数について、$f$のマイナス方面の定義域が開なのは、最小値が発散しないために必要な条件です。多分。
#### イェンセンの不等式と凸関数になるための必要十分条件
* 凸関数が成立⇔イェンセンの不等式が成立
関数 $f: \mathbb{R}^n \rightarrow(-\infty,+\infty]$ が凸関数であるための必要十分条件 は，任意の自然数 $m$ と任意の $x^1, \ldots, x^m \in \mathbb{R}^n$ および $\sum_{i=1}^m \alpha_i=1$ を満たす任意の $\alpha_i \geqq 0(i=1, \ldots, m)$ に対して次式が成立することでです。

$$

f\left(\sum_{i=1}^m \alpha_i x^i\right) \leqq \sum_{i=1}^m \alpha_i f\left(x^i\right)

$$
* この不等式をイェンセンの不等式という。証明でよく使います。
#### 狭義凸関数
* $f((1-\alpha) x+\alpha \boldsymbol{y})<(1-\alpha) f(\boldsymbol{x})+\alpha f(\boldsymbol{y})$が$\boldsymbol{x}, \boldsymbol{y} \in \operatorname{dom} f$ と $\alpha \in(0,1)$に対して成り立つとき、狭義凸関数といいます。
* これだけだと良く分からないので、可視化してみましょう。これは、任意の端点同士の直線よりもグラフが常に下にあるということを示しています。

#### 強凸関数(一様凸関数)
* $f((1-\alpha) \boldsymbol{x}+\alpha \boldsymbol{y}) \leqq(1-\alpha) f(\boldsymbol{x})+\alpha f(\boldsymbol{y})-\frac{1}{2} \sigma \alpha(1-\alpha)\|\boldsymbol{x}-\boldsymbol{y}\|^2$がすべての$\boldsymbol{x}, \boldsymbol{y} \in \operatorname{dom} f$ と $\alpha \in[0,1]$に対して成立するとそれは強凸関数
* なぜ？
	* $\frac{1}{2} \sigma \alpha(1-\alpha)\|\boldsymbol{x}-\boldsymbol{y}\|^2$　が意味するものとは何？
	* 可視化して考えてみましょう。
	*  $\frac{1}{2} \sigma \alpha(1-\alpha)\|\boldsymbol{x}-\boldsymbol{y}\|^2$は、**任意の線形補間(グラフを横切る線分)からグラフの底までの距離**と考えればよく、強凸関数は必ずこの距離が0以上(超過？)で存在しています。
* 強凸関数は狭義凸関数でです。
* 適切であるかはわからないが、強凸関数$\subset$狭義凸関数$\subset$凸関数かな？
#### 凸関数と超平面
* 凸集合の任意の内点に対しては、支持超平面が存在したはずなので、凸関数のエピグラフに対しても支持超平面が存在しています。

$f: \mathbb{R}^n \rightarrow(-\infty,+\infty]$ を定義域 $\operatorname{dom} f$ が開凸集合であるよ うな関数とします。 $f$ が $\operatorname{dom} f$ において微分可能であるとき，$f$ が（狭義）凸関数であるための必要十分条件は，$x \neq y$ であるような任意の $\boldsymbol{x}, \boldsymbol{y} \in \operatorname{dom} f$ に対してつぎの不等式が成り立つことです．

$$
f(y)-f(x) \geqq(>)\langle\nabla f(x), y-x\rangle
$$
#### 凸性とヘッセ行列
* 関数の定義域内の各点で、それを入力とする関数の二次勾配、つまりヘッセ行列が半正定値なら、関数は凸関数です。


### 共役関数
真凸関数に対して共役関数を定義します。
$f: \mathbb{R}^n \rightarrow(-\infty,+\infty]$ に対して、$f^*(\xi)=\sup \left\{\langle\boldsymbol{x}, \boldsymbol{\xi}\rangle-f(\boldsymbol{x}) \mid \boldsymbol{x} \in \mathbb{R}^n\right\}$を共役関数と言います。
幾何学的解釈は図2.15に関する説明を見ていただければわかります。

#### 共役関数と関係が深い関数

**閉真凸関数**
下半連続な真凸関数

**アフィン関数**
あるベクトル $\xi \in \mathbb{R}^n$ とスカラー $\beta \in \mathbb{R}$ を用いて
$$
h(\boldsymbol{x})=\langle\boldsymbol{x}, \boldsymbol{\xi}\rangle+\beta
$$
とあらわされる関数

#### 共役関数の性質
* 真凸関数の共役関数は**閉**真凸関数(性質1)
* 真凸関数が強凸関数なら、その共役関数の定義域は真凸関数と**同じ**
* 性質1より、共役関数の共役関数(**双共役関数**)$f^{* *}(x)=\sup \left\{\langle x, \xi\rangle-f^*(\boldsymbol{\xi}) \mid \xi \in \mathbb{R}^n\right\}$は閉真凸関数
