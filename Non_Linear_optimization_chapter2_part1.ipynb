#### 分離超平面

超平面が分ける2つの空間(超平面によって定義される半空間)を考えます。
$$
\begin{aligned}
H^{+} & =\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle a, x\rangle \leqq \alpha\right\} \\
H^{-} & =\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle a, \boldsymbol{x}\rangle \leqq \alpha\right\}
\end{aligned}
$$
半空間にそれぞれ属する集合 $S, T \subseteq \mathbb{R}^n$ が $S \subseteq H^{+}$かつ $T \subseteq H^{-}$を満たすとき，すなわち
$$
\begin{array}{ll}
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \leqq \alpha & (\boldsymbol{z} \in S) \\
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \leqq \alpha & (\boldsymbol{z} \in T)
\end{array}
$$
超平面は$S$、$T$を分離するといい、分離超平面と呼ばれます。

**定理**
凸集合$S$および、その凸包に属さない点$x$について、以下の式を満たす$S$ と $x$ の分離超平面 $H=\left\{\boldsymbol{x} \in \mathbb{R}^n \mid\langle\boldsymbol{a}, \boldsymbol{x}\rangle=\alpha\right\}$が存在します
。
$$
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \geqq \alpha>\langle\boldsymbol{a}, \boldsymbol{x}\rangle \quad(\boldsymbol{z} \in S)
$$

* $z$の内積 $\geq \alpha$
*  $x$の内積$< \alpha$
 $S$の点（$z$）は $H$ より上
 ↑
 ｜   ● $z ∈ S$
 ｜   
 ｜――――――――――――――――――― 超平面 $H$（内積 $= α$）
 ｜   
 ｜          ● $x$（外の点）は$H$より下
 ↓

#### 支持超平面
凸包に属する$x$が超平面にも属している場合、それを支持超平面といいます。
****
凸集合 $S \subseteq \mathbb{R}^n$ が超平面 $H$ によって定義される半空間 $H^{+}$または $H^{-}$ に含まれ，さらに点 $\boldsymbol{x} \in \mathrm{cl} S$ が $H$ に含まれるとき，$H$ を $\boldsymbol{x}$ における $S$ の支持超平面（supporting hyperplane）といいます。

#### 支持超平面定理3連発

**定理**
凸集合の境界の点はどこの点でも支持超平面を持ちます。←劣勾配(超平面を勾配をとりたい点に近づけていく)が存在するために凸性が必要な理由です。
****
空でない凸集合 $S \subseteq \mathbb{R}^n$ は，その境界の任意の点において支持超平面をもちます．すなわち， $\boldsymbol{x} \in \operatorname{bd} S$ ならば

$$
\langle\boldsymbol{a}, \boldsymbol{z}\rangle \geqq \alpha=\langle\boldsymbol{a}, \boldsymbol{x}\rangle \quad(\boldsymbol{z} \in S)
$$
を満たす $a \in \mathbb{R}^n(a \neq 0)$ と $\alpha \in \mathbb{R}$ が存在します．
****

**定理**
任意の凸集合と、それに属さない点に対してどの点でも凸集合と分離できる分離超平面があります。
****
任意の空でない凸集合 $S \subseteq \mathbb{R}^n$ と任意の点 $\boldsymbol{x} \notin S$ に対して，$S$ と $x$ の分離超平面が存在します．
****

**定理**
凸集合同士が交わっていない場合、その凸集合同士を分離できる超平面があります。

****
空でない二つの凸集合 $S, T \subseteq \mathbb{R}^n$ の共通集合 $S \cap T$ が空集合 であれば，$S$ と $T$ の分離超平面が存在します．
****

### 平均値、テイラー、陰関数定理

**平均値の定理**
関数の差分を内積で表現できます。これをつかうと、関数の勾配に関する一階条件を導出できます。
****
［平均値定理（mean value theorem）］関数 $f: \mathbb{R}^n \rightarrow \mathbb{R}$ は凸集合 $S \subseteq \mathbb{R}^n$ において微分可能とすします．そのとき，任意の $\boldsymbol{x}, \boldsymbol{y} \in S$ に対して

$$
f(\boldsymbol{x})-f(\boldsymbol{y})=\langle\nabla f(\tau \boldsymbol{x}+(1-\tau) \boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\rangle
$$


を満足する実数 $\tau \in(0,1)$ が存在します．
****
**テイラーの定理**
関数の差分を内積で表現できます。対象が2回微分可能かどうかが平均値の定理との違いです。
****
関数 $f: \mathbb{R}^n \rightarrow \mathbb{R}$ は凸集合 $S \subseteq \mathbb{R}^n$ において 2 回連続的微分可能とすします．そのとき，任意の $\boldsymbol{x}, \boldsymbol{y} \in S$ に対して次式を満足する実数 $\tau \in(0,1)$ が存在します。
$$
f(\boldsymbol{x})=f(\boldsymbol{y})+\langle\nabla f(\boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\rangle+\frac{1}{2}\left\langle\nabla^2 f(\tau \boldsymbol{x}+(1-\tau) \boldsymbol{y})(\boldsymbol{x}-\boldsymbol{y}), \boldsymbol{x}-\boldsymbol{y}\right\rangle
$$
****


**陰関数定理**
関数の一部の定義域をみると、そこだけ別の関数で表現できることがあります。ラグランジュ、最適性条件、双対性とかにでてきます。
****

関数 $f_i: \mathbb{R}^{n+p} \rightarrow \mathbb{R}(i=1, \ldots, n)$ が与えられており，
ある点 $\bar{x} \in \mathbb{R}^n, \bar{u} \in \mathbb{R}^p$ において，

$$
f_i(\bar{x}, \bar{u})=0 \quad(i=1, \ldots, n)
$$


が成り立っているとします。
また，各 $f_i$ は点 $(\bar{x}, \bar{u})$ のある近傍で連続的に微分可能であり，
次のヤコビ行列（ $\boldsymbol{x}$ に関する偏導関数からなる行列）：

$$
\nabla_x \boldsymbol{f}(\boldsymbol{x}, \boldsymbol{u})=\left[\begin{array}{ccc}
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_n}{\partial x_1} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_1}{\partial x_n} & \cdots & \frac{\partial f_n}{\partial x_n}
\end{array}\right] \in \mathbb{R}^{n \times n}
$$


が点 $(\bar{x}, \bar{u})$ において正則（逆行列を持つ）であるとします。

そのとき ：
点 $\bar{u}$ の適当な近傍 $U \subset \mathbb{R}^p$ において，
連続的微分可能な関数

$$
\phi_i: U \rightarrow \mathbb{R} \quad(i=1, \ldots, n)
$$


が存在して，

$$
f_i\left(\phi_1(u), \ldots, \phi_n(u), u\right)=0 \quad \text { for all } u \in U
$$


が成り立ちます。
また，$\phi_i(\bar{u})=\bar{x}_i$ となります。
****
